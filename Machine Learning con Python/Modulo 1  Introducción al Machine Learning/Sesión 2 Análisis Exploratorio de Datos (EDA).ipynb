## 📚 Parte 1: Fundamentos Teóricos del EDA

### 🎯 ¿Qué es el Análisis Exploratorio de Datos (EDA)?

Definición: El EDA es el proceso sistemático de investigar, resumir y visualizar datos para descubrir patrones, detectar anomalías, probar hipótesis y verificar supuestos usando estadísticas descriptivas y métodos gráficos.

"El EDA es como ser detective de datos: buscamos pistas, patrones y evidencia antes de llegar a conclusiones"

## 🔍 Importancia del EDA en Machine Learning

### **1. Fundamento para Decisiones Informadas 🧠**

+ Selección de algoritmos: Diferentes tipos de datos requieren diferentes enfoques
+ Feature engineering: Identificar qué transformaciones son necesarias
+ Preprocesamiento: Decidir estrategias de limpieza y normalización
+ Evaluación: Elegir métricas apropiadas según el contexto

Ejemplo práctico:

Si descubrimos que nuestro dataset tiene:
+ Distribuciones muy asimétricas → Necesitamos transformaciones log
+ Alta correlación entre features → Consideramos PCA o regularización
+ Clases desbalanceadas → Usamos métricas como F1-score en lugar de accuracy
+ Outliers extremos → Aplicamos métodos robustos como SVM

### **2. Prevención de Errores Costosos ⚠️**

+ Garbage In, Garbage Out: Datos de mala calidad = modelos inútiles
+ Sesgos ocultos: Detectar problemas antes de que contaminen el modelo
+ Supuestos incorrectos: Verificar que nuestras asunciones son válidas
+ Problemas de generalización: Identificar si el dataset es representativo

### **Casos reales de problemas evitados por EDA:**

+ Dataset médico: EDA reveló que todos los casos "positivos" venían del mismo hospital → sesgo de muestreo
Predicción financiera: EDA mostró que datos de entrenamiento solo incluían mercados alcistas → modelo inútil en recesión
+ Reconocimiento de imágenes: EDA detectó que todas las fotos de "gatos" tenían marca de agua → modelo aprendió la marca, no los gatos

### **3. Optimización del Rendimiento del Modelo 🚀**

+ Eficiencia computacional: Eliminar features redundantes
+ Mejor generalización: Entender la distribución real de los datos
+ Hiperparámetros informados: EDA guía la selección inicial de parámetros
+ Feature selection inteligente: Priorizar características más informativas

## **📊 Técnicas de Visualización para Entender los Datos**

### 1. **Visualizaciones Univariadas **📈

Analizan una variable a la vez para entender su distribución y características.

__A. Histogramas__ 📊

- Propósito: Ver la distribución de una variable numérica
- Qué detecta: Sesgo, multimodalidad, outliers
- Cuándo usar: Variables continuas o discretas con muchos valores

__B. Box Plots 📦__

- Propósito: Resumir distribución y detectar outliers
- Qué detecta: Cuartiles, mediana, valores atípicos
- Ventaja: Comparación fácil entre grupos

**C. Gráficos de Densidad 🌊**

- Propósito: Versión suave del histograma
- Ventaja: Mejor para comparar distribuciones superpuestas
- Ideal para: Comparar grupos (benign vs malignant)

### **2. Visualizaciones Bivariadas 🎯**

Analizan relaciones entre dos variables.

__A. Scatter Plots__ 🎯

- Propósito: Relaciones entre variables numéricas
- Qué detecta: Correlación, patrones no lineales, agrupaciones
- Cuándo usar: Explorar relaciones lineales y no lineales

__B. Heatmaps de Correlación__ 🔥

- Propósito: Visualizar correlaciones múltiples simultáneamente
- Ventaja: Identificar multicolinealidad rápidamente
- Interpretación: Colores intensos = correlaciones fuertes

### 3. **Visualizaciones Multivariadas**🌐

Analizan múltiples variables simultáneamente.

__A. Pair Plots__ 👫

- Propósito: Todas las combinaciones de scatter plots
- Ventaja: Vista completa de relaciones
- Limitación: Solo práctico para <10 variables

**B. Parallel Coordinates** 🌈

- Propósito: Visualizar patrones en alta dimensionalidad
- Cada línea: Un punto de datos
- Utilidad: Identificar clusters y outliers multidimensionales

## 🕵️ **Detección de Valores Atípicos y Datos Faltantes**

### 1. Tipos de Valores Atípicos ⚠️

__A. Outliers Univariados__ 📊

Valores extremos en una sola variable.

__B. Outliers Multivariados__ 🎯

Puntos que son normales individualmente pero anómalos en combinación.

### 2. Estrategias para Manejar Outliers 🛠️

__A. Investigación Primero__🔍

Preguntas clave:

- 1. ¿Es un error de medición/captura?
- 2. ¿Representa un caso real pero raro?
- 3. ¿Es relevante para nuestro problema?
- 4. ¿Afecta significativamente al modelo?

__B. Opciones de Tratamiento__ ⚡

+ Eliminar: Si son errores claros
+ Transformar: Log, sqrt para reducir impacto
+ Winsorizing: Reemplazar por percentiles
+ Mantener: Si representan casos importantes


### **3. Detección de Datos Faltantes **❓

__A. Tipos de Missing Data__ 📋

- MCAR (Missing Completely at Random): Aleatoria, sin patrón
- MAR (Missing at Random): Depende de otras variables observadas
- MNAR (Missing Not at Random): Depende del valor faltante mismo

__B. Visualización de Patrones Faltantes__

__C. Estrategias de Imputación 🔧__

- Simple: Media, mediana, moda
- Avanzada: Regresión, KNN, modelos específicos
- Domain-specific: Basada en conocimiento del dominio

## 🔗 Análisis de Correlaciones entre Variables

1. **Tipos de Correlación** 📊

__A. Correlación de Pearson__📈

>+ Mide: Relaciones lineales
>+ Rango: -1 a +1

Interpretación:

- |r| > 0.8: Correlación muy fuerte
- |r| 0.6-0.8: Correlación fuerte
- |r| 0.4-0.6: Correlación moderada
- |r| < 0.4: Correlación débil

__B. Correlación de Spearman__ 📉

+ Mide: Relaciones monotónicas (no necesariamente lineales)
+ Ventaja: Robusta ante outliers
+ Cuándo usar: Relaciones no lineales o datos ordinales

## 2. Problemas de Multicolinealidad ⚠️

### ¿Qué es la multicolinealidad?

Cuando variables predictoras están altamente correlacionadas entre sí.
Problemas que causa:

+ Inestabilidad: Pequeños cambios en datos causan grandes cambios en coeficientes
+ Interpretabilidad: Difícil determinar el efecto individual de cada variable
+ Overfitting: El modelo puede volverse demasiado específico

__Soluciones:__

+ Eliminar variables: Quitar una de las variables correlacionadas
+ Combinación: Crear índices o ratios
+ Regularización: L1 (Lasso) o L2 (Ridge)
+ PCA: Reducción de dimensionalidad

##3. Interpretación Avanzada de Correlaciones 🧠

__A. Correlación ≠ Causalidad__⚠️

Ejemplo clásico:

 Correlación alta entre:
 - Ventas de helados
 - Casos de ahogamiento

 ¿El helado causa ahogamientos? NO
 Variable oculta: Temperatura alta

__ B. Correlaciones Espurias__🎭

Correlaciones que aparecen por casualidad o por variables confusoras.

__C. Correlaciones No Lineales__ 🌊

Ejemplo: Relación en U

- Correlación de Pearson = 0 (parece no correlacionado)
- Pero hay relación fuerte no lineal
- Solución: Usar Spearman o análisis visual

##  📋 Checklist del EDA Sistemático

__1. Fase de Reconocimiento 🔍__

 + Dimensiones del dataset (filas, columnas)
 + Tipos de datos por variable
 + Primeras y últimas filas
 + Información general (.info(), .describe())

__2. Fase de Calidad 🧹__

 + Valores faltantes por variable
 + Valores duplicados
 + Outliers univariados
 + Outliers multivariados
 + Consistencia de tipos de datos

__3. Fase de Distribuciones 📊__

 + Histogramas de variables numéricas
 + Conteos de variables categóricas
 + Detección de asimetría y curtosis
 + Identificación de distribuciones

__4. Fase de Relaciones 🔗__

 + Matriz de correlación completa
 + Scatter plots de pares importantes
 + Análisis de la variable objetivo
 + Identificación de multicolinealidad

__5. Fase de Insights 💡__

 + Patrones principales identificados
 + Problemas potenciales documentados
 + Recomendaciones para preprocesamiento
 + Estrategia de modelado sugerida

## 📊 Parte 1: Importación de Librerías y Carga de Datos
### 🔧 Configuración del Entorno

```python
# Importar librerías esenciales
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer

# Configuraciones para mejores visualizaciones
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 12

# Suprimir warnings para output más limpio
import warnings
warnings.filterwarnings('ignore')

print("✅ Librerías importadas correctamente")
print("📊 Configuración de visualización establecida")
```

## 📥 Carga del Dataset

```python
# Cargar el dataset Breast Cancer Wisconsin
cancer = load_breast_cancer()

# Explorar la estructura básica del dataset
print("🔍 INFORMACIÓN BÁSICA DEL DATASET")
print("=" * 50)
print(f"📊 Dimensiones de los datos: {cancer.data.shape}")
print(f"🎯 Número de clases: {len(cancer.target_names)}")
print(f"📋 Clases disponibles: {cancer.target_names}")
print(f"🔢 Número de características: {len(cancer.feature_names)}")

# Verificar el tipo de objeto
print(f"\n📦 Tipo de dataset: {type(cancer)}")
print(f"🔑 Claves disponibles: {cancer.keys()}")
```

## 📋 Creación del DataFrame

```python
# Crear DataFrame para análisis más intuitivo
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)

# Agregar la columna target
df['target'] = cancer.target

# Agregar etiquetas descriptivas
df['diagnosis'] = df['target'].map({0: 'malignant', 1: 'benign'})

# Mostrar información básica del DataFrame
print("📊 ESTRUCTURA DEL DATAFRAME")
print("=" * 40)
print(f"📏 Dimensiones: {df.shape}")
print(f"💾 Memoria utilizada: {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"🔢 Tipos de datos únicos: {df.dtypes.value_counts().to_dict()}")

# Primeras 3 filas para verificar
print("\n👀 PRIMERAS 3 FILAS:")
print(df.head(3))
```

## 🔍 Parte 2: Exploración Inicial de Datos

### 📈 **Análisis de la Variable Objetivo**

```python
# Distribución de clases
print("🎯 ANÁLISIS DE LA VARIABLE OBJETIVO")
print("=" * 45)

# Conteo absoluto
target_counts = df['diagnosis'].value_counts()
print("📊 Distribución de diagnósticos:")
for diagnosis, count in target_counts.items():
    percentage = (count / len(df)) * 100
    print(f"   {diagnosis.capitalize()}: {count} pacientes ({percentage:.1f}%)")

# Verificar balance de clases
ratio = target_counts.max() / target_counts.min()
print(f"\n⚖️ Ratio de desbalance: {ratio:.2f}:1")
if ratio > 2:
    print("⚠️ Dataset moderadamente desbalanceado")
else:
    print("✅ Dataset relativamente balanceado")

# Crear visualización
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico de barras
target_counts.plot(kind='bar', ax=ax1, color=['#ff7f7f', '#7fbf7f'])
ax1.set_title('📊 Distribución de Diagnósticos', fontsize=14, fontweight='bold')
ax1.set_xlabel('Diagnóstico')
ax1.set_ylabel('Número de Pacientes')
ax1.tick_params(axis='x', rotation=0)

# Gráfico de pastel
ax2.pie(target_counts.values, labels=target_counts.index, autopct='%1.1f%%', 
        colors=['#ff7f7f', '#7fbf7f'], startangle=90)
ax2.set_title('🥧 Proporción de Diagnósticos', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 📋 **Información General del Dataset**

```python
# Información detallada del dataset
print("📋 INFORMACIÓN DETALLADA DEL DATASET")
print("=" * 50)

# Información básica
print("🔍 Info general:")
df.info()

print("\n📊 Estadísticas descriptivas (primeras 5 características):")
# Mostrar estadísticas de las primeras 5 features para no sobrecargar
stats = df.iloc[:, :5].describe()
print(stats)

# Verificar valores faltantes
print("\n❓ VALORES FALTANTES:")
missing_values = df.isnull().sum()
if missing_values.sum() == 0:
    print("✅ No hay valores faltantes en el dataset")
else:
    print("⚠️ Valores faltantes encontrados:")
    for col, missing in missing_values[missing_values > 0].items():
        print(f"   {col}: {missing} valores ({missing/len(df)*100:.1f}%)")
```

### 🏷️ **Análisis de Características (Features)**

```python
# Análizar los nombres de las características
print("🏷️ ANÁLISIS DE CARACTERÍSTICAS")
print("=" * 40)

feature_names = cancer.feature_names
print(f"📝 Total de características: {len(feature_names)}")

# Agrupar características por tipo de medición
mean_features = [f for f in feature_names if 'mean' in f]
se_features = [f for f in feature_names if 'error' in f]
worst_features = [f for f in feature_names if 'worst' in f]

print(f"\n📊 Características por tipo de medición:")
print(f"   📈 Medias (mean): {len(mean_features)} características")
print(f"   📉 Errores estándar (SE): {len(se_features)} características")
print(f"   ⚠️ Peores valores (worst): {len(worst_features)} características")

# Mostrar ejemplos de cada tipo
print(f"\n🔍 Ejemplos de características:")
print(f"   Medias: {mean_features[:3]} ...")
print(f"   Errores: {se_features[:3]} ...")
print(f"   Peores: {worst_features[:3]} ...")

# Características únicas (sin prefijos)
unique_measurements = list(set([f.replace('mean ', '').replace(' error', '').replace('worst ', '') for f in feature_names]))
print(f"\n📏 Tipos únicos de mediciones: {len(unique_measurements)}")
print(f"   {unique_measurements}")
```

## 📊 Parte 3: Análisis Estadístico Profundo

### 📈 **Estadísticas por Grupo de Diagnóstico**

```python
# Comparar estadísticas entre grupos
print("📊 COMPARACIÓN ESTADÍSTICA POR DIAGNÓSTICO")
print("=" * 55)

# Seleccionar características principales (mean features)
main_features = [f for f in df.columns if 'mean' in f]

# Estadísticas por diagnóstico
for diagnosis in ['malignant', 'benign']:
    subset = df[df['diagnosis'] == diagnosis]
    print(f"\n🔬 Estadísticas para casos {diagnosis.upper()}:")
    print(f"   📊 Número de casos: {len(subset)}")
    
    # Mostrar estadísticas de las primeras 5 características principales
    stats = subset[main_features[:5]].describe().loc[['mean', 'std', 'min', 'max']]
    print(stats.round(2))

# Calcular diferencias entre grupos
print("\n🔍 DIFERENCIAS ENTRE GRUPOS:")
malignant_data = df[df['diagnosis'] == 'malignant'][main_features]
benign_data = df[df['diagnosis'] == 'benign'][main_features]

differences = malignant_data.mean() - benign_data.mean()
print("📈 Diferencias en promedios (Maligno - Benigno):")
for feature, diff in differences.head(10).items():
    direction = "↑" if diff > 0 else "↓"
    print(f"   {feature}: {diff:.2f} {direction}")
```

### 📊 **Análisis de Correlaciones**

```python
# Matriz de correlación para características principales
print("🔗 ANÁLISIS DE CORRELACIONES")
print("=" * 35)

# Calcular correlación entre características principales
correlation_matrix = df[main_features].corr()

# Crear heatmap de correlaciones
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, cbar_kws={"shrink": .8}, fmt='.2f')
plt.title('🔗 Matriz de Correlación - Características Principales', 
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Identificar correlaciones fuertes
print("🔍 Correlaciones más fuertes (> 0.8):")
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) > 0.8:
            feature1 = correlation_matrix.columns[i]
            feature2 = correlation_matrix.columns[j]
            high_corr_pairs.append((feature1, feature2, corr_value))

for feat1, feat2, corr in sorted(high_corr_pairs, key=lambda x: abs(x[2]), reverse=True)[:5]:
    print(f"   {feat1} ↔ {feat2}: {corr:.3f}")
```

## 📊 Parte 4: Visualizaciones Exploratorias

### 📈 **Distribuciones de Características Clave**

```python
# Visualizar distribuciones de características más importantes
print("📈 VISUALIZACIÓN DE DISTRIBUCIONES")
print("=" * 40)

# Seleccionar 6 características clave para análisis
key_features = [
    'mean radius', 'mean texture', 'mean perimeter',
    'mean area', 'mean smoothness', 'mean compactness'
]

# Crear subplots para distribuciones
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for i, feature in enumerate(key_features):
    # Histograma por diagnóstico
    for diagnosis in ['malignant', 'benign']:
        data = df[df['diagnosis'] == diagnosis][feature]
        axes[i].hist(data, alpha=0.7, label=diagnosis, bins=20, density=True)
    
    axes[i].set_title(f'📊 Distribución: {feature.title()}', fontweight='bold')
    axes[i].set_xlabel(feature.replace('mean ', '').title())
    axes[i].set_ylabel('Densidad')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.suptitle('📈 Distribuciones de Características Clave por Diagnóstico', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Análisis estadístico de separabilidad
print("\n🔍 Análisis de separabilidad de características:")
for feature in key_features:
    malignant_vals = df[df['diagnosis'] == 'malignant'][feature]
    benign_vals = df[df['diagnosis'] == 'benign'][feature]
    
    # Calcular solapamiento aproximado
    overlap = len(set(malignant_vals.round(1)) & set(benign_vals.round(1)))
    total_unique = len(set(malignant_vals.round(1)) | set(benign_vals.round(1)))
    separation_score = 1 - (overlap / total_unique)
    
    print(f"   {feature}: Separabilidad = {separation_score:.2f}")
```

### 📊 **Box Plots para Comparación**

```python
# Box plots para comparar distribuciones
print("📦 ANÁLISIS CON BOX PLOTS")
print("=" * 30)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for i, feature in enumerate(key_features):
    # Box plot por diagnóstico
    df.boxplot(column=feature, by='diagnosis', ax=axes[i])
    axes[i].set_title(f'📦 {feature.title()}')
    axes[i].set_xlabel('Diagnóstico')
    axes[i].set_ylabel(feature.replace('mean ', '').title())
    
    # Remover título automático de pandas
    axes[i].set_title(f'📦 {feature.title()}')

plt.suptitle('📦 Box Plots: Características por Diagnóstico', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Identificar outliers potenciales
print("⚠️ Detección de valores atípicos:")
for feature in key_features[:3]:  # Analizar solo 3 para no saturar
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    print(f"   {feature}: {len(outliers)} valores atípicos ({len(outliers)/len(df)*100:.1f}%)")
```

### 🎯 **Scatter Plots para Relaciones**

```python
# Scatter plots para visualizar relaciones entre características
print("🎯 ANÁLISIS DE RELACIONES ENTRE CARACTERÍSTICAS")
print("=" * 55)

# Seleccionar pares de características interesantes
feature_pairs = [
    ('mean radius', 'mean area'),
    ('mean perimeter', 'mean area'),
    ('mean compactness', 'mean concavity')
]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, (feat1, feat2) in enumerate(feature_pairs):
    # Scatter plot por diagnóstico
    for diagnosis in ['malignant', 'benign']:
        subset = df[df['diagnosis'] == diagnosis]
        axes[i].scatter(subset[feat1], subset[feat2], 
                       label=diagnosis, alpha=0.6, s=50)
    
    axes[i].set_xlabel(feat1.replace('mean ', '').title())
    axes[i].set_ylabel(feat2.replace('mean ', '').title())
    axes[i].set_title(f'🎯 {feat1.title()} vs {feat2.title()}')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.suptitle('🎯 Relaciones entre Características Clave', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Calcular correlaciones específicas
print("🔗 Correlaciones entre pares analizados:")
for feat1, feat2 in feature_pairs:
    correlation = df[feat1].corr(df[feat2])
    print(f"   {feat1} ↔ {feat2}: {correlation:.3f}")
```

### 🔍 **Resumen de Hallazgos**

```python
print("🔍 RESUMEN DE HALLAZGOS DEL EDA")
print("=" * 45)

# 1. Balance del dataset
target_counts = df['diagnosis'].value_counts()
balance_ratio = target_counts.max() / target_counts.min()
print(f"1️⃣ BALANCE DEL DATASET:")
print(f"   • Casos benignos: {target_counts['benign']} ({target_counts['benign']/len(df)*100:.1f}%)")
print(f"   • Casos malignos: {target_counts['malignant']} ({target_counts['malignant']/len(df)*100:.1f}%)")
print(f"   • Ratio de balance: {balance_ratio:.2f}:1 {'✅ Balanceado' if balance_ratio < 2 else '⚠️ Desbalanceado'}")

# 2. Calidad de los datos
print(f"\n2️⃣ CALIDAD DE LOS DATOS:")
print(f"   • Valores faltantes: {df.isnull().sum().sum()} ✅")
print(f"   • Tipos de datos: Todos numéricos ✅")
print(f"   • Dimensionalidad: {df.shape[1]-2} características, {df.shape[0]} muestras")

# 3. Características más discriminativas
malignant_means = df[df['diagnosis'] == 'malignant'][main_features].mean()
benign_means = df[df['diagnosis'] == 'benign'][main_features].mean()
differences = (malignant_means - benign_means).abs().sort_values(ascending=False)

print(f"\n3️⃣ CARACTERÍSTICAS MÁS DISCRIMINATIVAS:")
for i, (feature, diff) in enumerate(differences.head(5).items(), 1):
    print(f"   {i}. {feature}: Diferencia = {diff:.2f}")

# 4. Correlaciones importantes
correlation_matrix = df[main_features].corr()
high_corr_count = (correlation_matrix.abs() > 0.8).sum().sum() - len(main_features)
print(f"\n4️⃣ ESTRUCTURA DE CORRELACIONES:")
print(f"   • Correlaciones altas (>0.8): {high_corr_count//2} pares")
print(f"   • Potencial multicolinealidad: {'⚠️ Presente' if high_corr_count > 20 else '✅ Baja'}")

# 5. Separabilidad de clases
print(f"\n5️⃣ SEPARABILIDAD DE CLASES:")
separable_features = []
for feature in main_features[:5]:
    malignant_vals = df[df['diagnosis'] == 'malignant'][feature]
    benign_vals = df[df['diagnosis'] == 'benign'][feature]
    
    # Test simple de separabilidad
    malignant_max = malignant_vals.min()
    benign_min = benign_vals.max()
    
    if malignant_max > benign_min or benign_vals.min() > malignant_vals.max():
        separable_features.append(feature)

print(f"   • Características con buena separación: {len(separable_features)}")
print(f"   • Predictibilidad esperada: {'🎯 Alta' if len(separable_features) > 2 else '📊 Moderada'}")
```
