## üìö Parte 1: Fundamentos Te√≥ricos del EDA

### üéØ ¬øQu√© es el An√°lisis Exploratorio de Datos (EDA)?

Definici√≥n: El EDA es el proceso sistem√°tico de investigar, resumir y visualizar datos para descubrir patrones, detectar anomal√≠as, probar hip√≥tesis y verificar supuestos usando estad√≠sticas descriptivas y m√©todos gr√°ficos.

"El EDA es como ser detective de datos: buscamos pistas, patrones y evidencia antes de llegar a conclusiones"

## üîç Importancia del EDA en Machine Learning

### **1. Fundamento para Decisiones Informadas üß†**

+ Selecci√≥n de algoritmos: Diferentes tipos de datos requieren diferentes enfoques
+ Feature engineering: Identificar qu√© transformaciones son necesarias
+ Preprocesamiento: Decidir estrategias de limpieza y normalizaci√≥n
+ Evaluaci√≥n: Elegir m√©tricas apropiadas seg√∫n el contexto

Ejemplo pr√°ctico:

Si descubrimos que nuestro dataset tiene:
+ Distribuciones muy asim√©tricas ‚Üí Necesitamos transformaciones log
+ Alta correlaci√≥n entre features ‚Üí Consideramos PCA o regularizaci√≥n
+ Clases desbalanceadas ‚Üí Usamos m√©tricas como F1-score en lugar de accuracy
+ Outliers extremos ‚Üí Aplicamos m√©todos robustos como SVM

### **2. Prevenci√≥n de Errores Costosos ‚ö†Ô∏è**

+ Garbage In, Garbage Out: Datos de mala calidad = modelos in√∫tiles
+ Sesgos ocultos: Detectar problemas antes de que contaminen el modelo
+ Supuestos incorrectos: Verificar que nuestras asunciones son v√°lidas
+ Problemas de generalizaci√≥n: Identificar si el dataset es representativo

### **Casos reales de problemas evitados por EDA:**

+ Dataset m√©dico: EDA revel√≥ que todos los casos "positivos" ven√≠an del mismo hospital ‚Üí sesgo de muestreo
Predicci√≥n financiera: EDA mostr√≥ que datos de entrenamiento solo inclu√≠an mercados alcistas ‚Üí modelo in√∫til en recesi√≥n
+ Reconocimiento de im√°genes: EDA detect√≥ que todas las fotos de "gatos" ten√≠an marca de agua ‚Üí modelo aprendi√≥ la marca, no los gatos

### **3. Optimizaci√≥n del Rendimiento del Modelo üöÄ**

+ Eficiencia computacional: Eliminar features redundantes
+ Mejor generalizaci√≥n: Entender la distribuci√≥n real de los datos
+ Hiperpar√°metros informados: EDA gu√≠a la selecci√≥n inicial de par√°metros
+ Feature selection inteligente: Priorizar caracter√≠sticas m√°s informativas

## **üìä T√©cnicas de Visualizaci√≥n para Entender los Datos**

### 1. **Visualizaciones Univariadas **üìà

Analizan una variable a la vez para entender su distribuci√≥n y caracter√≠sticas.

__A. Histogramas__ üìä

- Prop√≥sito: Ver la distribuci√≥n de una variable num√©rica
- Qu√© detecta: Sesgo, multimodalidad, outliers
- Cu√°ndo usar: Variables continuas o discretas con muchos valores

__B. Box Plots üì¶__

- Prop√≥sito: Resumir distribuci√≥n y detectar outliers
- Qu√© detecta: Cuartiles, mediana, valores at√≠picos
- Ventaja: Comparaci√≥n f√°cil entre grupos

**C. Gr√°ficos de Densidad üåä**

- Prop√≥sito: Versi√≥n suave del histograma
- Ventaja: Mejor para comparar distribuciones superpuestas
- Ideal para: Comparar grupos (benign vs malignant)

### **2. Visualizaciones Bivariadas üéØ**

Analizan relaciones entre dos variables.

__A. Scatter Plots__ üéØ

- Prop√≥sito: Relaciones entre variables num√©ricas
- Qu√© detecta: Correlaci√≥n, patrones no lineales, agrupaciones
- Cu√°ndo usar: Explorar relaciones lineales y no lineales

__B. Heatmaps de Correlaci√≥n__ üî•

- Prop√≥sito: Visualizar correlaciones m√∫ltiples simult√°neamente
- Ventaja: Identificar multicolinealidad r√°pidamente
- Interpretaci√≥n: Colores intensos = correlaciones fuertes

### 3. **Visualizaciones Multivariadas**üåê

Analizan m√∫ltiples variables simult√°neamente.

__A. Pair Plots__ üë´

- Prop√≥sito: Todas las combinaciones de scatter plots
- Ventaja: Vista completa de relaciones
- Limitaci√≥n: Solo pr√°ctico para <10 variables

**B. Parallel Coordinates** üåà

- Prop√≥sito: Visualizar patrones en alta dimensionalidad
- Cada l√≠nea: Un punto de datos
- Utilidad: Identificar clusters y outliers multidimensionales

## üïµÔ∏è **Detecci√≥n de Valores At√≠picos y Datos Faltantes**

### 1. Tipos de Valores At√≠picos ‚ö†Ô∏è

__A. Outliers Univariados__ üìä

Valores extremos en una sola variable.

__B. Outliers Multivariados__ üéØ

Puntos que son normales individualmente pero an√≥malos en combinaci√≥n.

### 2. Estrategias para Manejar Outliers üõ†Ô∏è

__A. Investigaci√≥n Primero__üîç

Preguntas clave:

- 1. ¬øEs un error de medici√≥n/captura?
- 2. ¬øRepresenta un caso real pero raro?
- 3. ¬øEs relevante para nuestro problema?
- 4. ¬øAfecta significativamente al modelo?

__B. Opciones de Tratamiento__ ‚ö°

+ Eliminar: Si son errores claros
+ Transformar: Log, sqrt para reducir impacto
+ Winsorizing: Reemplazar por percentiles
+ Mantener: Si representan casos importantes


### **3. Detecci√≥n de Datos Faltantes **‚ùì

__A. Tipos de Missing Data__ üìã

- MCAR (Missing Completely at Random): Aleatoria, sin patr√≥n
- MAR (Missing at Random): Depende de otras variables observadas
- MNAR (Missing Not at Random): Depende del valor faltante mismo

__B. Visualizaci√≥n de Patrones Faltantes__

__C. Estrategias de Imputaci√≥n üîß__

- Simple: Media, mediana, moda
- Avanzada: Regresi√≥n, KNN, modelos espec√≠ficos
- Domain-specific: Basada en conocimiento del dominio

## üîó An√°lisis de Correlaciones entre Variables

1. **Tipos de Correlaci√≥n** üìä

__A. Correlaci√≥n de Pearson__üìà

>+ Mide: Relaciones lineales
>+ Rango: -1 a +1

Interpretaci√≥n:

- |r| > 0.8: Correlaci√≥n muy fuerte
- |r| 0.6-0.8: Correlaci√≥n fuerte
- |r| 0.4-0.6: Correlaci√≥n moderada
- |r| < 0.4: Correlaci√≥n d√©bil

__B. Correlaci√≥n de Spearman__ üìâ

+ Mide: Relaciones monot√≥nicas (no necesariamente lineales)
+ Ventaja: Robusta ante outliers
+ Cu√°ndo usar: Relaciones no lineales o datos ordinales

## 2. Problemas de Multicolinealidad ‚ö†Ô∏è

### ¬øQu√© es la multicolinealidad?

Cuando variables predictoras est√°n altamente correlacionadas entre s√≠.
Problemas que causa:

+ Inestabilidad: Peque√±os cambios en datos causan grandes cambios en coeficientes
+ Interpretabilidad: Dif√≠cil determinar el efecto individual de cada variable
+ Overfitting: El modelo puede volverse demasiado espec√≠fico

__Soluciones:__

+ Eliminar variables: Quitar una de las variables correlacionadas
+ Combinaci√≥n: Crear √≠ndices o ratios
+ Regularizaci√≥n: L1 (Lasso) o L2 (Ridge)
+ PCA: Reducci√≥n de dimensionalidad

##3. Interpretaci√≥n Avanzada de Correlaciones üß†

__A. Correlaci√≥n ‚â† Causalidad__‚ö†Ô∏è

Ejemplo cl√°sico:

 Correlaci√≥n alta entre:
 - Ventas de helados
 - Casos de ahogamiento

 ¬øEl helado causa ahogamientos? NO
 Variable oculta: Temperatura alta

__ B. Correlaciones Espurias__üé≠

Correlaciones que aparecen por casualidad o por variables confusoras.

__C. Correlaciones No Lineales__ üåä

Ejemplo: Relaci√≥n en U

- Correlaci√≥n de Pearson = 0 (parece no correlacionado)
- Pero hay relaci√≥n fuerte no lineal
- Soluci√≥n: Usar Spearman o an√°lisis visual

##  üìã Checklist del EDA Sistem√°tico

__1. Fase de Reconocimiento üîç__

 + Dimensiones del dataset (filas, columnas)
 + Tipos de datos por variable
 + Primeras y √∫ltimas filas
 + Informaci√≥n general (.info(), .describe())

__2. Fase de Calidad üßπ__

 + Valores faltantes por variable
 + Valores duplicados
 + Outliers univariados
 + Outliers multivariados
 + Consistencia de tipos de datos

__3. Fase de Distribuciones üìä__

 + Histogramas de variables num√©ricas
 + Conteos de variables categ√≥ricas
 + Detecci√≥n de asimetr√≠a y curtosis
 + Identificaci√≥n de distribuciones

__4. Fase de Relaciones üîó__

 + Matriz de correlaci√≥n completa
 + Scatter plots de pares importantes
 + An√°lisis de la variable objetivo
 + Identificaci√≥n de multicolinealidad

__5. Fase de Insights üí°__

 + Patrones principales identificados
 + Problemas potenciales documentados
 + Recomendaciones para preprocesamiento
 + Estrategia de modelado sugerida

## üìä Parte 1: Importaci√≥n de Librer√≠as y Carga de Datos
### üîß Configuraci√≥n del Entorno

```python
# Importar librer√≠as esenciales
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer

# Configuraciones para mejores visualizaciones
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 12

# Suprimir warnings para output m√°s limpio
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Librer√≠as importadas correctamente")
print("üìä Configuraci√≥n de visualizaci√≥n establecida")
```

## üì• Carga del Dataset

```python
# Cargar el dataset Breast Cancer Wisconsin
cancer = load_breast_cancer()

# Explorar la estructura b√°sica del dataset
print("üîç INFORMACI√ìN B√ÅSICA DEL DATASET")
print("=" * 50)
print(f"üìä Dimensiones de los datos: {cancer.data.shape}")
print(f"üéØ N√∫mero de clases: {len(cancer.target_names)}")
print(f"üìã Clases disponibles: {cancer.target_names}")
print(f"üî¢ N√∫mero de caracter√≠sticas: {len(cancer.feature_names)}")

# Verificar el tipo de objeto
print(f"\nüì¶ Tipo de dataset: {type(cancer)}")
print(f"üîë Claves disponibles: {cancer.keys()}")
```

## üìã Creaci√≥n del DataFrame

```python
# Crear DataFrame para an√°lisis m√°s intuitivo
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)

# Agregar la columna target
df['target'] = cancer.target

# Agregar etiquetas descriptivas
df['diagnosis'] = df['target'].map({0: 'malignant', 1: 'benign'})

# Mostrar informaci√≥n b√°sica del DataFrame
print("üìä ESTRUCTURA DEL DATAFRAME")
print("=" * 40)
print(f"üìè Dimensiones: {df.shape}")
print(f"üíæ Memoria utilizada: {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
print(f"üî¢ Tipos de datos √∫nicos: {df.dtypes.value_counts().to_dict()}")

# Primeras 3 filas para verificar
print("\nüëÄ PRIMERAS 3 FILAS:")
print(df.head(3))
```

## üîç Parte 2: Exploraci√≥n Inicial de Datos

### üìà **An√°lisis de la Variable Objetivo**

```python
# Distribuci√≥n de clases
print("üéØ AN√ÅLISIS DE LA VARIABLE OBJETIVO")
print("=" * 45)

# Conteo absoluto
target_counts = df['diagnosis'].value_counts()
print("üìä Distribuci√≥n de diagn√≥sticos:")
for diagnosis, count in target_counts.items():
    percentage = (count / len(df)) * 100
    print(f"   {diagnosis.capitalize()}: {count} pacientes ({percentage:.1f}%)")

# Verificar balance de clases
ratio = target_counts.max() / target_counts.min()
print(f"\n‚öñÔ∏è Ratio de desbalance: {ratio:.2f}:1")
if ratio > 2:
    print("‚ö†Ô∏è Dataset moderadamente desbalanceado")
else:
    print("‚úÖ Dataset relativamente balanceado")

# Crear visualizaci√≥n
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gr√°fico de barras
target_counts.plot(kind='bar', ax=ax1, color=['#ff7f7f', '#7fbf7f'])
ax1.set_title('üìä Distribuci√≥n de Diagn√≥sticos', fontsize=14, fontweight='bold')
ax1.set_xlabel('Diagn√≥stico')
ax1.set_ylabel('N√∫mero de Pacientes')
ax1.tick_params(axis='x', rotation=0)

# Gr√°fico de pastel
ax2.pie(target_counts.values, labels=target_counts.index, autopct='%1.1f%%', 
        colors=['#ff7f7f', '#7fbf7f'], startangle=90)
ax2.set_title('ü•ß Proporci√≥n de Diagn√≥sticos', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

### üìã **Informaci√≥n General del Dataset**

```python
# Informaci√≥n detallada del dataset
print("üìã INFORMACI√ìN DETALLADA DEL DATASET")
print("=" * 50)

# Informaci√≥n b√°sica
print("üîç Info general:")
df.info()

print("\nüìä Estad√≠sticas descriptivas (primeras 5 caracter√≠sticas):")
# Mostrar estad√≠sticas de las primeras 5 features para no sobrecargar
stats = df.iloc[:, :5].describe()
print(stats)

# Verificar valores faltantes
print("\n‚ùì VALORES FALTANTES:")
missing_values = df.isnull().sum()
if missing_values.sum() == 0:
    print("‚úÖ No hay valores faltantes en el dataset")
else:
    print("‚ö†Ô∏è Valores faltantes encontrados:")
    for col, missing in missing_values[missing_values > 0].items():
        print(f"   {col}: {missing} valores ({missing/len(df)*100:.1f}%)")
```

### üè∑Ô∏è **An√°lisis de Caracter√≠sticas (Features)**

```python
# An√°lizar los nombres de las caracter√≠sticas
print("üè∑Ô∏è AN√ÅLISIS DE CARACTER√çSTICAS")
print("=" * 40)

feature_names = cancer.feature_names
print(f"üìù Total de caracter√≠sticas: {len(feature_names)}")

# Agrupar caracter√≠sticas por tipo de medici√≥n
mean_features = [f for f in feature_names if 'mean' in f]
se_features = [f for f in feature_names if 'error' in f]
worst_features = [f for f in feature_names if 'worst' in f]

print(f"\nüìä Caracter√≠sticas por tipo de medici√≥n:")
print(f"   üìà Medias (mean): {len(mean_features)} caracter√≠sticas")
print(f"   üìâ Errores est√°ndar (SE): {len(se_features)} caracter√≠sticas")
print(f"   ‚ö†Ô∏è Peores valores (worst): {len(worst_features)} caracter√≠sticas")

# Mostrar ejemplos de cada tipo
print(f"\nüîç Ejemplos de caracter√≠sticas:")
print(f"   Medias: {mean_features[:3]} ...")
print(f"   Errores: {se_features[:3]} ...")
print(f"   Peores: {worst_features[:3]} ...")

# Caracter√≠sticas √∫nicas (sin prefijos)
unique_measurements = list(set([f.replace('mean ', '').replace(' error', '').replace('worst ', '') for f in feature_names]))
print(f"\nüìè Tipos √∫nicos de mediciones: {len(unique_measurements)}")
print(f"   {unique_measurements}")
```

## üìä Parte 3: An√°lisis Estad√≠stico Profundo

### üìà **Estad√≠sticas por Grupo de Diagn√≥stico**

```python
# Comparar estad√≠sticas entre grupos
print("üìä COMPARACI√ìN ESTAD√çSTICA POR DIAGN√ìSTICO")
print("=" * 55)

# Seleccionar caracter√≠sticas principales (mean features)
main_features = [f for f in df.columns if 'mean' in f]

# Estad√≠sticas por diagn√≥stico
for diagnosis in ['malignant', 'benign']:
    subset = df[df['diagnosis'] == diagnosis]
    print(f"\nüî¨ Estad√≠sticas para casos {diagnosis.upper()}:")
    print(f"   üìä N√∫mero de casos: {len(subset)}")
    
    # Mostrar estad√≠sticas de las primeras 5 caracter√≠sticas principales
    stats = subset[main_features[:5]].describe().loc[['mean', 'std', 'min', 'max']]
    print(stats.round(2))

# Calcular diferencias entre grupos
print("\nüîç DIFERENCIAS ENTRE GRUPOS:")
malignant_data = df[df['diagnosis'] == 'malignant'][main_features]
benign_data = df[df['diagnosis'] == 'benign'][main_features]

differences = malignant_data.mean() - benign_data.mean()
print("üìà Diferencias en promedios (Maligno - Benigno):")
for feature, diff in differences.head(10).items():
    direction = "‚Üë" if diff > 0 else "‚Üì"
    print(f"   {feature}: {diff:.2f} {direction}")
```

### üìä **An√°lisis de Correlaciones**

```python
# Matriz de correlaci√≥n para caracter√≠sticas principales
print("üîó AN√ÅLISIS DE CORRELACIONES")
print("=" * 35)

# Calcular correlaci√≥n entre caracter√≠sticas principales
correlation_matrix = df[main_features].corr()

# Crear heatmap de correlaciones
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, cbar_kws={"shrink": .8}, fmt='.2f')
plt.title('üîó Matriz de Correlaci√≥n - Caracter√≠sticas Principales', 
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Identificar correlaciones fuertes
print("üîç Correlaciones m√°s fuertes (> 0.8):")
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        if abs(corr_value) > 0.8:
            feature1 = correlation_matrix.columns[i]
            feature2 = correlation_matrix.columns[j]
            high_corr_pairs.append((feature1, feature2, corr_value))

for feat1, feat2, corr in sorted(high_corr_pairs, key=lambda x: abs(x[2]), reverse=True)[:5]:
    print(f"   {feat1} ‚Üî {feat2}: {corr:.3f}")
```

## üìä Parte 4: Visualizaciones Exploratorias

### üìà **Distribuciones de Caracter√≠sticas Clave**

```python
# Visualizar distribuciones de caracter√≠sticas m√°s importantes
print("üìà VISUALIZACI√ìN DE DISTRIBUCIONES")
print("=" * 40)

# Seleccionar 6 caracter√≠sticas clave para an√°lisis
key_features = [
    'mean radius', 'mean texture', 'mean perimeter',
    'mean area', 'mean smoothness', 'mean compactness'
]

# Crear subplots para distribuciones
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for i, feature in enumerate(key_features):
    # Histograma por diagn√≥stico
    for diagnosis in ['malignant', 'benign']:
        data = df[df['diagnosis'] == diagnosis][feature]
        axes[i].hist(data, alpha=0.7, label=diagnosis, bins=20, density=True)
    
    axes[i].set_title(f'üìä Distribuci√≥n: {feature.title()}', fontweight='bold')
    axes[i].set_xlabel(feature.replace('mean ', '').title())
    axes[i].set_ylabel('Densidad')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.suptitle('üìà Distribuciones de Caracter√≠sticas Clave por Diagn√≥stico', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# An√°lisis estad√≠stico de separabilidad
print("\nüîç An√°lisis de separabilidad de caracter√≠sticas:")
for feature in key_features:
    malignant_vals = df[df['diagnosis'] == 'malignant'][feature]
    benign_vals = df[df['diagnosis'] == 'benign'][feature]
    
    # Calcular solapamiento aproximado
    overlap = len(set(malignant_vals.round(1)) & set(benign_vals.round(1)))
    total_unique = len(set(malignant_vals.round(1)) | set(benign_vals.round(1)))
    separation_score = 1 - (overlap / total_unique)
    
    print(f"   {feature}: Separabilidad = {separation_score:.2f}")
```

### üìä **Box Plots para Comparaci√≥n**

```python
# Box plots para comparar distribuciones
print("üì¶ AN√ÅLISIS CON BOX PLOTS")
print("=" * 30)

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for i, feature in enumerate(key_features):
    # Box plot por diagn√≥stico
    df.boxplot(column=feature, by='diagnosis', ax=axes[i])
    axes[i].set_title(f'üì¶ {feature.title()}')
    axes[i].set_xlabel('Diagn√≥stico')
    axes[i].set_ylabel(feature.replace('mean ', '').title())
    
    # Remover t√≠tulo autom√°tico de pandas
    axes[i].set_title(f'üì¶ {feature.title()}')

plt.suptitle('üì¶ Box Plots: Caracter√≠sticas por Diagn√≥stico', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Identificar outliers potenciales
print("‚ö†Ô∏è Detecci√≥n de valores at√≠picos:")
for feature in key_features[:3]:  # Analizar solo 3 para no saturar
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    print(f"   {feature}: {len(outliers)} valores at√≠picos ({len(outliers)/len(df)*100:.1f}%)")
```

### üéØ **Scatter Plots para Relaciones**

### üîç **Resumen de Hallazgos**

```python
print("üîç RESUMEN DE HALLAZGOS DEL EDA")
print("=" * 45)

# 1. Balance del dataset
target_counts = df['diagnosis'].value_counts()
balance_ratio = target_counts.max() / target_counts.min()
print(f"1Ô∏è‚É£ BALANCE DEL DATASET:")
print(f"   ‚Ä¢ Casos benignos: {target_counts['benign']} ({target_counts['benign']/len(df)*100:.1f}%)")
print(f"   ‚Ä¢ Casos malignos: {target_counts['malignant']} ({target_counts['malignant']/len(df)*100:.1f}%)")
print(f"   ‚Ä¢ Ratio de balance: {balance_ratio:.2f}:1 {'‚úÖ Balanceado' if balance_ratio < 2 else '‚ö†Ô∏è Desbalanceado'}")

# 2. Calidad de los datos
print(f"\n2Ô∏è‚É£ CALIDAD DE LOS DATOS:")
print(f"   ‚Ä¢ Valores faltantes: {df.isnull().sum().sum()} ‚úÖ")
print(f"   ‚Ä¢ Tipos de datos: Todos num√©ricos ‚úÖ")
print(f"   ‚Ä¢ Dimensionalidad: {df.shape[1]-2} caracter√≠sticas, {df.shape[0]} muestras")

# 3. Caracter√≠sticas m√°s discriminativas
malignant_means = df[df['diagnosis'] == 'malignant'][main_features].mean()
benign_means = df[df['diagnosis'] == 'benign'][main_features].mean()
differences = (malignant_means - benign_means).abs().sort_values(ascending=False)

print(f"\n3Ô∏è‚É£ CARACTER√çSTICAS M√ÅS DISCRIMINATIVAS:")
for i, (feature, diff) in enumerate(differences.head(5).items(), 1):
    print(f"   {i}. {feature}: Diferencia = {diff:.2f}")

# 4. Correlaciones importantes
correlation_matrix = df[main_features].corr()
high_corr_count = (correlation_matrix.abs() > 0.8).sum().sum() - len(main_features)
print(f"\n4Ô∏è‚É£ ESTRUCTURA DE CORRELACIONES:")
print(f"   ‚Ä¢ Correlaciones altas (>0.8): {high_corr_count//2} pares")
print(f"   ‚Ä¢ Potencial multicolinealidad: {'‚ö†Ô∏è Presente' if high_corr_count > 20 else '‚úÖ Baja'}")

# 5. Separabilidad de clases
print(f"\n5Ô∏è‚É£ SEPARABILIDAD DE CLASES:")
separable_features = []
for feature in main_features[:5]:
    malignant_vals = df[df['diagnosis'] == 'malignant'][feature]
    benign_vals = df[df['diagnosis'] == 'benign'][feature]
    
    # Test simple de separabilidad
    malignant_max = malignant_vals.min()
    benign_min = benign_vals.max()
    
    if malignant_max > benign_min or benign_vals.min() > malignant_vals.max():
        separable_features.append(feature)

print(f"   ‚Ä¢ Caracter√≠sticas con buena separaci√≥n: {len(separable_features)}")
print(f"   ‚Ä¢ Predictibilidad esperada: {'üéØ Alta' if len(separable_features) > 2 else 'üìä Moderada'}")
```
