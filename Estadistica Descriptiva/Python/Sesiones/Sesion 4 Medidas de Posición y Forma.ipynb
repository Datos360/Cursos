## üìä **PARTE 2: Percentiles y Cuartiles Avanzados**

### **üî¢ C√°lculo de Percentiles**

2. **Calcular posici√≥n**: `posici√≥n = (k/100) √ó (n + 1)`

#### **üí° Ejemplo Detallado:**

3. **Valores**: x‚ÇÉ = 18, x‚ÇÑ = 20  
4. **Interpolaci√≥n**: P30 = 18 + 0.3 √ó (20 - 18) = 18 + 0.6 = 18.6

### **üìà Aplicaciones Pr√°cticas de Percentiles**

#### **1. Evaluaci√≥n Acad√©mica:**
- "Est√°s en el percentil 85 en matem√°ticas"

#### **2. Medicina:**

#### **3. Finanzas:**

### **‚úÖ Ventajas de los Percentiles**

1. **Intuitivos**: F√°ciles de interpretar en contexto

### **‚ö†Ô∏è Limitaciones de los Percentiles**

## CONFIGURACI√ìN INICIAL Y PREPARACI√ìN DE DATOS

````python
# Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import skew, kurtosis, zscore
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaci√≥n
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 8)
sns.set_palette("husl")
````

````python
n_estudiantes = 200
estudiantes_df = pd.DataFrame({
        'id': range(1, n_estudiantes + 1),
        'edad': np.clip(np.random.normal(20, 2, n_estudiantes).round().astype(int), 16, 35),
        'nota_promedio': np.clip(np.random.normal(3.5, 0.8, n_estudiantes).round(2), 0, 5),
        'horas_estudio': np.clip(np.random.gamma(2, 3, n_estudiantes).round().astype(int), 0, 50),
        'genero': np.random.choice(['Masculino', 'Femenino'], n_estudiantes, p=[0.48, 0.52]),
        'carrera': np.random.choice(['Ingenier√≠a', 'Medicina', 'Derecho', 'Psicolog√≠a', 'Econom√≠a'], 
         n_estudiantes, p=[0.3, 0.2, 0.15, 0.2, 0.15]),
        'estrato': np.random.choice([1, 2, 3, 4, 5, 6], n_estudiantes, p=[0.1, 0.3, 0.3, 0.2, 0.08, 0.02])
    })

estudiantes_df.head()
````

## C√ÅLCULO DETALLADO DE PERCENTILES Y CUARTILES

````python
print("\n" + "="*70)
print("üìè AN√ÅLISIS DETALLADO DE PERCENTILES Y CUARTILES")
print("="*70)

# Tomar las notas como ejemplo principal
notas = estudiantes_df['nota_promedio'].copy()

print(f"üìä Analizando notas promedio de {len(notas)} estudiantes")
print(f"   Rango: {notas.min():.2f} - {notas.max():.2f}")

# C√°lculo manual de percentiles espec√≠ficos
percentiles_interes = [5, 10, 25, 50, 75, 90, 95]
print(f"\nüìã PERCENTILES CLAVE:")
print("-" * 40)
print(f"{'Percentil':<12} {'Valor':<8} {'Interpretaci√≥n'}")
print("-" * 40)

percentiles_calculados = {}
for p in percentiles_interes:
    valor = np.percentile(notas, p)
    percentiles_calculados[p] = valor
    
    # Interpretar cada percentil
    if p == 50:
        interpretacion = "Mediana"
    elif p <= 25:
        interpretacion = "Cuarto inferior"
    elif p >= 75:
        interpretacion = "Cuarto superior"
    else:
        interpretacion = f"{p}% por debajo"
    
    print(f"P{p:<11} {valor:<8.2f} {interpretacion}")

# An√°lisis espec√≠fico de cuartiles
q1 = percentiles_calculados[25]
q2 = percentiles_calculados[50]  # Mediana
q3 = percentiles_calculados[75]
iqr = q3 - q1

print(f"\nüéØ AN√ÅLISIS DE CUARTILES:")
print(f"   Q1 (P25): {q1:.2f} - 25% de estudiantes tienen nota ‚â§ {q1:.2f}")
print(f"   Q2 (P50): {q2:.2f} - 50% de estudiantes tienen nota ‚â§ {q2:.2f} (Mediana)")
print(f"   Q3 (P75): {q3:.2f} - 75% de estudiantes tienen nota ‚â§ {q3:.2f}")
print(f"   IQR: {iqr:.2f} - Rango del 50% central de los datos")

# Ejemplos pr√°cticos con estudiantes espec√≠ficos
print(f"\nüë• EJEMPLOS CON ESTUDIANTES ESPEC√çFICOS:")
estudiantes_ejemplo = [
    estudiantes_df.loc[estudiantes_df['nota_promedio'].idxmax()],  # Mejor nota
    estudiantes_df.loc[estudiantes_df['nota_promedio'].idxmin()],  # Peor nota
    estudiantes_df.iloc[len(estudiantes_df)//2]  # Estudiante del medio
]

for i, estudiante in enumerate(estudiantes_ejemplo):
    nota = estudiante['nota_promedio']
    # Calcular en qu√© percentil est√°
    percentil = stats.percentileofscore(notas, nota)
    
    if i == 0:
        tipo = "MEJOR estudiante"
    elif i == 1:
        tipo = "PEOR estudiante"
    else:
        tipo = "Estudiante PROMEDIO"
    
    print(f"   {tipo}:")
    print(f"      Nota: {nota:.2f} ‚Üí Percentil {percentil:.1f}")
    print(f"      Supera al {percentil:.1f}% de los estudiantes")
````

## PUNTUACIONES Z (Z-SCORES) - ESTANDARIZACI√ìN Y COMPARACIONES

````python
print("\n" + "="*70)
print("‚öñÔ∏è PUNTUACIONES Z - ESTANDARIZACI√ìN Y COMPARACIONES")
print("="*70)

# Calcular Z-scores para m√∫ltiples variables
variables_numericas = ['edad', 'nota_promedio', 'horas_estudio']

print("üìä C√ÅLCULO DE Z-SCORES PASO A PASO:")
print("-" * 50)

z_scores_df = pd.DataFrame()

for var in variables_numericas:
    datos = estudiantes_df[var]
    media = datos.mean()
    std = datos.std()
    
    # Calcular Z-scores manualmente
    z_manual = (datos - media) / std
    
    # Verificar con scipy
    z_scipy = zscore(datos)
    
    # Guardar en DataFrame
    z_scores_df[f'z_{var}'] = z_manual
    
    print(f"\nüî¢ {var.upper().replace('_', ' ')}:")
    print(f"   Media: {media:.2f}")
    print(f"   Desv.Std: {std:.2f}")
    print(f"   Z-score = (x - {media:.2f}) / {std:.2f}")
    
    # Mostrar algunos ejemplos
    indices_ejemplo = [datos.idxmax(), datos.idxmin(), datos.median()]
    for idx in [datos.idxmax(), datos.idxmin()]:
        valor = datos.loc[idx]
        z_val = z_manual.loc[idx]
        if idx == datos.idxmax():
            tipo = "M√ÅXIMO"
        else:
            tipo = "M√çNIMO"
        print(f"   Ejemplo {tipo}: {valor} ‚Üí Z = {z_val:.2f}")

# Comparaci√≥n entre variables usando Z-scores
print(f"\n‚öñÔ∏è COMPARACI√ìN ENTRE VARIABLES CON Z-SCORES:")
print("-" * 60)

# Tomar 5 estudiantes aleatorios para an√°lisis comparativo
estudiantes_muestra = estudiantes_df.sample(5, random_state=42)

print(f"{'ID':<4} {'Edad':<6} {'Z-Edad':<8} {'Nota':<6} {'Z-Nota':<8} {'Horas':<6} {'Z-Horas':<8} {'Mejor En'}")
print("-" * 60)

for _, estudiante in estudiantes_muestra.iterrows():
    idx = estudiante.name
    z_edad = z_scores_df.loc[idx, 'z_edad']
    z_nota = z_scores_df.loc[idx, 'z_nota_promedio']
    z_horas = z_scores_df.loc[idx, 'z_horas_estudio']
    
    # Determinar en qu√© es mejor (mayor Z-score)
    z_values = {'Edad': z_edad, 'Notas': z_nota, 'Estudio': z_horas}
    mejor_en = max(z_values, key=z_values.get)
    
    print(f"{estudiante['id']:<4} {estudiante['edad']:<6} {z_edad:<8.2f} "
          f"{estudiante['nota_promedio']:<6.2f} {z_nota:<8.2f} "
          f"{estudiante['horas_estudio']:<6} {z_horas:<8.2f} {mejor_en}")

# Detecci√≥n de outliers usando Z-scores
print(f"\nüéØ DETECCI√ìN DE OUTLIERS CON Z-SCORES:")
print("-" * 50)

for var in variables_numericas:
    z_col = f'z_{var}'
    outliers_moderados = np.abs(z_scores_df[z_col]) > 2
    outliers_extremos = np.abs(z_scores_df[z_col]) > 3
    
    n_moderados = outliers_moderados.sum()
    n_extremos = outliers_extremos.sum()
    
    print(f"üìä {var.upper().replace('_', ' ')}:")
    print(f"   Outliers moderados (|Z| > 2): {n_moderados} ({n_moderados/len(estudiantes_df)*100:.1f}%)")
    print(f"   Outliers extremos (|Z| > 3): {n_extremos} ({n_extremos/len(estudiantes_df)*100:.1f}%)")
    
    if n_extremos > 0:
        valores_extremos = estudiantes_df.loc[outliers_extremos, var].values
        print(f"   Valores extremos: {valores_extremos}")
````

# AN√ÅLISIS DE ASIMETR√çA (SKEWNESS)

````python
print("\n" + "="*70)
print("üìà AN√ÅLISIS DE ASIMETR√çA (SKEWNESS)")
print("="*70)

print("üìä C√ÅLCULO DE ASIMETR√çA PARA CADA VARIABLE:")
print("-" * 65)
print(f"{'Variable':<15} {'Asimetr√≠a':<10} {'Interpretaci√≥n':<20} {'Recomendaci√≥n'}")
print("-" * 65)

resultados_asimetria = {}

for var in variables_numericas:
    datos = estudiantes_df[var]
    
    # Calcular asimetr√≠a
    asimetria = skew(datos)
    resultados_asimetria[var] = asimetria
    
    # Interpretar asimetr√≠a
    if abs(asimetria) < 0.5:
        interpretacion = "Aproxim. sim√©trica"
        recomendacion = "Media"
    elif abs(asimetria) < 1:
        if asimetria > 0:
            interpretacion = "Moderada positiva"
        else:
            interpretacion = "Moderada negativa"
        recomendacion = "Media o Mediana"
    else:
        if asimetria > 0:
            interpretacion = "Muy positiva"
        else:
            interpretacion = "Muy negativa"
        recomendacion = "Mediana"
    
    print(f"{var:<15} {asimetria:<10.3f} {interpretacion:<20} {recomendacion}")

# An√°lisis detallado de la variable m√°s asim√©trica
var_mas_asimetrica = max(resultados_asimetria, key=lambda k: abs(resultados_asimetria[k]))
asimetria_max = resultados_asimetria[var_mas_asimetrica]

print(f"\nüîç AN√ÅLISIS DETALLADO - {var_mas_asimetrica.upper().replace('_', ' ')}:")
print(f"   Asimetr√≠a: {asimetria_max:.3f}")

datos_asimetrica = estudiantes_df[var_mas_asimetrica]
media_asim = datos_asimetrica.mean()
mediana_asim = datos_asimetrica.median()
moda_asim = datos_asimetrica.mode().iloc[0] if len(datos_asimetrica.mode()) > 0 else "N/A"

print(f"   Media: {media_asim:.2f}")
print(f"   Mediana: {mediana_asim:.2f}")
print(f"   Moda: {moda_asim}")

if asimetria_max > 0:
    print(f"   ‚ö†Ô∏è Asimetr√≠a POSITIVA: Media > Mediana")
    print(f"      La mayor√≠a de valores est√°n por debajo de la media")
    print(f"      Hay algunos valores muy altos que 'jalan' la media hacia arriba")
else:
    print(f"   ‚ö†Ô∏è Asimetr√≠a NEGATIVA: Media < Mediana")
    print(f"      La mayor√≠a de valores est√°n por encima de la media")
    print(f"      Hay algunos valores muy bajos que 'jalan' la media hacia abajo")
````

# AN√ÅLISIS DE CURTOSIS

````python
print("\n" + "="*70)
print("üìä AN√ÅLISIS DE CURTOSIS")
print("="*70)

print("üìä C√ÅLCULO DE CURTOSIS PARA CADA VARIABLE:")
print("-" * 65)
print(f"{'Variable':<15} {'Curtosis':<10} {'Tipo':<15} {'Interpretaci√≥n'}")
print("-" * 65)

resultados_curtosis = {}

for var in variables_numericas:
    datos = estudiantes_df[var]
    
    # Calcular curtosis (usando fisher=True para exceso de curtosis)
    kurt = kurtosis(datos, fisher=True)
    resultados_curtosis[var] = kurt
    
    # Interpretar curtosis
    if abs(kurt) < 1:
        tipo = "Mesoc√∫rtica"
        interpretacion = "Similar a normal"
    elif kurt > 1:
        tipo = "Leptoc√∫rtica"
        interpretacion = "M√°s puntiaguda"
    else:
        tipo = "Platic√∫rtica"
        interpretacion = "M√°s aplanada"
    
    print(f"{var:<15} {kurt:<10.3f} {tipo:<15} {interpretacion}")

# An√°lisis detallado de curtosis
print(f"\nüîç INTERPRETACI√ìN DETALLADA DE CURTOSIS:")

for var in variables_numericas:
    kurt = resultados_curtosis[var]
    print(f"\nüìä {var.upper().replace('_', ' ')} (Curtosis: {kurt:.3f}):")
    
    if kurt > 1:
        print(f"   üî∫ LEPTOC√öRTICA:")
        print(f"      - Distribuci√≥n m√°s 'puntiaguda' que la normal")
        print(f"      - Mayor concentraci√≥n alrededor de la media")
        print(f"      - Colas m√°s 'pesadas' (m√°s valores extremos)")
        print(f"      - Mayor probabilidad de outliers")
    elif kurt < -1:
        print(f"   üîª PLATIC√öRTICA:")
        print(f"      - Distribuci√≥n m√°s 'aplanada' que la normal")
        print(f"      - Menor concentraci√≥n alrededor de la media")
        print(f"      - Colas m√°s 'ligeras' (pocos valores extremos)")
        print(f"      - Distribuci√≥n m√°s uniforme")
    else:
        print(f"   üü¢ MESOC√öRTICA:")
        print(f"      - Distribuci√≥n similar a la normal")
        print(f"      - Concentraci√≥n moderada alrededor de la media")
        print(f"      - Patr√≥n de colas normal")
````

# AN√ÅLISIS ESTAD√çSTICO COMPLETO CON LO VISTO HASTA LA SESION 4

````python
print("\n" + "="*70)
print("üìã AN√ÅLISIS ESTAD√çSTICO COMPLETO - INTEGRACI√ìN DE TODAS LAS MEDIDAS")
print("="*70)

def crear_perfil_estadistico(datos, nombre_variable):
    """Crear un perfil estad√≠stico completo para una variable"""
    
    # Medidas de tendencia central
    media = datos.mean()
    mediana = datos.median()
    moda = datos.mode().iloc[0] if len(datos.mode()) > 0 else None
    
    # Medidas de dispersi√≥n
    std = datos.std()
    cv = (std / media) * 100
    rango = datos.max() - datos.min()
    iqr = datos.quantile(0.75) - datos.quantile(0.25)
    
    # Medidas de posici√≥n
    percentiles = [datos.quantile(p/100) for p in [5, 10, 25, 50, 75, 90, 95]]
    
    # Medidas de forma
    asimetria = skew(datos)
    curtosis_val = kurtosis(datos, fisher=True)
    
    # Outliers
    z_scores = np.abs((datos - media) / std)
    outliers_count = (z_scores > 2).sum()
    
    return {
        'nombre': nombre_variable,
        'n': len(datos),
        'media': media,
        'mediana': mediana,
        'moda': moda,
        'std': std,
        'cv': cv,
        'rango': rango,
        'iqr': iqr,
        'percentiles': percentiles,
        'asimetria': asimetria,
        'curtosis': curtosis_val,
        'outliers': outliers_count
    }

# Crear perfiles para todas las variables
perfiles = {}
for var in variables_numericas:
    perfiles[var] = crear_perfil_estadistico(estudiantes_df[var], var)

# Mostrar tabla resumen
print("üìä TABLA RESUMEN EJECUTIVA:")
print("-" * 90)
print(f"{'Variable':<15} {'Media':<8} {'Mediana':<8} {'CV%':<8} {'Asimetr√≠a':<10} {'Curtosis':<10} {'Outliers'}")
print("-" * 90)

for var in variables_numericas:
    p = perfiles[var]
    print(f"{var:<15} {p['media']:<8.2f} {p['mediana']:<8.2f} {p['cv']:<8.1f} "
          f"{p['asimetria']:<10.3f} {p['curtosis']:<10.3f} {p['outliers']}")

# Recomendaciones basadas en el perfil
print(f"\nüí° RECOMENDACIONES METODOL√ìGICAS:")
print("-" * 50)

for var in variables_numericas:
    p = perfiles[var]
    print(f"\nüìä {var.upper().replace('_', ' ')}:")
    
    # Recomendaci√≥n de medida central
    if abs(p['asimetria']) < 0.5:
        medida_central = "Media (distribuci√≥n aproximadamente sim√©trica)"
    else:
        medida_central = "Mediana (distribuci√≥n asim√©trica)"
    
    # Recomendaci√≥n de dispersi√≥n
    if p['outliers'] > len(estudiantes_df) * 0.05:  # M√°s del 5% outliers
        medida_dispersion = "IQR (presencia significativa de outliers)"
    else:
        medida_dispersion = "Desviaci√≥n Est√°ndar (pocos outliers)"
    
    # Evaluaci√≥n de normalidad
    if abs(p['asimetria']) < 0.5 and abs(p['curtosis']) < 1:
        normalidad = "Aproximadamente normal (aplicar regla emp√≠rica)"
    else:
        normalidad = "No normal (cuidado con supuestos de normalidad)"
    
    print(f"   ‚úì Medida central: {medida_central}")
    print(f"   ‚úì Medida dispersi√≥n: {medida_dispersion}")
    print(f"   ‚úì Normalidad: {normalidad}")
````
