## 📊 **PARTE 2: Percentiles y Cuartiles Avanzados**

### **🎯 Conceptos Fundamentales**

#### **Percentiles (Pk):**
> El **percentil k** es el valor que deja el k% de los datos por debajo de él.

#### **Relaciones Importantes:**
```
CUARTILES = PERCENTILES ESPECIALES
├── Q1 = P25 (Primer Cuartil)
├── Q2 = P50 = Mediana (Segundo Cuartil)  
└── Q3 = P75 (Tercer Cuartil)

DECILES = CADA 10%
├── D1 = P10, D2 = P20, ..., D9 = P90
```

### **🔢 Cálculo de Percentiles**

#### **Método General (Interpolación Lineal):**

1. **Ordenar** los datos de menor a mayor
2. **Calcular posición**: `posición = (k/100) × (n + 1)`
3. **Si la posición es entera**: Ese es el percentil
4. **Si no es entera**: Interpolar entre los valores adyacentes

#### **💡 Ejemplo Detallado:**

**Datos**: 12, 15, 18, 20, 22, 25, 28, 30, 35, 40 (n = 10)

**Calcular P30 (percentil 30):**

1. **Posición**: (30/100) × (10 + 1) = 3.3
2. **No es entero**, entonces interpolamos entre posición 3 y 4
3. **Valores**: x₃ = 18, x₄ = 20  
4. **Interpolación**: P30 = 18 + 0.3 × (20 - 18) = 18 + 0.6 = 18.6

**Interpretación**: El 30% de los datos son ≤ 18.6

### **📈 Aplicaciones Prácticas de Percentiles**

#### **1. Evaluación Académica:**
- "Estás en el percentil 85 en matemáticas"
- Significa: superas al 85% de los estudiantes

#### **2. Medicina:**
- Tablas de crecimiento infantil por percentiles
- P50 = crecimiento promedio, P95 = crecimiento excepcional

#### **3. Finanzas:**
- VaR (Value at Risk): P5 de pérdidas esperadas
- "En el 5% de los casos peores, esperamos perder más de X"

#### **4. Recursos Humanos:**
- Benchmarking salarial por percentiles
- P75 de salarios = "salario competitivo"

### **✅ Ventajas de los Percentiles**

1. **Intuitivos**: Fáciles de interpretar en contexto
2. **Robustos**: No afectados por valores extremos
3. **Flexibles**: Cualquier división porcentual
4. **Comparativos**: Excelentes para ranking y benchmarking

### **⚠️ Limitaciones de los Percentiles**

1. **Dependientes del orden**: Solo funcionan con datos ordenables
2. **Menos precisos con muestras pequeñas**: Interpolación menos confiable
3. **No aditivos**: P50 + P50 ≠ P100

## 📊 **PARTE 3: Puntuaciones Z (Z-scores) - Estandarización Universal**

### **🎯 El Problema de las Escalas Diferentes**

¿Qué es mejor?
- **Juan**: 85 en un examen de matemáticas (media = 75, σ = 10)
- **María**: 78 en un examen de historia (media = 70, σ = 4)

**¡No podemos compararlos directamente!** Necesitamos **estandarizar**.

### **🔢 Puntuación Z (Z-score)**

La **puntuación Z** indica cuántas desviaciones estándar está un valor por encima o por debajo de la media.

#### **Fórmula:**
```
Z = (X - μ) / σ     (población)
Z = (X - x̄) / s     (muestra)
```

#### **💡 Resolviendo el Ejemplo:**

**Juan**: Z = (85 - 75) / 10 = 1.0
**María**: Z = (78 - 70) / 4 = 2.0

**Conclusión**: María tuvo mejor rendimiento relativo (2.0 > 1.0)

### **📊 Interpretación de Z-scores**

#### **Valores Típicos:**
- **Z = 0**: Exactamente en la media
- **Z = 1**: Una desviación estándar por encima de la media
- **Z = -1**: Una desviación estándar por debajo de la media
- **|Z| > 2**: Valores poco comunes (en los extremos)
- **|Z| > 3**: Valores muy raros (posibles outliers)

#### **En Distribución Normal:**
- **68%** de los datos tienen |Z| ≤ 1
- **95%** de los datos tienen |Z| ≤ 2  
- **99.7%** de los datos tienen |Z| ≤ 3

### **📈 Aplicaciones de Z-scores**

#### **1. Comparación entre Variables Diferentes:**
```python
# Comparar altura vs peso de atletas
z_altura = (altura - altura_promedio) / altura_std
z_peso = (peso - peso_promedio) / peso_std
# Ahora podemos compararlos directamente
```

#### **2. Detección de Outliers:**
```python
outliers = datos[abs(z_scores) > 3]  # Valores muy extremos
```

#### **3. Estandarización para Machine Learning:**
```python
datos_estandarizados = (datos - datos.mean()) / datos.std()
# Media = 0, Desv.Std = 1
```

#### **4. Percentiles en Distribución Normal:**
- Z = -1.645 → P5 (percentil 5)
- Z = 0 → P50 (mediana)
- Z = 1.645 → P95 (percentil 95)

### **✅ Ventajas de Z-scores**

1. **Estandarización universal**: Todas las variables quedan en la misma escala
2. **Comparación directa**: Entre diferentes medidas y grupos
3. **Detección de outliers**: Criterio objetivo (|Z| > 2 o 3)
4. **Base para otros cálculos**: Correlaciones, regresiones, etc.

### **⚠️ Limitaciones de Z-scores**

1. **Asume normalidad**: La interpretación es óptima con distribuciones normales
2. **Sensible a outliers**: Media y desviación estándar pueden estar distorsionadas
3. **Pérdida de unidades originales**: No conserva la interpretación original

## 📊 **PARTE 4: Asimetría (Skewness) - La Inclinación de los Datos**

### **🎯 ¿Qué es la Asimetría?**

La **asimetría** mide qué tan "inclinada" está una distribución respecto a la simetría perfecta.

> Una distribución **simétrica** tiene la misma forma a ambos lados de la media. Una distribución **asimétrica** se "inclina" hacia un lado.

### **📊 Tipos de Asimetría**

#### **1. Asimetría Negativa (Sesgo a la Izquierda)**
```
    📊
   /|
  / |
 /  |     ← Cola larga hacia la izquierda
/___|
```
- **Característica**: Cola larga hacia valores bajos
- **Relación**: Media < Mediana < Moda
- **Ejemplos**: Edad al morir (muchos viven mucho, pocos mueren jóvenes), calificaciones fáciles

#### **2. Distribución Simétrica**
```
  📊
 /|\
/ | \     ← Forma de campana perfecta
\___/
```
- **Característica**: Perfectamente balanceada
- **Relación**: Media = Mediana = Moda
- **Ejemplos**: Altura, peso (en poblaciones homogéneas), errores de medición

#### **3. Asimetría Positiva (Sesgo a la Derecha)**
```
📊
|\
| \
|  \      ← Cola larga hacia la derecha  
|___\
```
- **Característica**: Cola larga hacia valores altos
- **Relación**: Moda < Mediana < Media
- **Ejemplos**: Ingresos, precios de casas, tiempo de respuesta

### **🔢 Medición de la Asimetría**

#### **Coeficiente de Asimetría de Fisher:**
```
Skewness = E[(X - μ)³] / σ³
```

**Para una muestra:**
```
Sk = [n / ((n-1)(n-2))] × Σ[(xi - x̄) / s]³
```

#### **Interpretación del Coeficiente:**
- **Sk = 0**: Distribución perfectamente simétrica
- **Sk > 0**: Asimetría positiva (cola derecha)
- **Sk < 0**: Asimetría negativa (cola izquierda)

#### **Reglas Prácticas:**
- **|Sk| < 0.5**: Aproximadamente simétrica
- **0.5 ≤ |Sk| < 1**: Moderadamente asimétrica  
- **|Sk| ≥ 1**: Altamente asimétrica

### **💡 Ejemplo Práctico:**

**Salarios en una empresa** (en miles):
30, 32, 35, 38, 40, 42, 45, 48, 120

- **Media**: 47.8K (afectada por el CEO)
- **Mediana**: 40K (más representativa)
- **Asimetría**: Positiva (Sk > 0)

**Interpretación**: La mayoría gana cerca de 40K, pero unos pocos salarios muy altos jalan la media hacia arriba.

### **📈 Aplicaciones de la Asimetría**

#### **1. Selección de Medidas Centrales:**
- **Si Sk ≈ 0**: Usar media (distribución simétrica)
- **Si Sk ≠ 0**: Preferir mediana (más robusta)

#### **2. Análisis de Riesgo Financiero:**
- **Sk > 0**: Pocos eventos de gran ganancia, muchas pérdidas pequeñas
- **Sk < 0**: Pocos eventos de gran pérdida, muchas ganancias pequeñas

#### **3. Control de Calidad:**
- **Proceso simétrico (Sk ≈ 0)**: Proceso controlado
- **Proceso asimétrico**: Investigar causas de sesgo

#### **4. Modelado Estadístico:**
- **Sk ≠ 0**: La distribución normal no es apropiada
- Considerar transformaciones o distribuciones alternativas

## 📊 **PARTE 5: Curtosis - La "Altura" de la Distribución**

### **🎯 ¿Qué es la Curtosis?**

La **curtosis** mide qué tan "puntiaguda" o "aplanada" es una distribución comparada con la distribución normal.

> La curtosis describe la **concentración** de datos alrededor del centro y el **grosor** de las colas.

### **📊 Tipos de Curtosis**

#### **1. Mesocúrtica (Curtosis Normal)**
```
   📊
  /|\
 / | \     ← Como la distribución normal
/  |  \
\___|___/
```
- **Kurt = 3** (o Kurt = 0 en exceso de curtosis)
- **Referencia**: Distribución normal estándar

#### **2. Leptocúrtica (Más Puntiaguda)**
```
    📊
   /|\
  / | \    ← Pico más alto, colas más pesadas
 /  |  \
/____|____\
```
- **Kurt > 3** (o Kurt > 0 en exceso)
- **Características**: 
  - Pico más alto y agudo
  - Colas más pesadas (más valores extremos)
  - Mayor concentración alrededor de la media

#### **3. Platicúrtica (Más Aplanada)**
```
  ___📊___
 /   |   \   ← Más plana, colas más ligeras
/    |    \
\_______/
```
- **Kurt < 3** (o Kurt < 0 en exceso)
- **Características**:
  - Distribución más plana
  - Colas más ligeras (pocos valores extremos)
  - Menor concentración alrededor de la media

### **🔢 Medición de la Curtosis**

#### **Coeficiente de Curtosis:**
```
Curtosis = E[(X - μ)⁴] / σ⁴
```

#### **Exceso de Curtosis (más común):**
```
Exceso = Curtosis - 3
```

**Para una muestra:**
```
Kurt = [n(n+1)/((n-1)(n-2)(n-3))] × Σ[(xi - x̄)/s]⁴ - 3(n-1)²/((n-2)(n-3))
```

#### **Interpretación:**
- **Kurt = 0**: Curtosis normal (como distribución normal)
- **Kurt > 0**: Leptocúrtica (más puntiaguda)
- **Kurt < 0**: Platicúrtica (más aplanada)

#### **Reglas Prácticas:**
- **|Kurt| < 1**: Curtosis similar a la normal
- **1 ≤ |Kurt| < 3**: Curtosis moderadamente diferente
- **|Kurt| ≥ 3**: Curtosis muy diferente a la normal

### **💡 Ejemplo Conceptual:**

**Tiempo de respuesta en un sistema:**

**Sistema A (Leptocúrtica)**:
- La mayoría responde en ~2 segundos
- Pero ocasionalmente hay respuestas de 10+ segundos
- **Interpretación**: Sistema generalmente rápido pero con fallas ocasionales severas

**Sistema B (Platicúrtica)**:
- Respuestas distribuidas uniformemente entre 1-5 segundos
- Pocos valores extremos
- **Interpretación**: Sistema más predecible, sin picos de rendimiento ni fallas severas

### **📈 Aplicaciones de la Curtosis**

#### **1. Análisis de Riesgo Financiero:**
- **Alta curtosis**: Mayor probabilidad de eventos extremos (crisis o booms)
- **Baja curtosis**: Rendimientos más predecibles

#### **2. Control de Calidad:**
- **Alta curtosis**: Proceso muy controlado pero con fallas ocasionales graves
- **Baja curtosis**: Proceso menos controlado pero más predecible

#### **3. Detección de Outliers:**
- **Alta curtosis**: Indica presencia de valores extremos frecuentes
- **Baja curtosis**: Pocos outliers en los datos

#### **4. Selección de Modelos:**
- **Kurt ≠ 0**: La distribución normal no es apropiada
- Considerar distribuciones con curtosis específica (t-Student, Laplace, etc.)

````python
print("\n" + "="*70)
print("📏 ANÁLISIS DETALLADO DE PERCENTILES Y CUARTILES")
print("="*70)

# Tomar las notas como ejemplo principal
notas = estudiantes_df['nota_promedio'].copy()

print(f"📊 Analizando notas promedio de {len(notas)} estudiantes")
print(f"   Rango: {notas.min():.2f} - {notas.max():.2f}")

# Cálculo manual de percentiles específicos
percentiles_interes = [5, 10, 25, 50, 75, 90, 95]
print(f"\n📋 PERCENTILES CLAVE:")
print("-" * 40)
print(f"{'Percentil':<12} {'Valor':<8} {'Interpretación'}")
print("-" * 40)

percentiles_calculados = {}
for p in percentiles_interes:
    valor = np.percentile(notas, p)
    percentiles_calculados[p] = valor
    
    # Interpretar cada percentil
    if p == 50:
        interpretacion = "Mediana"
    elif p <= 25:
        interpretacion = "Cuarto inferior"
    elif p >= 75:
        interpretacion = "Cuarto superior"
    else:
        interpretacion = f"{p}% por debajo"
    
    print(f"P{p:<11} {valor:<8.2f} {interpretacion}")

# Análisis específico de cuartiles
q1 = percentiles_calculados[25]
q2 = percentiles_calculados[50]  # Mediana
q3 = percentiles_calculados[75]
iqr = q3 - q1

print(f"\n🎯 ANÁLISIS DE CUARTILES:")
print(f"   Q1 (P25): {q1:.2f} - 25% de estudiantes tienen nota ≤ {q1:.2f}")
print(f"   Q2 (P50): {q2:.2f} - 50% de estudiantes tienen nota ≤ {q2:.2f} (Mediana)")
print(f"   Q3 (P75): {q3:.2f} - 75% de estudiantes tienen nota ≤ {q3:.2f}")
print(f"   IQR: {iqr:.2f} - Rango del 50% central de los datos")

# Ejemplos prácticos con estudiantes específicos
print(f"\n👥 EJEMPLOS CON ESTUDIANTES ESPECÍFICOS:")
estudiantes_ejemplo = [
    estudiantes_df.loc[estudiantes_df['nota_promedio'].idxmax()],  # Mejor nota
    estudiantes_df.loc[estudiantes_df['nota_promedio'].idxmin()],  # Peor nota
    estudiantes_df.iloc[len(estudiantes_df)//2]  # Estudiante del medio
]

for i, estudiante in enumerate(estudiantes_ejemplo):
    nota = estudiante['nota_promedio']
    # Calcular en qué percentil está
    percentil = stats.percentileofscore(notas, nota)
    
    if i == 0:
        tipo = "MEJOR estudiante"
    elif i == 1:
        tipo = "PEOR estudiante"
    else:
        tipo = "Estudiante PROMEDIO"
    
    print(f"   {tipo}:")
    print(f"      Nota: {nota:.2f} → Percentil {percentil:.1f}")
    print(f"      Supera al {percentil:.1f}% de los estudiantes")
````

## PUNTUACIONES Z (Z-SCORES) - ESTANDARIZACIÓN Y COMPARACIONES

````python
print("\n" + "="*70)
print("⚖️ PUNTUACIONES Z - ESTANDARIZACIÓN Y COMPARACIONES")
print("="*70)

# Calcular Z-scores para múltiples variables
variables_numericas = ['edad', 'nota_promedio', 'horas_estudio']

print("📊 CÁLCULO DE Z-SCORES PASO A PASO:")
print("-" * 50)

z_scores_df = pd.DataFrame()

for var in variables_numericas:
    datos = estudiantes_df[var]
    media = datos.mean()
    std = datos.std()
    
    # Calcular Z-scores manualmente
    z_manual = (datos - media) / std
    
    # Verificar con scipy
    z_scipy = zscore(datos)
    
    # Guardar en DataFrame
    z_scores_df[f'z_{var}'] = z_manual
    
    print(f"\n🔢 {var.upper().replace('_', ' ')}:")
    print(f"   Media: {media:.2f}")
    print(f"   Desv.Std: {std:.2f}")
    print(f"   Z-score = (x - {media:.2f}) / {std:.2f}")
    
    # Mostrar algunos ejemplos
    indices_ejemplo = [datos.idxmax(), datos.idxmin(), datos.median()]
    for idx in [datos.idxmax(), datos.idxmin()]:
        valor = datos.loc[idx]
        z_val = z_manual.loc[idx]
        if idx == datos.idxmax():
            tipo = "MÁXIMO"
        else:
            tipo = "MÍNIMO"
        print(f"   Ejemplo {tipo}: {valor} → Z = {z_val:.2f}")

# Comparación entre variables usando Z-scores
print(f"\n⚖️ COMPARACIÓN ENTRE VARIABLES CON Z-SCORES:")
print("-" * 60)

# Tomar 5 estudiantes aleatorios para análisis comparativo
estudiantes_muestra = estudiantes_df.sample(5, random_state=42)

print(f"{'ID':<4} {'Edad':<6} {'Z-Edad':<8} {'Nota':<6} {'Z-Nota':<8} {'Horas':<6} {'Z-Horas':<8} {'Mejor En'}")
print("-" * 60)

for _, estudiante in estudiantes_muestra.iterrows():
    idx = estudiante.name
    z_edad = z_scores_df.loc[idx, 'z_edad']
    z_nota = z_scores_df.loc[idx, 'z_nota_promedio']
    z_horas = z_scores_df.loc[idx, 'z_horas_estudio']
    
    # Determinar en qué es mejor (mayor Z-score)
    z_values = {'Edad': z_edad, 'Notas': z_nota, 'Estudio': z_horas}
    mejor_en = max(z_values, key=z_values.get)
    
    print(f"{estudiante['id']:<4} {estudiante['edad']:<6} {z_edad:<8.2f} "
          f"{estudiante['nota_promedio']:<6.2f} {z_nota:<8.2f} "
          f"{estudiante['horas_estudio']:<6} {z_horas:<8.2f} {mejor_en}")

# Detección de outliers usando Z-scores
print(f"\n🎯 DETECCIÓN DE OUTLIERS CON Z-SCORES:")
print("-" * 50)

for var in variables_numericas:
    z_col = f'z_{var}'
    outliers_moderados = np.abs(z_scores_df[z_col]) > 2
    outliers_extremos = np.abs(z_scores_df[z_col]) > 3
    
    n_moderados = outliers_moderados.sum()
    n_extremos = outliers_extremos.sum()
    
    print(f"📊 {var.upper().replace('_', ' ')}:")
    print(f"   Outliers moderados (|Z| > 2): {n_moderados} ({n_moderados/len(estudiantes_df)*100:.1f}%)")
    print(f"   Outliers extremos (|Z| > 3): {n_extremos} ({n_extremos/len(estudiantes_df)*100:.1f}%)")
    
    if n_extremos > 0:
        valores_extremos = estudiantes_df.loc[outliers_extremos, var].values
        print(f"   Valores extremos: {valores_extremos}")
````

# ANÁLISIS DE ASIMETRÍA (SKEWNESS)

````python
print("\n" + "="*70)
print("📈 ANÁLISIS DE ASIMETRÍA (SKEWNESS)")
print("="*70)

print("📊 CÁLCULO DE ASIMETRÍA PARA CADA VARIABLE:")
print("-" * 65)
print(f"{'Variable':<15} {'Asimetría':<10} {'Interpretación':<20} {'Recomendación'}")
print("-" * 65)

resultados_asimetria = {}

for var in variables_numericas:
    datos = estudiantes_df[var]
    
    # Calcular asimetría
    asimetria = skew(datos)
    resultados_asimetria[var] = asimetria
    
    # Interpretar asimetría
    if abs(asimetria) < 0.5:
        interpretacion = "Aproxim. simétrica"
        recomendacion = "Media"
    elif abs(asimetria) < 1:
        if asimetria > 0:
            interpretacion = "Moderada positiva"
        else:
            interpretacion = "Moderada negativa"
        recomendacion = "Media o Mediana"
    else:
        if asimetria > 0:
            interpretacion = "Muy positiva"
        else:
            interpretacion = "Muy negativa"
        recomendacion = "Mediana"
    
    print(f"{var:<15} {asimetria:<10.3f} {interpretacion:<20} {recomendacion}")

# Análisis detallado de la variable más asimétrica
var_mas_asimetrica = max(resultados_asimetria, key=lambda k: abs(resultados_asimetria[k]))
asimetria_max = resultados_asimetria[var_mas_asimetrica]

print(f"\n🔍 ANÁLISIS DETALLADO - {var_mas_asimetrica.upper().replace('_', ' ')}:")
print(f"   Asimetría: {asimetria_max:.3f}")

datos_asimetrica = estudiantes_df[var_mas_asimetrica]
media_asim = datos_asimetrica.mean()
mediana_asim = datos_asimetrica.median()
moda_asim = datos_asimetrica.mode().iloc[0] if len(datos_asimetrica.mode()) > 0 else "N/A"

print(f"   Media: {media_asim:.2f}")
print(f"   Mediana: {mediana_asim:.2f}")
print(f"   Moda: {moda_asim}")

if asimetria_max > 0:
    print(f"   ⚠️ Asimetría POSITIVA: Media > Mediana")
    print(f"      La mayoría de valores están por debajo de la media")
    print(f"      Hay algunos valores muy altos que 'jalan' la media hacia arriba")
else:
    print(f"   ⚠️ Asimetría NEGATIVA: Media < Mediana")
    print(f"      La mayoría de valores están por encima de la media")
    print(f"      Hay algunos valores muy bajos que 'jalan' la media hacia abajo")
````

# ANÁLISIS DE CURTOSIS

````python
print("\n" + "="*70)
print("📊 ANÁLISIS DE CURTOSIS")
print("="*70)

print("📊 CÁLCULO DE CURTOSIS PARA CADA VARIABLE:")
print("-" * 65)
print(f"{'Variable':<15} {'Curtosis':<10} {'Tipo':<15} {'Interpretación'}")
print("-" * 65)

resultados_curtosis = {}

for var in variables_numericas:
    datos = estudiantes_df[var]
    
    # Calcular curtosis (usando fisher=True para exceso de curtosis)
    kurt = kurtosis(datos, fisher=True)
    resultados_curtosis[var] = kurt
    
    # Interpretar curtosis
    if abs(kurt) < 1:
        tipo = "Mesocúrtica"
        interpretacion = "Similar a normal"
    elif kurt > 1:
        tipo = "Leptocúrtica"
        interpretacion = "Más puntiaguda"
    else:
        tipo = "Platicúrtica"
        interpretacion = "Más aplanada"
    
    print(f"{var:<15} {kurt:<10.3f} {tipo:<15} {interpretacion}")

# Análisis detallado de curtosis
print(f"\n🔍 INTERPRETACIÓN DETALLADA DE CURTOSIS:")

for var in variables_numericas:
    kurt = resultados_curtosis[var]
    print(f"\n📊 {var.upper().replace('_', ' ')} (Curtosis: {kurt:.3f}):")
    
    if kurt > 1:
        print(f"   🔺 LEPTOCÚRTICA:")
        print(f"      - Distribución más 'puntiaguda' que la normal")
        print(f"      - Mayor concentración alrededor de la media")
        print(f"      - Colas más 'pesadas' (más valores extremos)")
        print(f"      - Mayor probabilidad de outliers")
    elif kurt < -1:
        print(f"   🔻 PLATICÚRTICA:")
        print(f"      - Distribución más 'aplanada' que la normal")
        print(f"      - Menor concentración alrededor de la media")
        print(f"      - Colas más 'ligeras' (pocos valores extremos)")
        print(f"      - Distribución más uniforme")
    else:
        print(f"   🟢 MESOCÚRTICA:")
        print(f"      - Distribución similar a la normal")
        print(f"      - Concentración moderada alrededor de la media")
        print(f"      - Patrón de colas normal")
````

````python
print("\n" + "="*70)
print("📊 VISUALIZACIÓN INTEGRAL - POSICIÓN Y FORMA")
print("="*70)

# Crear dashboard completo
fig = plt.figure(figsize=(20, 15))
fig.suptitle('Dashboard Completo: Medidas de Posición y Forma', fontsize=16, y=0.98)

for i, var in enumerate(variables_numericas):
    datos = estudiantes_df[var]
    
    # Subplot 1: Histograma con estadísticas
    ax1 = plt.subplot(3, 3, i*3 + 1)
    
    # Histograma
    n, bins, patches = ax1.hist(datos, bins=20, alpha=0.7, color=f'C{i}', edgecolor='black')
    
    # Añadir líneas de estadísticas
    media = datos.mean()
    mediana = datos.median()
    ax1.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
    ax1.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
    
    # Añadir percentiles
    p25 = np.percentile(datos, 25)
    p75 = np.percentile(datos, 75)
    ax1.axvline(p25, color='orange', linestyle=':', alpha=0.7, label=f'Q1: {p25:.2f}')
    ax1.axvline(p75, color='orange', linestyle=':', alpha=0.7, label=f'Q3: {p75:.2f}')
    
    ax1.set_title(f'{var.replace("_", " ").title()}\nAsimetría: {resultados_asimetria[var]:.3f}, Curtosis: {resultados_curtosis[var]:.3f}')
    ax1.legend(fontsize=8)
    ax1.grid(True, alpha=0.3)
    
    # Subplot 2: Box plot con outliers
    ax2 = plt.subplot(3, 3, i*3 + 2)
    
    bp = ax2.boxplot([datos], patch_artist=True, labels=[var.replace("_", " ")])
    bp['boxes'][0].set_facecolor(f'C{i}')
    bp['boxes'][0].set_alpha(0.7)
    
    # Añadir media al boxplot
    ax2.plot(1, media, 'ro', markersize=8, label=f'Media: {media:.2f}')
    
    # Marcar outliers con Z-score
    z_scores = np.abs((datos - media) / datos.std())
    outliers_idx = z_scores > 2
    if outliers_idx.any():
        outliers_vals = datos[outliers_idx]
        ax2.scatter([1] * len(outliers_vals), outliers_vals, 
                   color='red', s=50, alpha=0.7, label=f'Outliers: {len(outliers_vals)}')
    
    ax2.set_title('Box Plot con Outliers')
    ax2.legend(fontsize=8)
    ax2.grid(True, alpha=0.3)
    
    # Subplot 3: Q-Q plot para evaluar normalidad
    ax3 = plt.subplot(3, 3, i*3 + 3)
    
    stats.probplot(datos, dist="norm", plot=ax3)
    ax3.set_title('Q-Q Plot (vs Normal)')
    ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

# ANÁLISIS ESTADÍSTICO COMPLETO CON LO VISTO HASTA LA SESION 4

````python
print("\n" + "="*70)
print("📋 ANÁLISIS ESTADÍSTICO COMPLETO - INTEGRACIÓN DE TODAS LAS MEDIDAS")
print("="*70)

def crear_perfil_estadistico(datos, nombre_variable):
    """Crear un perfil estadístico completo para una variable"""
    
    # Medidas de tendencia central
    media = datos.mean()
    mediana = datos.median()
    moda = datos.mode().iloc[0] if len(datos.mode()) > 0 else None
    
    # Medidas de dispersión
    std = datos.std()
    cv = (std / media) * 100
    rango = datos.max() - datos.min()
    iqr = datos.quantile(0.75) - datos.quantile(0.25)
    
    # Medidas de posición
    percentiles = [datos.quantile(p/100) for p in [5, 10, 25, 50, 75, 90, 95]]
    
    # Medidas de forma
    asimetria = skew(datos)
    curtosis_val = kurtosis(datos, fisher=True)
    
    # Outliers
    z_scores = np.abs((datos - media) / std)
    outliers_count = (z_scores > 2).sum()
    
    return {
        'nombre': nombre_variable,
        'n': len(datos),
        'media': media,
        'mediana': mediana,
        'moda': moda,
        'std': std,
        'cv': cv,
        'rango': rango,
        'iqr': iqr,
        'percentiles': percentiles,
        'asimetria': asimetria,
        'curtosis': curtosis_val,
        'outliers': outliers_count
    }

# Crear perfiles para todas las variables
perfiles = {}
for var in variables_numericas:
    perfiles[var] = crear_perfil_estadistico(estudiantes_df[var], var)

# Mostrar tabla resumen
print("📊 TABLA RESUMEN EJECUTIVA:")
print("-" * 90)
print(f"{'Variable':<15} {'Media':<8} {'Mediana':<8} {'CV%':<8} {'Asimetría':<10} {'Curtosis':<10} {'Outliers'}")
print("-" * 90)

for var in variables_numericas:
    p = perfiles[var]
    print(f"{var:<15} {p['media']:<8.2f} {p['mediana']:<8.2f} {p['cv']:<8.1f} "
          f"{p['asimetria']:<10.3f} {p['curtosis']:<10.3f} {p['outliers']}")

# Recomendaciones basadas en el perfil
print(f"\n💡 RECOMENDACIONES METODOLÓGICAS:")
print("-" * 50)

for var in variables_numericas:
    p = perfiles[var]
    print(f"\n📊 {var.upper().replace('_', ' ')}:")
    
    # Recomendación de medida central
    if abs(p['asimetria']) < 0.5:
        medida_central = "Media (distribución aproximadamente simétrica)"
    else:
        medida_central = "Mediana (distribución asimétrica)"
    
    # Recomendación de dispersión
    if p['outliers'] > len(estudiantes_df) * 0.05:  # Más del 5% outliers
        medida_dispersion = "IQR (presencia significativa de outliers)"
    else:
        medida_dispersion = "Desviación Estándar (pocos outliers)"
    
    # Evaluación de normalidad
    if abs(p['asimetria']) < 0.5 and abs(p['curtosis']) < 1:
        normalidad = "Aproximadamente normal (aplicar regla empírica)"
    else:
        normalidad = "No normal (cuidado con supuestos de normalidad)"
    
    print(f"   ✓ Medida central: {medida_central}")
    print(f"   ✓ Medida dispersión: {medida_dispersion}")
    print(f"   ✓ Normalidad: {normalidad}")
````
