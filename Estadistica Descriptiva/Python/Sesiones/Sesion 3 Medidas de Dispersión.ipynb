## 📖 **PARTE 1: ¿Por qué Importa la Dispersión?**

### **🤔 La Historia Incompleta de las Medidas Centrales**

Imagina dos clases de estudiantes:

**Clase A**: Notas = 70, 75, 80, 85, 90 → **Media = 80**  
**Clase B**: Notas = 50, 60, 80, 100, 110 → **Media = 80**

¿Son estas clases iguales? ¡Definitivamente NO!

- **Clase A**: Rendimiento consistente, todos cerca del promedio
- **Clase B**: Rendimiento muy variable, gran dispersión

### **🎯 Definición de Dispersión**

> Las **medidas de dispersión** (o variabilidad) nos indican qué tan **esparcidos** o **concentrados** están los datos alrededor de la medida de tendencia central.

### **💡 ¿Por qué es Importante la Dispersión?**

1. **Predicción**: Mayor dispersión = mayor incertidumbre
2. **Riesgo**: En finanzas, mayor variabilidad = mayor riesgo
3. **Control de Calidad**: Menor dispersión = mayor consistencia
4. **Toma de Decisiones**: Ayuda a evaluar la confiabilidad de los datos
5. **Comparación**: Permite comparar la variabilidad entre grupos

### **📊 Tipos de Medidas de Dispersión**

```
MEDIDAS DE DISPERSIÓN
├── SIMPLES
│   ├── Rango (Máximo - Mínimo)
│   └── Rango Intercuartílico (Q3 - Q1)
├── BASADAS EN LA MEDIA
│   ├── Varianza (σ² o s²)
│   └── Desviación Estándar (σ o s)
└── RELATIVAS
    └── Coeficiente de Variación (CV)
```

## 📊 **PARTE 2: Rango y Rango Intercuartílico**

### **📏 Rango Simple**

El **rango** es la medida de dispersión más simple: la diferencia entre el valor máximo y mínimo.

#### **Fórmula:**
```
Rango = Valor Máximo - Valor Mínimo
```

#### **💡 Ejemplo:**
**Temperaturas de la semana**: 18°C, 20°C, 22°C, 25°C, 28°C
- Máximo: 28°C
- Mínimo: 18°C
- **Rango: 28 - 18 = 10°C**

**Interpretación**: Las temperaturas varían en un rango de 10°C durante la semana.

#### **✅ Ventajas del Rango:**
1. **Fácil de calcular**: Solo necesitas dos valores
2. **Intuitivo**: Todos entienden "de X a Y"
3. **Rápido**: Ideal para análisis preliminares

#### **⚠️ Desventajas del Rango:**
1. **Sensible a outliers**: Un solo valor extremo lo afecta completamente
2. **No usa toda la información**: Ignora la distribución interna
3. **Poco informativo**: No dice cómo se distribuyen los datos internamente

### **📦 Rango Intercuartílico (IQR)**

El **rango intercuartílico** es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Representa el rango del 50% central de los datos.

#### **Fórmula:**
```
IQR = Q3 - Q1
```

#### **📋 Pasos para Calcularlo:**
1. **Ordenar** los datos de menor a mayor
2. **Encontrar Q1** (percentil 25): 25% de datos por debajo
3. **Encontrar Q3** (percentil 75): 75% de datos por debajo  
4. **Calcular IQR** = Q3 - Q1

#### **💡 Ejemplo Detallado:**
**Datos**: 12, 15, 18, 20, 22, 25, 28, 30, 35, 40

1. Ya están ordenados ✅
2. n = 10
3. **Q1**: Posición = 0.25 × (10 + 1) = 2.75 → Q1 = 16.5 (interpolando entre 15 y 18)
4. **Q3**: Posición = 0.75 × (10 + 1) = 8.25 → Q3 = 31.25 (interpolando entre 30 y 35)
5. **IQR = 31.25 - 16.5 = 14.75**

**Interpretación**: El 50% central de los datos tiene un rango de 14.75 unidades.

#### **✅ Ventajas del IQR:**
1. **Resistente a outliers**: Los valores extremos no lo afectan
2. **Más informativo que el rango**: Se enfoca en la parte central
3. **Útil para detectar outliers**: Base para la regla de 1.5 × IQR
4. **Robusto**: Funciona bien con distribuciones asimétricas

#### **⚠️ Desventajas del IQR:**
1. **Ignora el 50% de los datos**: Solo usa los cuartiles
2. **Menos preciso**: No tan sensible a cambios pequeños
3. **Cálculo más complejo**: Requiere encontrar cuartiles

## 📊 **PARTE 3: Varianza - El Corazón de la Dispersión**

### **🎯 Concepto Fundamental**

La **varianza** mide qué tan dispersos están los datos respecto a la media. Es el promedio de los cuadrados de las desviaciones respecto a la media.

> **Idea clave**: Si todos los datos fueran iguales a la media, la varianza sería cero. Cuanto más se alejen los datos de la media, mayor será la varianza.

### **📐 ¿Cómo se Construye la Varianza?**

#### **Paso a Paso Conceptual:**

1. **Calcular desviaciones**: Cada dato menos la media
   - `(xi - x̄)`

2. **¿Por qué elevar al cuadrado?**
   - Las desviaciones positivas y negativas se cancelan
   - Al cuadrado, todas se vuelven positivas
   - Da más peso a las desviaciones grandes

3. **Promediar los cuadrados**: Obtener la varianza

### **🔢 Fórmulas de la Varianza**

#### **Varianza Poblacional (σ²):**
```
σ² = Σ(xi - μ)² / N
```

#### **Varianza Muestral (s²):**
```
s² = Σ(xi - x̄)² / (n-1)
```

**¿Por qué n-1 en la muestra?** ¡Grados de libertad!
- Al calcular la media muestral, "perdemos" un grado de libertad
- n-1 corrige este sesgo y hace que s² sea un mejor estimador de σ²

### **💡 Ejemplo Paso a Paso**

**Datos**: Horas de sueño: 6, 7, 8, 7, 9

#### **Paso 1**: Calcular la media
```
x̄ = (6 + 7 + 8 + 7 + 9) ÷ 5 = 37 ÷ 5 = 7.4 horas
```

#### **Paso 2**: Calcular desviaciones
```
x₁ - x̄ = 6 - 7.4 = -1.4
x₂ - x̄ = 7 - 7.4 = -0.4
x₃ - x̄ = 8 - 7.4 = 0.6
x₄ - x̄ = 7 - 7.4 = -0.4  
x₅ - x̄ = 9 - 7.4 = 1.6
```

#### **Paso 3**: Elevar al cuadrado
```
(-1.4)² = 1.96
(-0.4)² = 0.16
(0.6)² = 0.36
(-0.4)² = 0.16
(1.6)² = 2.56
```

#### **Paso 4**: Sumar y dividir
```
Suma = 1.96 + 0.16 + 0.36 + 0.16 + 2.56 = 5.20
s² = 5.20 ÷ (5-1) = 5.20 ÷ 4 = 1.30 (horas)²
```

### **🤔 Interpretación de la Varianza**

- **Varianza = 1.30 (horas)²**
- **Problema**: Las unidades están al cuadrado
- **Solución**: La desviación estándar (raíz de la varianza)

## 📊 **PARTE 4: Desviación Estándar - La Medida Más Importante**

### **🎯 Definición**

La **desviación estándar** es la raíz cuadrada positiva de la varianza. Representa la dispersión típica de los datos respecto a la media, en las mismas unidades originales.

### **🔢 Fórmulas**

#### **Desviación Estándar Poblacional (σ):**
```
σ = √σ² = √[Σ(xi - μ)² / N]
```

#### **Desviación Estándar Muestral (s):**
```
s = √s² = √[Σ(xi - x̄)² / (n-1)]
```

### **💡 Continuando Nuestro Ejemplo**
```
s = √1.30 = 1.14 horas
```

**Interpretación**: En promedio, las horas de sueño se desvían 1.14 horas respecto a la media de 7.4 horas.

### **📊 La Regla Empírica (68-95-99.7)**

Para **distribuciones aproximadamente normales** (en forma de campana):

```
📊 REGLA EMPÍRICA:
├── 68% de los datos están dentro de 1σ de la media [μ ± σ]
├── 95% de los datos están dentro de 2σ de la media [μ ± 2σ]  
└── 99.7% de los datos están dentro de 3σ de la media [μ ± 3σ]
```

#### **💡 Ejemplo con Alturas**
Si las alturas tienen media = 170 cm y σ = 10 cm:

- **68%** de las personas miden entre **160-180 cm** (170 ± 10)
- **95%** de las personas miden entre **150-190 cm** (170 ± 20)
- **99.7%** de las personas miden entre **140-200 cm** (170 ± 30)

### **✅ Ventajas de la Desviación Estándar**

1. **Mismas unidades**: Interpretación directa en unidades originales
2. **Sensible**: Refleja cambios en la variabilidad
3. **Matemáticamente útil**: Base para muchos cálculos estadísticos
4. **Regla empírica**: Permite estimaciones en distribuciones normales
5. **Ampliamente utilizada**: Estándar en literatura científica

### **⚠️ Limitaciones de la Desviación Estándar**

1. **Sensible a outliers**: Valores extremos la aumentan mucho
2. **Asume normalidad**: La regla empírica solo aplica a distribuciones normales
3. **Difícil de comparar**: Entre variables con diferentes escalas o unidades

## 📊 **PARTE 5: Coeficiente de Variación - La Medida Relativa**

### **🎯 El Problema de Comparación**

¿Qué es más variable?
- **Grupo A**: Salarios con σ = $5,000 y media = $50,000
- **Grupo B**: Edades con σ = 8 años y media = 25 años

¡No podemos comparar directamente! Necesitamos una **medida relativa**.

### **🔢 Coeficiente de Variación (CV)**

El **coeficiente de variación** expresa la desviación estándar como porcentaje de la media:

#### **Fórmula:**
```
CV = (σ/μ) × 100    (población)
CV = (s/x̄) × 100    (muestra)
```

### **💡 Resolviendo el Ejemplo**

**Grupo A (Salarios):**
```
CV_A = (5,000 ÷ 50,000) × 100 = 10%
```

**Grupo B (Edades):**
```
CV_B = (8 ÷ 25) × 100 = 32%
```

**Conclusión**: Las edades son más variables relativamente (32% vs 10%).

### **🎯 Interpretación del Coeficiente de Variación**

#### **Guía General:**
- **CV < 10%**: Variabilidad baja (datos homogéneos)
- **CV 10-20%**: Variabilidad moderada
- **CV 20-30%**: Variabilidad alta
- **CV > 30%**: Variabilidad muy alta (datos heterogéneos)

#### **Criterios de Calidad (según DANE):**
- **CV ≤ 7%**: Estimaciones precisas
- **CV 8-14%**: Estimaciones con precisión aceptable
- **CV 15-20%**: Precisión regular
- **CV > 20%**: Estimación poco precisa

### **✅ Usos del Coeficiente de Variación**

1. **Comparar variabilidad** entre variables de diferentes escalas
2. **Evaluar consistencia** en procesos de producción
3. **Análisis de riesgo** en inversiones (mayor CV = mayor riesgo)
4. **Control de calidad** en mediciones
5. **Selección de métodos** de predicción más estables

### **⚠️ Limitaciones del CV**

1. **No definido** cuando la media es cero
2. **Problemático** cuando la media es negativa
3. **Menos útil** para variables que pueden ser cero (como errores)

## 📊 **PARTE 7: Relación entre Medidas de Dispersión**

### **🔗 Conexiones Importantes**

#### **1. Varianza y Desviación Estándar**
```
σ = √σ²    o    s = √s²
```
- Siempre relacionadas por la raíz cuadrada
- Misma información, diferentes unidades

#### **2. Desviación Estándar y CV**
```
CV = (σ/μ) × 100
```
- CV relativiza la desviación estándar
- Permite comparaciones entre variables diferentes

#### **3. IQR y Desviación Estándar**
```
Para distribuciones normales: IQR ≈ 1.35 × σ
```
- Ambas miden dispersión
- IQR más robusto, σ más sensible

### **📊 Cuándo Usar Cada Medida**

| Situación | Medida Recomendada | Razón |
|-----------|-------------------|-------|
| **Datos con outliers** | **IQR** | Robusta a valores extremos |
| **Distribución normal** | **Desviación Estándar** | Permite usar regla empírica |
| **Comparar variables diferentes** | **Coeficiente de Variación** | Medida relativa |
| **Análisis rápido** | **Rango** | Simple y directo |
| **Cálculos estadísticos posteriores** | **Varianza/Desv. Estándar** | Base matemática sólida |

## 🧠 **PARTE 8: Conceptos Avanzados de Dispersión**

### **📊 Desviación Absoluta Media (MAD)**

Una alternativa robusta a la desviación estándar:

```
MAD = Σ|xi - x̄| / n
```

**Ventajas**: Menos sensible a outliers que la desviación estándar.

### **📊 Dispersión vs. Forma**

La dispersión interactúa con la forma de la distribución:

- **Distribuciones simétricas**: σ y IQR dan información similar
- **Distribuciones asimétricas**: IQR puede ser más representativo
- **Distribuciones con colas pesadas**: σ puede sobrestimar la dispersión típica

### **📊 Regla de Chebyshev**

Para **cualquier distribución** (no solo normal):

> Al menos (1 - 1/k²) de los datos están dentro de k desviaciones estándar de la media.

**Ejemplos**:
- k = 2: Al menos 75% dentro de μ ± 2σ
- k = 3: Al menos 89% dentro de μ ± 3σ

# CÁLCULO MANUAL VS AUTOMÁTICO - COMPRENDIENDO LAS FÓRMULAS

````python
print("\n" + "="*70)
print("🧮 CÁLCULO MANUAL DE MEDIDAS DE DISPERSIÓN")
print("="*70)

# Dataset pequeño para cálculo manual paso a paso
horas_sueno = [6, 7, 8, 7, 9]
print(f"📊 Datos de ejemplo (horas de sueño): {horas_sueno}")

# PASO 1: Calcular la media
media_manual = sum(horas_sueno) / len(horas_sueno)
print(f"\n1️⃣ MEDIA:")
print(f"   Media = {sum(horas_sueno)} ÷ {len(horas_sueno)} = {media_manual:.2f} horas")
````

````python
# PASO 2: Calcular desviaciones
print(f"\n2️⃣ DESVIACIONES (xi - x̄):")
desviaciones = []
for i, valor in enumerate(horas_sueno):
    desv = valor - media_manual
    desviaciones.append(desv)
    print(f"   x{i+1} - x̄ = {valor} - {media_manual:.2f} = {desv:.2f}")

# Verificar que suman cero
suma_desviaciones = sum(desviaciones)
print(f"   Suma de desviaciones = {suma_desviaciones:.10f} ≈ 0 ✅")
````

````python
# PASO 3: Calcular cuadrados de desviaciones
print(f"\n3️⃣ CUADRADOS DE DESVIACIONES (xi - x̄)²:")
cuadrados_desv = []
for i, desv in enumerate(desviaciones):
    cuadrado = desv ** 2
    cuadrados_desv.append(cuadrado)
    print(f"   ({desv:.2f})² = {cuadrado:.4f}")

suma_cuadrados = sum(cuadrados_desv)
print(f"   Suma de cuadrados = {suma_cuadrados:.4f}")
````

````python
# PASO 4: Calcular varianza (muestral)
n = len(horas_sueno)
varianza_manual = suma_cuadrados / (n - 1)
print(f"\n4️⃣ VARIANZA MUESTRAL:")
print(f"   s² = {suma_cuadrados:.4f} ÷ ({n} - 1) = {suma_cuadrados:.4f} ÷ {n-1} = {varianza_manual:.4f} (horas)²")
````

````python
# PASO 5: Calcular desviación estándar
desv_std_manual = varianza_manual ** 0.5
print(f"\n5️⃣ DESVIACIÓN ESTÁNDAR:")
print(f"   s = √{varianza_manual:.4f} = {desv_std_manual:.4f} horas")
````

````python
# PASO 6: Calcular coeficiente de variación
cv_manual = (desv_std_manual / media_manual) * 100
print(f"\n6️⃣ COEFICIENTE DE VARIACIÓN:")
print(f"   CV = ({desv_std_manual:.4f} ÷ {media_manual:.2f}) × 100 = {cv_manual:.2f}%")
````

````python
# Comparar con cálculos automáticos de Python
print(f"\n🔍 VERIFICACIÓN CON PYTHON:")
datos_series = pd.Series(horas_sueno)
print(f"   Media - pandas: {datos_series.mean():.4f}")
print(f"   Varianza - pandas: {datos_series.var():.4f}")
print(f"   Desv.Std - pandas: {datos_series.std():.4f}")
print(f"   CV calculado: {(datos_series.std()/datos_series.mean()*100):.2f}%")
print(f"   ✅ ¡Todos los cálculos coinciden!")
````

# COMPARACIÓN DE MEDIDAS DE DISPERSIÓN

````python
print("\n" + "="*70)
print("📏 COMPARACIÓN ENTRE DIFERENTES MEDIDAS DE DISPERSIÓN")
print("="*70)

# Crear dataset más grande para análisis completo
np.random.seed(42)
datos_ejemplo = np.random.normal(100, 15, 50)  # Media=100, Desv.Std=15
datos_ejemplo = np.round(datos_ejemplo, 1)

print(f"📊 Dataset de ejemplo (50 valores): Media≈100, σ≈15")
print(f"   Primeros 10 valores: {datos_ejemplo[:10]}")

# Calcular todas las medidas de dispersión
rango = np.max(datos_ejemplo) - np.min(datos_ejemplo)
q1 = np.percentile(datos_ejemplo, 25)
q3 = np.percentile(datos_ejemplo, 75)
iqr = q3 - q1
varianza = np.var(datos_ejemplo, ddof=1)
desv_std = np.std(datos_ejemplo, ddof=1)
media = np.mean(datos_ejemplo)
cv = (desv_std / media) * 100

print(f"\n📋 RESUMEN DE MEDIDAS DE DISPERSIÓN:")
print(f"{'Medida':<20} {'Valor':<12} {'Interpretación'}")
print("-" * 55)
print(f"{'Rango':<20} {rango:<12.2f} Diferencia máx-mín")
print(f"{'Rango Intercuartil':<20} {iqr:<12.2f} Rango del 50% central")
print(f"{'Varianza':<20} {varianza:<12.2f} (unidades)²")
print(f"{'Desviación Estándar':<20} {desv_std:<12.2f} Dispersión típica")
print(f"{'Coef. Variación':<20} {cv:<12.2f}% Dispersión relativa")
````

# IMPACTO DE OUTLIERS EN MEDIDAS DE DISPERSIÓN

````python
print("\n" + "="*70)
print("🎯 IMPACTO DE OUTLIERS EN MEDIDAS DE DISPERSIÓN")
print("="*70)

# Datos normales sin outliers
datos_normales = [95, 98, 100, 102, 105, 97, 103, 99, 101, 96]
print(f"📊 Datos NORMALES: {datos_normales}")

# Calcular medidas sin outliers
rango_normal = max(datos_normales) - min(datos_normales)
q1_normal = np.percentile(datos_normales, 25)
q3_normal = np.percentile(datos_normales, 75)
iqr_normal = q3_normal - q1_normal
std_normal = np.std(datos_normales, ddof=1)
cv_normal = (std_normal / np.mean(datos_normales)) * 100

print(f"   Rango: {rango_normal:.1f}")
print(f"   IQR: {iqr_normal:.1f}")
print(f"   Desv.Std: {std_normal:.2f}")
print(f"   CV: {cv_normal:.1f}%")

# Agregar outliers extremos
datos_con_outliers = datos_normales + [150, 200]  # Agregar dos outliers
print(f"\n📊 Datos CON OUTLIERS: {datos_con_outliers}")

# Calcular medidas con outliers
rango_outlier = max(datos_con_outliers) - min(datos_con_outliers)
q1_outlier = np.percentile(datos_con_outliers, 25)
q3_outlier = np.percentile(datos_con_outliers, 75)
iqr_outlier = q3_outlier - q1_outlier
std_outlier = np.std(datos_con_outliers, ddof=1)
cv_outlier = (std_outlier / np.mean(datos_con_outliers)) * 100

print(f"   Rango: {rango_outlier:.1f}")
print(f"   IQR: {iqr_outlier:.1f}")
print(f"   Desv.Std: {std_outlier:.2f}")
print(f"   CV: {cv_outlier:.1f}%")

# Análisis del impacto
print(f"\n📈 IMPACTO DE LOS OUTLIERS:")
cambio_rango = ((rango_outlier - rango_normal) / rango_normal) * 100
cambio_iqr = ((iqr_outlier - iqr_normal) / iqr_normal) * 100
cambio_std = ((std_outlier - std_normal) / std_normal) * 100

print(f"   Rango cambió: +{cambio_rango:.0f}% 📈 ¡MUY AFECTADO!")
print(f"   IQR cambió: +{cambio_iqr:.1f}% 📊 Prácticamente sin cambio")
print(f"   Desv.Std cambió: +{cambio_std:.0f}% 📈 ¡MUY AFECTADA!")

print(f"\n💡 CONCLUSIÓN: El IQR es la medida más ROBUSTA ante outliers")

# Visualización del impacto
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Box plot sin outliers
bp1 = ax1.boxplot([datos_normales], labels=['Sin Outliers'], patch_artist=True)
bp1['boxes'][0].set_facecolor('lightblue')
ax1.set_title('Distribución SIN Outliers')
ax1.set_ylabel('Valores')
ax1.grid(True, alpha=0.3)
ax1.text(0.7, max(datos_normales)-1, f'IQR: {iqr_normal:.1f}\nStd: {std_normal:.2f}', 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))

# Box plot con outliers
bp2 = ax2.boxplot([datos_con_outliers], labels=['Con Outliers'], patch_artist=True)
bp2['boxes'][0].set_facecolor('lightcoral')
ax2.set_title('Distribución CON Outliers')
ax2.set_ylabel('Valores')
ax2.grid(True, alpha=0.3)
ax2.text(0.7, max(datos_con_outliers)-20, f'IQR: {iqr_outlier:.1f}\nStd: {std_outlier:.2f}', 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral"))

plt.tight_layout()
plt.show()
````

# COEFICIENTE DE VARIACIÓN PARA COMPARACIONES

````python
print("\n" + "="*70)
print("⚖️ COEFICIENTE DE VARIACIÓN - COMPARANDO VARIABILIDAD")
print("="*70)

# Crear datasets para comparación
np.random.seed(42)

# Tres grupos con diferentes características
grupo_salarios = np.random.normal(50000, 10000, 100)  # Salarios en pesos
grupo_edades = np.random.normal(30, 8, 100)          # Edades en años  
grupo_alturas = np.random.normal(170, 10, 100)       # Alturas en cm

# Asegurar valores positivos y realistas
grupo_salarios = np.abs(grupo_salarios)
grupo_edades = np.clip(grupo_edades, 18, 65)
grupo_alturas = np.clip(grupo_alturas, 150, 200)

datasets = {
    'Salarios (pesos)': grupo_salarios,
    'Edades (años)': grupo_edades,
    'Alturas (cm)': grupo_alturas
}

print("📊 ANÁLISIS COMPARATIVO DE VARIABILIDAD:")
print("-" * 80)
print(f"{'Variable':<18} {'Media':<12} {'Desv.Std':<12} {'CV (%)':<10} {'Interpretación'}")
print("-" * 80)

resultados_cv = {}
for nombre, datos in datasets.items():
    media = np.mean(datos)
    std = np.std(datos, ddof=1)
    cv = (std / media) * 100
    
    # Interpretación del CV
    if cv < 10:
        interpretacion = "Baja variabilidad"
    elif cv < 20:
        interpretacion = "Moderada variab."
    elif cv < 30:
        interpretacion = "Alta variabilidad"
    else:
        interpretacion = "Muy alta variab."
    
    resultados_cv[nombre] = cv
    print(f"{nombre:<18} {media:<12.0f} {std:<12.2f} {cv:<10.1f} {interpretacion}")

# Identificar el más y menos variable
mas_variable = max(resultados_cv, key=resultados_cv.get)
menos_variable = min(resultados_cv, key=resultados_cv.get)

print(f"\n📈 CONCLUSIONES:")
print(f"   Más variable: {mas_variable} (CV = {resultados_cv[mas_variable]:.1f}%)")
print(f"   Menos variable: {menos_variable} (CV = {resultados_cv[menos_variable]:.1f}%)")

# Visualización comparativa
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Comparación de Variabilidad usando Coeficiente de Variación', fontsize=16)

nombres = list(datasets.keys())
for i, (nombre, datos) in enumerate(datasets.items()):
    # Histogramas (fila superior)
    ax1 = axes[0, i]
    ax1.hist(datos, bins=20, alpha=0.7, color=f'C{i}', edgecolor='black')
    media = np.mean(datos)
    std = np.std(datos, ddof=1)
    cv = (std / media) * 100
    
    ax1.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.0f}')
    ax1.axvline(media-std, color='orange', linestyle=':', alpha=0.7, label=f'±1σ')
    ax1.axvline(media+std, color='orange', linestyle=':', alpha=0.7)
    
    ax1.set_title(f'{nombre}\nCV = {cv:.1f}%')
    ax1.legend(fontsize=8)
    ax1.grid(True, alpha=0.3)
    
    # Box plots (fila inferior)  
    ax2 = axes[1, i]
    bp = ax2.boxplot([datos], patch_artist=True)
    bp['boxes'][0].set_facecolor(f'C{i}')
    bp['boxes'][0].set_alpha(0.7)
    ax2.set_title(f'Dispersión Relativa')
    ax2.set_xticklabels([f'{cv:.1f}%'])
    ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
