## ğŸ“Š **PARTE 2: Rango y Rango IntercuartÃ­lico**

### **ğŸ“ Rango Simple**

El **rango** es la medida de dispersiÃ³n mÃ¡s simple: la diferencia entre el valor mÃ¡ximo y mÃ­nimo.

#### **FÃ³rmula:**

**InterpretaciÃ³n**: Las temperaturas varÃ­an en un rango de 10Â°C durante la semana.

#### **âœ… Ventajas del Rango:**
1. **FÃ¡cil de calcular**: Solo necesitas dos valores

#### **âš ï¸ Desventajas del Rango:**

### **ğŸ“¦ Rango IntercuartÃ­lico (IQR)**

El **rango intercuartÃ­lico** es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Representa el rango del 50% central de los datos.

#### **FÃ³rmula:**

#### **ğŸ“‹ Pasos para Calcularlo:**
1. **Ordenar** los datos de menor a mayor
2. **Encontrar Q1** (percentil 25): 25% de datos por debajo
3. **Encontrar Q3** (percentil 75): 75% de datos por debajo  
4. **Calcular IQR** = Q3 - Q1

#### **ğŸ’¡ Ejemplo Detallado:**
**Datos**: 12, 15, 18, 20, 22, 25, 28, 30, 35, 40

1. Ya estÃ¡n ordenados âœ…
2. n = 10
3. **Q1**: PosiciÃ³n = 0.25 Ã— (10 + 1) = 2.75 â†’ Q1 = 16.5 (interpolando entre 15 y 18)
4. **Q3**: PosiciÃ³n = 0.75 Ã— (10 + 1) = 8.25 â†’ Q3 = 31.25 (interpolando entre 30 y 35)
5. **IQR = 31.25 - 16.5 = 14.75**

**InterpretaciÃ³n**: El 50% central de los datos tiene un rango de 14.75 unidades.

#### **âœ… Ventajas del IQR:**
1. **Resistente a outliers**: Los valores extremos no lo afectan
2. **MÃ¡s informativo que el rango**: Se enfoca en la parte central
3. **Ãštil para detectar outliers**: Base para la regla de 1.5 Ã— IQR
4. **Robusto**: Funciona bien con distribuciones asimÃ©tricas

#### **âš ï¸ Desventajas del IQR:**
1. **Ignora el 50% de los datos**: Solo usa los cuartiles
2. **Menos preciso**: No tan sensible a cambios pequeÃ±os
3. **CÃ¡lculo mÃ¡s complejo**: Requiere encontrar cuartiles

## ğŸ“Š **PARTE 3: Varianza - El CorazÃ³n de la DispersiÃ³n**

### **ğŸ¯ Concepto Fundamental**

La **varianza** mide quÃ© tan dispersos estÃ¡n los datos respecto a la media. Es el promedio de los cuadrados de las desviaciones respecto a la media.

> **Idea clave**: Si todos los datos fueran iguales a la media, la varianza serÃ­a cero. Cuanto mÃ¡s se alejen los datos de la media, mayor serÃ¡ la varianza.

### **ğŸ“ Â¿CÃ³mo se Construye la Varianza?**

#### **Paso a Paso Conceptual:**

1. **Calcular desviaciones**: Cada dato menos la media
   - `(xi - xÌ„)`

2. **Â¿Por quÃ© elevar al cuadrado?**
   - Las desviaciones positivas y negativas se cancelan
   - Al cuadrado, todas se vuelven positivas
   - Da mÃ¡s peso a las desviaciones grandes

3. **Promediar los cuadrados**: Obtener la varianza

### **ğŸ”¢ FÃ³rmulas de la Varianza**

#### **Varianza Poblacional (ÏƒÂ²):**
```
ÏƒÂ² = Î£(xi - Î¼)Â² / N
```

#### **Varianza Muestral (sÂ²):**

**Â¿Por quÃ© n-1 en la muestra?** Â¡Grados de libertad!
- Al calcular la media muestral, "perdemos" un grado de libertad
- n-1 corrige este sesgo y hace que sÂ² sea un mejor estimador de ÏƒÂ²

### **ğŸ’¡ Ejemplo Paso a Paso**

**Datos**: Horas de sueÃ±o: 6, 7, 8, 7, 9

#### **Paso 1**: Calcular la media

#### **Paso 2**: Calcular desviaciones

#### **Paso 3**: Elevar al cuadrado

### **ğŸ¤” InterpretaciÃ³n de la Varianza**

## ğŸ“Š **PARTE 5: Coeficiente de VariaciÃ³n - La Medida Relativa**

### **ğŸ¯ El Problema de ComparaciÃ³n**

Â¿QuÃ© es mÃ¡s variable?
- **Grupo A**: Salarios con Ïƒ = $5,000 y media = $50,000
- **Grupo B**: Edades con Ïƒ = 8 aÃ±os y media = 25 aÃ±os

Â¡No podemos comparar directamente! Necesitamos una **medida relativa**.

### **ğŸ”¢ Coeficiente de VariaciÃ³n (CV)**

El **coeficiente de variaciÃ³n** expresa la desviaciÃ³n estÃ¡ndar como porcentaje de la media:

#### **FÃ³rmula:**

### **ğŸ’¡ Resolviendo el Ejemplo**

**Grupo B (Edades):**

### **ğŸ¯ InterpretaciÃ³n del Coeficiente de VariaciÃ³n**

#### **GuÃ­a General:**
- **CV < 10%**: Variabilidad baja (datos homogÃ©neos)

#### **Criterios de Calidad (segÃºn DANE):**
- **CV â‰¤ 7%**: Estimaciones precisas

- **CV 15-20%**: PrecisiÃ³n regular
- **CV > 20%**: EstimaciÃ³n poco precisa

### **âœ… Usos del Coeficiente de VariaciÃ³n**

1. **Comparar variabilidad** entre variables de diferentes escalas
2. **Evaluar consistencia** en procesos de producciÃ³n
3. **AnÃ¡lisis de riesgo** en inversiones (mayor CV = mayor riesgo)
4. **Control de calidad** en mediciones
5. **SelecciÃ³n de mÃ©todos** de predicciÃ³n mÃ¡s estables

### **âš ï¸ Limitaciones del CV**

1. **No definido** cuando la media es cero
2. **ProblemÃ¡tico** cuando la media es negativa
3. **Menos Ãºtil** para variables que pueden ser cero (como errores)

# CÃLCULO MANUAL VS AUTOMÃTICO - COMPRENDIENDO LAS FÃ“RMULAS

````python
print("\n" + "="*70)
print("ğŸ§® CÃLCULO MANUAL DE MEDIDAS DE DISPERSIÃ“N")
print("="*70)

# Dataset pequeÃ±o para cÃ¡lculo manual paso a paso
horas_sueno = [6, 7, 8, 7, 9]
print(f"ğŸ“Š Datos de ejemplo (horas de sueÃ±o): {horas_sueno}")

# PASO 1: Calcular la media
media_manual = sum(horas_sueno) / len(horas_sueno)
print(f"\n1ï¸âƒ£ MEDIA:")
print(f"   Media = {sum(horas_sueno)} Ã· {len(horas_sueno)} = {media_manual:.2f} horas")
````

````python
# PASO 2: Calcular desviaciones
print(f"\n2ï¸âƒ£ DESVIACIONES (xi - xÌ„):")
desviaciones = []
for i, valor in enumerate(horas_sueno):
    desv = valor - media_manual
    desviaciones.append(desv)
    print(f"   x{i+1} - xÌ„ = {valor} - {media_manual:.2f} = {desv:.2f}")

# Verificar que suman cero
suma_desviaciones = sum(desviaciones)
print(f"   Suma de desviaciones = {suma_desviaciones:.10f} â‰ˆ 0 âœ…")
````

````python
# PASO 3: Calcular cuadrados de desviaciones
print(f"\n3ï¸âƒ£ CUADRADOS DE DESVIACIONES (xi - xÌ„)Â²:")
cuadrados_desv = []
for i, desv in enumerate(desviaciones):
    cuadrado = desv ** 2
    cuadrados_desv.append(cuadrado)
    print(f"   ({desv:.2f})Â² = {cuadrado:.4f}")

suma_cuadrados = sum(cuadrados_desv)
print(f"   Suma de cuadrados = {suma_cuadrados:.4f}")
````

````python
# PASO 4: Calcular varianza (muestral)
n = len(horas_sueno)
varianza_manual = suma_cuadrados / (n - 1)
print(f"\n4ï¸âƒ£ VARIANZA MUESTRAL:")
print(f"   sÂ² = {suma_cuadrados:.4f} Ã· ({n} - 1) = {suma_cuadrados:.4f} Ã· {n-1} = {varianza_manual:.4f} (horas)Â²")
````

````python
# PASO 5: Calcular desviaciÃ³n estÃ¡ndar
desv_std_manual = varianza_manual ** 0.5
print(f"\n5ï¸âƒ£ DESVIACIÃ“N ESTÃNDAR:")
print(f"   s = âˆš{varianza_manual:.4f} = {desv_std_manual:.4f} horas")
````

````python
# PASO 6: Calcular coeficiente de variaciÃ³n
cv_manual = (desv_std_manual / media_manual) * 100
print(f"\n6ï¸âƒ£ COEFICIENTE DE VARIACIÃ“N:")
print(f"   CV = ({desv_std_manual:.4f} Ã· {media_manual:.2f}) Ã— 100 = {cv_manual:.2f}%")
````

````python
# Comparar con cÃ¡lculos automÃ¡ticos de Python
print(f"\nğŸ” VERIFICACIÃ“N CON PYTHON:")
datos_series = pd.Series(horas_sueno)
print(f"   Media - pandas: {datos_series.mean():.4f}")
print(f"   Varianza - pandas: {datos_series.var():.4f}")
print(f"   Desv.Std - pandas: {datos_series.std():.4f}")
print(f"   CV calculado: {(datos_series.std()/datos_series.mean()*100):.2f}%")
print(f"   âœ… Â¡Todos los cÃ¡lculos coinciden!")
````

# IMPACTO DE OUTLIERS EN MEDIDAS DE DISPERSIÃ“N

````python
print("\n" + "="*70)
print("ğŸ¯ IMPACTO DE OUTLIERS EN MEDIDAS DE DISPERSIÃ“N")
print("="*70)

# Datos normales sin outliers
datos_normales = [95, 98, 100, 102, 105, 97, 103, 99, 101, 96]
print(f"ğŸ“Š Datos NORMALES: {datos_normales}")

# Calcular medidas sin outliers
rango_normal = max(datos_normales) - min(datos_normales)
q1_normal = np.percentile(datos_normales, 25)
q3_normal = np.percentile(datos_normales, 75)
iqr_normal = q3_normal - q1_normal
std_normal = np.std(datos_normales, ddof=1)
cv_normal = (std_normal / np.mean(datos_normales)) * 100

print(f"   Rango: {rango_normal:.1f}")
print(f"   IQR: {iqr_normal:.1f}")
print(f"   Desv.Std: {std_normal:.2f}")
print(f"   CV: {cv_normal:.1f}%")

# Agregar outliers extremos
datos_con_outliers = datos_normales + [150, 200]  # Agregar dos outliers
print(f"\nğŸ“Š Datos CON OUTLIERS: {datos_con_outliers}")

# Calcular medidas con outliers
rango_outlier = max(datos_con_outliers) - min(datos_con_outliers)
q1_outlier = np.percentile(datos_con_outliers, 25)
q3_outlier = np.percentile(datos_con_outliers, 75)
iqr_outlier = q3_outlier - q1_outlier
std_outlier = np.std(datos_con_outliers, ddof=1)
cv_outlier = (std_outlier / np.mean(datos_con_outliers)) * 100

print(f"   Rango: {rango_outlier:.1f}")
print(f"   IQR: {iqr_outlier:.1f}")
print(f"   Desv.Std: {std_outlier:.2f}")
print(f"   CV: {cv_outlier:.1f}%")

# AnÃ¡lisis del impacto
print(f"\nğŸ“ˆ IMPACTO DE LOS OUTLIERS:")
cambio_rango = ((rango_outlier - rango_normal) / rango_normal) * 100
cambio_iqr = ((iqr_outlier - iqr_normal) / iqr_normal) * 100
cambio_std = ((std_outlier - std_normal) / std_normal) * 100

print(f"   Rango cambiÃ³: +{cambio_rango:.0f}% ğŸ“ˆ Â¡MUY AFECTADO!")
print(f"   IQR cambiÃ³: +{cambio_iqr:.1f}% ğŸ“Š PrÃ¡cticamente sin cambio")
print(f"   Desv.Std cambiÃ³: +{cambio_std:.0f}% ğŸ“ˆ Â¡MUY AFECTADA!")

print(f"\nğŸ’¡ CONCLUSIÃ“N: El IQR es la medida mÃ¡s ROBUSTA ante outliers")

# VisualizaciÃ³n del impacto
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Box plot sin outliers
bp1 = ax1.boxplot([datos_normales], labels=['Sin Outliers'], patch_artist=True)
bp1['boxes'][0].set_facecolor('lightblue')
ax1.set_title('DistribuciÃ³n SIN Outliers')
ax1.set_ylabel('Valores')
ax1.grid(True, alpha=0.3)
ax1.text(0.7, max(datos_normales)-1, f'IQR: {iqr_normal:.1f}\nStd: {std_normal:.2f}', 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))

# Box plot con outliers
bp2 = ax2.boxplot([datos_con_outliers], labels=['Con Outliers'], patch_artist=True)
bp2['boxes'][0].set_facecolor('lightcoral')
ax2.set_title('DistribuciÃ³n CON Outliers')
ax2.set_ylabel('Valores')
ax2.grid(True, alpha=0.3)
ax2.text(0.7, max(datos_con_outliers)-20, f'IQR: {iqr_outlier:.1f}\nStd: {std_outlier:.2f}', 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral"))

plt.tight_layout()
plt.show()
````

# COEFICIENTE DE VARIACIÃ“N PARA COMPARACIONES

````python
print("\n" + "="*70)
print("âš–ï¸ COEFICIENTE DE VARIACIÃ“N - COMPARANDO VARIABILIDAD")
print("="*70)

# Crear datasets para comparaciÃ³n
np.random.seed(42)

# Tres grupos con diferentes caracterÃ­sticas
grupo_salarios = np.random.normal(50000, 10000, 100)  # Salarios en pesos
grupo_edades = np.random.normal(30, 8, 100)          # Edades en aÃ±os  
grupo_alturas = np.random.normal(170, 10, 100)       # Alturas en cm

# Asegurar valores positivos y realistas
grupo_salarios = np.abs(grupo_salarios)
grupo_edades = np.clip(grupo_edades, 18, 65)
grupo_alturas = np.clip(grupo_alturas, 150, 200)

datasets = {
    'Salarios (pesos)': grupo_salarios,
    'Edades (aÃ±os)': grupo_edades,
    'Alturas (cm)': grupo_alturas
}

print("ğŸ“Š ANÃLISIS COMPARATIVO DE VARIABILIDAD:")
print("-" * 80)
print(f"{'Variable':<18} {'Media':<12} {'Desv.Std':<12} {'CV (%)':<10} {'InterpretaciÃ³n'}")
print("-" * 80)

resultados_cv = {}
for nombre, datos in datasets.items():
    media = np.mean(datos)
    std = np.std(datos, ddof=1)
    cv = (std / media) * 100
    
    # InterpretaciÃ³n del CV
    if cv < 10:
        interpretacion = "Baja variabilidad"
    elif cv < 20:
        interpretacion = "Moderada variab."
    elif cv < 30:
        interpretacion = "Alta variabilidad"
    else:
        interpretacion = "Muy alta variab."
    
    resultados_cv[nombre] = cv
    print(f"{nombre:<18} {media:<12.0f} {std:<12.2f} {cv:<10.1f} {interpretacion}")

# Identificar el mÃ¡s y menos variable
mas_variable = max(resultados_cv, key=resultados_cv.get)
menos_variable = min(resultados_cv, key=resultados_cv.get)

print(f"\nğŸ“ˆ CONCLUSIONES:")
print(f"   MÃ¡s variable: {mas_variable} (CV = {resultados_cv[mas_variable]:.1f}%)")
print(f"   Menos variable: {menos_variable} (CV = {resultados_cv[menos_variable]:.1f}%)")

# VisualizaciÃ³n comparativa
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('ComparaciÃ³n de Variabilidad usando Coeficiente de VariaciÃ³n', fontsize=16)

nombres = list(datasets.keys())
for i, (nombre, datos) in enumerate(datasets.items()):
    # Histogramas (fila superior)
    ax1 = axes[0, i]
    ax1.hist(datos, bins=20, alpha=0.7, color=f'C{i}', edgecolor='black')
    media = np.mean(datos)
    std = np.std(datos, ddof=1)
    cv = (std / media) * 100
    
    ax1.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.0f}')
    ax1.axvline(media-std, color='orange', linestyle=':', alpha=0.7, label=f'Â±1Ïƒ')
    ax1.axvline(media+std, color='orange', linestyle=':', alpha=0.7)
    
    ax1.set_title(f'{nombre}\nCV = {cv:.1f}%')
    ax1.legend(fontsize=8)
    ax1.grid(True, alpha=0.3)
    
    # Box plots (fila inferior)  
    ax2 = axes[1, i]
    bp = ax2.boxplot([datos], patch_artist=True)
    bp['boxes'][0].set_facecolor(f'C{i}')
    bp['boxes'][0].set_alpha(0.7)
    ax2.set_title(f'DispersiÃ³n Relativa')
    ax2.set_xticklabels([f'{cv:.1f}%'])
    ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
