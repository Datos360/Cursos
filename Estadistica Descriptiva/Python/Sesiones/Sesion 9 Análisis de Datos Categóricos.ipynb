## 📖 FUNDAMENTOS TEÓRICOS

### ¿Qué son los Datos Categóricos?

Los **datos categóricos** (o cualitativos) representan características que pueden ser divididas en **categorías** o **grupos**, pero que no tienen un orden numérico inherente o donde las diferencias numéricas no son significativas.

#### **🏷️ Características Fundamentales:**
- **Valores discretos** que representan categorías
- **No tienen operaciones aritméticas** significativas (no se puede "promediar" colores)
- **Requieren técnicas estadísticas especializadas**
- **Información valiosa** sobre distribuciones y patrones de comportamiento

### Importancia de los Datos Categóricos

#### **¿Por qué son cruciales en el análisis de datos?**

1. **📊 Abundantes en la realidad**
   - Género, ocupación, estado civil, nivel educativo

2. **🎯 Información estratégica**

   - Patrones de comportamiento
   - Clasificaciones de riesgo

3. **💡 Base para decisiones**
   - Estrategias de marketing dirigidas
   - Políticas públicas diferenciadas
   - Personalización de servicios

## 1. TIPOS DE VARIABLES CATEGÓRICAS

### 1.1 Variables Nominales

#### **📝 Definición:**
Categorías **sin orden intrínseco**, donde el cambio de posición no altera el significado.

#### **🔍 Características:**
- **No hay jerarquía** entre categorías
- **Diferencias cualitativas** únicamente  
- **Cada categoría es mutuamente excluyente**

#### **📋 Ejemplos:**
- **Color**: Rojo, Azul, Verde, Amarillo
- **Género**: Masculino, Femenino, Otro
- **Estado civil**: Soltero, Casado, Divorciado, Viudo
- **Tipo de sangre**: A, B, AB, O
- **Marca de producto**: Apple, Samsung, Huawei
- **Ciudad de origen**: Madrid, Barcelona, Sevilla

#### **⚠️ Importante:**
Con variables nominales, la **moda** es la única medida de tendencia central válida.

### 1.2 Variables Ordinales

#### **📝 Definición:**
Categorías con **orden jerárquico** claro, donde la secuencia tiene significado.

#### **🔍 Características:**
- **Existe jerarquía** natural entre categorías
- **Las diferencias entre niveles** pueden no ser equidistantes
- **Se puede establecer** "mayor que" o "menor que"

#### **📋 Ejemplos:**
- **Nivel educativo**: Primaria < Secundaria < Universitario < Posgrado
- **Satisfacción**: Muy insatisfecho < Insatisfecho < Neutral < Satisfecho < Muy satisfecho
- **Tamaño**: Pequeño < Mediano < Grande < Extra grande
- **Nivel socioeconómico**: Bajo < Medio-bajo < Medio < Medio-alto < Alto
- **Calificación crediticia**: D < C < B < A < AAA
- **Severidad médica**: Leve < Moderada < Grave < Crítica

#### **📊 Medidas Aplicables:**
- **Moda**: Categoría más frecuente
- **Mediana**: Categoría central (percentil 50)
- **Percentiles**: Divisiones de la distribución ordenada

### 1.3 Diferencias Clave

| Aspecto | Nominal | Ordinal |
|---------|---------|---------|
| **Orden** | ❌ No hay orden | ✅ Orden jerárquico |
| **Comparación** | Solo igualdad/diferencia | Mayor/menor/igual |
| **Mediana** | ❌ No aplicable | ✅ Aplicable |
| **Percentiles** | ❌ No aplicable | ✅ Aplicable |
| **Codificación** | Cualquier número | Números que respeten orden |

## 2. ESTADÍSTICAS DESCRIPTIVAS ESPECIALIZADAS

### 2.1 Distribuciones de Frecuencia

#### **📊 Tipos de Frecuencias para Datos Categóricos:**

**1. Frecuencia Absoluta (ni)**
- Número de observaciones en cada categoría
- Suma total = tamaño de la muestra

**2. Frecuencia Relativa (fi)**
- Proporción de cada categoría respecto al total
- fi = ni / n
- Suma de todas las fi = 1.0 (100%)

**3. Frecuencia Acumulada (Ni)**
- **Solo válida para variables ordinales**
- Suma acumulativa de frecuencias hasta cada categoría
- Útil para calcular percentiles

#### **📋 Ejemplo Ilustrativo:**

**Nivel de Satisfacción del Cliente (n=200)**

| Categoría | Frecuencia Absoluta | Frecuencia Relativa | Freq. Acumulada |
|-----------|-------------------|-------------------|-----------------|
| Muy Insatisfecho | 15 | 0.075 (7.5%) | 15 |
| Insatisfecho | 25 | 0.125 (12.5%) | 40 |
| Neutral | 50 | 0.250 (25.0%) | 90 |
| Satisfecho | 80 | 0.400 (40.0%) | 170 |
| Muy Satisfecho | 30 | 0.150 (15.0%) | 200 |

### 2.2 La Moda en Datos Categóricos

#### **📊 Definición Ampliada:**
La **moda** es la categoría con mayor frecuencia absoluta.

#### **🔍 Tipos de Distribuciones Modales:**

**1. Unimodal**
- Una sola categoría predominante
- Distribución con preferencia clara

**2. Bimodal**  
- Dos categorías con frecuencias máximas iguales
- Indica polarización o dos segmentos principales

**3. Multimodal**
- Varias categorías con frecuencias máximas similares
- Distribución más uniforme

**4. Amodal**
- Todas las categorías tienen frecuencia similar
- No hay preferencia clara

#### **📈 Interpretación Estratégica:**
- **Moda clara**: Concentración, preferencia dominante
- **Múltiples modas**: Segmentación, diversidad de preferencias
- **Distribución plana**: Fragmentación, falta de consenso

### 2.3 Medidas de Diversidad

#### **🌈 ¿Por qué medir diversidad?**
- Evaluar la **heterogeneidad** de la distribución
- Comparar **niveles de concentración** entre grupos
- Identificar **segmentos** o **monopolios**

#### **📊 Índice de Diversidad de Shannon (Entropía)**

**Fórmula:**
```
H = -Σ(pi × log(pi))
```

Donde:
- pi = proporción de la categoría i
- log = logaritmo natural o base 2

**Interpretación:**
- **H = 0**: Concentración total (una sola categoría)
- **H máximo**: Distribución perfectamente uniforme
- **H alto**: Alta diversidad, distribución dispersa
- **H bajo**: Baja diversidad, concentración

#### **📊 Índice de Gini (Concentración)**

**Fórmula:**
```
G = 1 - Σ(pi²)
```

**Interpretación:**
- **G = 0**: Concentración máxima
- **G = 1 - (1/k)**: Distribución uniforme (k = número de categorías)
- **G alto**: Alta dispersión
- **G bajo**: Alta concentración

#### **📊 Coeficiente de Variación Cualitativo**

**Fórmula:**
```
CVQ = (k × Σ(pi²) - 1) / (k - 1)
```

**Interpretación:**
- **CVQ = 0**: Distribución uniforme
- **CVQ = 1**: Concentración máxima

### 2.4 Medidas de Concentración

#### **📊 Proporción de la Moda**
```
Concentración Modal = f_moda / n
```
- Indica qué proporción representa la categoría más frecuente

#### **📊 Razón de Concentración (CR)**
```
CR_k = Suma de las k categorías más frecuentes / Total
```
- CR_3 = proporción de las 3 categorías principales
- Útil para evaluar concentración de mercado

### 4.2 Matrices de Transición

#### **🔄 Concepto:**
Muestran la **probabilidad de cambio** de una categoría a otra en períodos sucesivos.

#### **📋 Ejemplo - Lealtad de Marca:**
```
          Período t+1
         A    B    C
Período A  0.7  0.2  0.1
   t    B  0.1  0.8  0.1  
        C  0.2  0.3  0.5
```

**Interpretación:**
- 70% de clientes de marca A permanecen fieles
- 20% de clientes de marca A cambian a marca B
- Marca B tiene mayor lealtad (80%)

## 5. ANÁLISIS DE ASOCIACIÓN AVANZADO

### 5.1 Más Allá del Chi-cuadrado

#### **📊 Limitaciones del Chi-cuadrado:**
- Solo detecta **asociación**, no su **naturaleza**
- Sensible al **tamaño de muestra**
- No indica **fuerza** de la asociación

#### **🔍 Medidas Alternativas:**

**1. V de Cramér**
```
V = √(χ² / (n × min(r-1, c-1)))
```
- Rango: 0 a 1
- 0 = independencia, 1 = asociación perfecta

**2. Coeficiente de Contingencia**
```
C = √(χ² / (χ² + n))
```
- Ajustado por tamaño de muestra

**3. Lambda (λ)**
```
λ = (E₁ - E₂) / E₁
```
- Mide reducción proporcional de error
- Interpretación de mejora predictiva

### 5.2 Análisis de Correspondencias

#### **📊 Técnica Avanzada:**
Método para **visualizar** y **analizar** la asociación entre variables categóricas en un espacio dimensional reducido.

#### **💡 Beneficios:**
- **Visualización intuitiva** de relaciones complejas
- **Identificación de patrones** no evidentes en tablas
- **Reducción dimensional** para múltiples categorías

## 6. CONSIDERACIONES PRÁCTICAS

### 6.1 Codificación de Variables Categóricas

#### **🔧 Métodos de Codificación:**

**1. Codificación Nominal (One-Hot)**
- Crear variables binarias para cada categoría
- Útil para modelos que requieren variables numéricas

**2. Codificación Ordinal**
- Asignar números que respeten el orden
- Mantener distancias apropiadas cuando sea posible

**3. Codificación por Frecuencia**
- Ordenar categorías por frecuencia
- Útil para variables con muchas categorías

#### **⚠️ Precauciones:**
- **No usar codificación ordinal** para variables nominales
- **Cuidar el manejo** de categorías poco frecuentes
- **Considerar categorías de "Otros"** para simplificar

### 6.2 Tratamiento de Categorías Raras

#### **🎯 Problema:**
Categorías con muy baja frecuencia pueden:
- **Sesgar análisis** estadísticos
- **Complicar visualizaciones**
- **Reducir robustez** de modelos

#### **🛠️ Estrategias:**

**1. Agrupación Lógica**
- Combinar categorías similares
- Crear categoría "Otros" para las menos frecuentes

**2. Análisis de Frecuencia Mínima**
- Establecer umbral mínimo (ej: >5% o >10 observaciones)
- Reagrupar categorías por debajo del umbral

**3. Análisis Separado**
- Análisis principal sin categorías raras
- Análisis específico de casos especiales

### 6.3 Validación de Calidad

#### **🔍 Verificaciones Esenciales:**

**1. Consistencia de Categorías**
- Verificar **escritura uniforme** ("SI" vs "Sí" vs "si")
- Detectar **duplicados semánticos** ("Estudiante" vs "Student")

**2. Completitud**
- Identificar **valores faltantes**
- Evaluar **categorías no válidas** o "N/A"

**3. Distribución**
- Verificar **concentración excesiva** en pocas categorías
- Identificar **desequilibrios** problemáticos

## CREACIÓN DEL DATASET DE ENCUESTA DE SATISFACCIÓN

````python
# Configurar semilla para reproducibilidad
np.random.seed(42)
n_samples = 1500

print("📋 Generando dataset de encuesta de satisfacción del cliente...")

# Generar demografía básica
edades = np.random.choice(['18-25', '26-35', '36-45', '46-55', '56+'], 
                         n_samples, p=[0.15, 0.25, 0.30, 0.20, 0.10])

generos = np.random.choice(['Masculino', 'Femenino', 'Otro'], 
                          n_samples, p=[0.48, 0.50, 0.02])

# Nivel educativo (variable ordinal)
niveles_educacion = ['Primaria', 'Secundaria', 'Técnico', 'Universitario', 'Posgrado']
educacion = np.random.choice(niveles_educacion, n_samples, 
                            p=[0.05, 0.20, 0.25, 0.35, 0.15])

# Ubicación geográfica
regiones = ['Norte', 'Sur', 'Este', 'Oeste', 'Centro']
region = np.random.choice(regiones, n_samples, p=[0.22, 0.18, 0.25, 0.20, 0.15])

# Tipo de cliente
tipos_cliente = ['Nuevo', 'Regular', 'Premium', 'VIP']
tipo_cliente = np.random.choice(tipos_cliente, n_samples, 
                               p=[0.30, 0.40, 0.25, 0.05])

# Canal de servicio
canales = ['Online', 'Telefónico', 'Presencial', 'App Móvil', 'Chat']
canal_servicio = np.random.choice(canales, n_samples, 
                                 p=[0.35, 0.20, 0.15, 0.25, 0.05])

# Producto/Servicio utilizado
productos = ['Básico', 'Estándar', 'Premium', 'Enterprise']
producto = np.random.choice(productos, n_samples, 
                           p=[0.40, 0.35, 0.20, 0.05])

# Satisfacción general (variable ordinal clave)
# Introducir correlaciones realistas
satisfaccion_scores = []
for i in range(n_samples):
    # Factores que influyen en la satisfacción
    base_score = 3.0  # Neutral base
    
    # Tipo de cliente influye
    if tipo_cliente[i] == 'VIP':
        base_score += 0.8
    elif tipo_cliente[i] == 'Premium':
        base_score += 0.5
    elif tipo_cliente[i] == 'Regular':
        base_score += 0.2
    
    # Canal influye
    if canal_servicio[i] == 'Presencial':
        base_score += 0.3
    elif canal_servicio[i] == 'Online':
        base_score += 0.1
    elif canal_servicio[i] == 'Telefónico':
        base_score -= 0.2
    
    # Producto influye
    if producto[i] == 'Premium':
        base_score += 0.4
    elif producto[i] == 'Enterprise':
        base_score += 0.6
    elif producto[i] == 'Básico':
        base_score -= 0.3
    
    # Añadir ruido aleatorio
    base_score += np.random.normal(0, 0.5)
    
    # Convertir a escala ordinal
    if base_score <= 1.5:
        satisfaccion_scores.append('Muy Insatisfecho')
    elif base_score <= 2.5:
        satisfaccion_scores.append('Insatisfecho')
    elif base_score <= 3.5:
        satisfaccion_scores.append('Neutral')
    elif base_score <= 4.5:
        satisfaccion_scores.append('Satisfecho')
    else:
        satisfaccion_scores.append('Muy Satisfecho')

satisfaccion_general = satisfaccion_scores

# Probabilidad de recomendación (NPS style)
probabilidad_recomendacion = []
for satisf in satisfaccion_general:
    if satisf == 'Muy Satisfecho':
        prob_rec = np.random.choice(['Promotor', 'Neutral', 'Detractor'], 
                                   p=[0.85, 0.10, 0.05])
    elif satisf == 'Satisfecho':
        prob_rec = np.random.choice(['Promotor', 'Neutral', 'Detractor'], 
                                   p=[0.60, 0.35, 0.05])
    elif satisf == 'Neutral':
        prob_rec = np.random.choice(['Promotor', 'Neutral', 'Detractor'], 
                                   p=[0.10, 0.70, 0.20])
    elif satisf == 'Insatisfecho':
        prob_rec = np.random.choice(['Promotor', 'Neutral', 'Detractor'], 
                                   p=[0.05, 0.15, 0.80])
    else:  # Muy Insatisfecho
        prob_rec = np.random.choice(['Promotor', 'Neutral', 'Detractor'], 
                                   p=[0.02, 0.08, 0.90])
    probabilidad_recomendacion.append(prob_rec)

# Frecuencia de uso
frecuencias_uso = ['Diario', 'Semanal', 'Quincenal', 'Mensual', 'Ocasional']
frecuencia_uso = np.random.choice(frecuencias_uso, n_samples, 
                                 p=[0.20, 0.30, 0.15, 0.25, 0.10])

# Motivo de contacto
motivos = ['Consulta', 'Problema técnico', 'Reclamo', 'Sugerencia', 'Información']
motivo_contacto = np.random.choice(motivos, n_samples, 
                                  p=[0.35, 0.25, 0.20, 0.05, 0.15])

# Crear DataFrame
df_satisfaccion = pd.DataFrame({
    'edad_grupo': edades,
    'genero': generos,
    'nivel_educativo': educacion,
    'region': region,
    'tipo_cliente': tipo_cliente,
    'canal_servicio': canal_servicio,
    'producto_servicio': producto,
    'satisfaccion_general': satisfaccion_general,
    'probabilidad_recomendacion': probabilidad_recomendacion,
    'frecuencia_uso': frecuencia_uso,
    'motivo_contacto': motivo_contacto
})

print("✅ Dataset de satisfacción creado exitosamente")
print("📊 Información del dataset:")
print("="*50)
print(f"Tamaño: {df_satisfaccion.shape}")
print(f"Variables: {list(df_satisfaccion.columns)}")

print("\n🔍 Primeras observaciones:")
print(df_satisfaccion.head())

print("\n📋 Información básica del dataset:")
print(df_satisfaccion.info())
````

## 1. ANÁLISIS DESCRIPTIVO BÁSICO DE VARIABLES CATEGÓRICAS

````python
def analisis_basico_categorica(df, columna, mostrar_grafico=True):
    """
    Realizar análisis descriptivo básico de una variable categórica
    """
    print(f"\n🔍 ANÁLISIS BÁSICO - {columna.upper()}")
    print("="*50)
    
    # Calcular frecuencias
    freq_abs = df[columna].value_counts()
    freq_rel = df[columna].value_counts(normalize=True) * 100
    freq_acum = freq_abs.cumsum()  # Solo para variables ordinales
    
    # Crear tabla resumen
    tabla_resumen = pd.DataFrame({
        'Frecuencia_Absoluta': freq_abs,
        'Frecuencia_Relativa_Pct': freq_rel.round(2),
        'Frecuencia_Acumulada': freq_acum
    })
    
    print("📊 Tabla de Frecuencias:")
    print(tabla_resumen)
    
    # Estadísticas básicas
    n_categorias = len(freq_abs)
    moda = freq_abs.index[0]
    freq_moda = freq_abs.iloc[0]
    porcentaje_moda = freq_rel.iloc[0]
    
    print(f"\n📈 Estadísticas Básicas:")
    print(f"   Número total de observaciones: {len(df)}")
    print(f"   Número de categorías únicas: {n_categorias}")
    print(f"   Moda (categoría más frecuente): {moda}")
    print(f"   Frecuencia de la moda: {freq_moda} ({porcentaje_moda:.1f}%)")
    
    # Verificar si hay valores faltantes
    valores_faltantes = df[columna].isna().sum()
    if valores_faltantes > 0:
        print(f"   ⚠️ Valores faltantes: {valores_faltantes}")
    else:
        print(f"   ✅ Sin valores faltantes")
    
    # Visualización
    if mostrar_grafico:
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        
        # Gráfico de barras
        freq_abs.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
        axes[0].set_title(f'Frecuencias Absolutas - {columna}', fontweight='bold')
        axes[0].set_xlabel(columna)
        axes[0].set_ylabel('Frecuencia')
        axes[0].tick_params(axis='x', rotation=45)
        axes[0].grid(True, alpha=0.3)
        
        # Gráfico de pastel (solo si hay ≤ 6 categorías)
        if n_categorias <= 6:
            axes[1].pie(freq_abs.values, labels=freq_abs.index, autopct='%1.1f%%', 
                       startangle=90)
            axes[1].set_title(f'Distribución Porcentual - {columna}', fontweight='bold')
        else:
            # Gráfico de barras horizontal para muchas categorías
            freq_abs.plot(kind='barh', ax=axes[1], color='lightcoral')
            axes[1].set_title(f'Frecuencias (Horizontal) - {columna}', fontweight='bold')
            axes[1].set_xlabel('Frecuencia')
        
        plt.tight_layout()
        plt.show()
    
    return tabla_resumen, {
        'n_categorias': n_categorias,
        'moda': moda,
        'freq_moda': freq_moda,
        'porcentaje_moda': porcentaje_moda,
        'valores_faltantes': valores_faltantes
    }

# Analizar variables categóricas principales
variables_categoricas = ['satisfaccion_general', 'tipo_cliente', 'canal_servicio', 
                        'probabilidad_recomendacion', 'region']

resultados_basicos = {}
for variable in variables_categoricas[:3]:  # Analizar las primeras 3
    tabla, stats = analisis_basico_categorica(df_satisfaccion, variable)
    resultados_basicos[variable] = {'tabla': tabla, 'stats': stats}
````

## 2. ANÁLISIS DE VARIABLES ORDINALES

````python
print("\n" + "="*70)
print("📊 2. ANÁLISIS ESPECIALIZADO DE VARIABLES ORDINALES")
print("="*70)

def analisis_variable_ordinal(df, columna, orden_categorias=None):
    """
    Análisis específico para variables ordinales
    """
    print(f"\n🔍 ANÁLISIS ORDINAL - {columna.upper()}")
    print("="*50)
    
    if orden_categorias is None:
        # Usar el orden natural de los datos
        orden_categorias = df[columna].value_counts().index.tolist()
        print(f"⚠️ Usando orden natural de los datos: {orden_categorias}")
    else:
        print(f"📋 Orden especificado: {orden_categorias}")
    
    # Crear variable categórica ordenada
    df_temp = df.copy()
    df_temp[f'{columna}_ordenada'] = pd.Categorical(df_temp[columna], 
                                                   categories=orden_categorias, 
                                                   ordered=True)
    
    # Frecuencias respetando el orden
    freq_abs = df_temp[f'{columna}_ordenada'].value_counts().sort_index()
    freq_rel = (freq_abs / len(df_temp)) * 100
    freq_acum = freq_abs.cumsum()
    freq_rel_acum = (freq_acum / len(df_temp)) * 100
    
    # Tabla completa
    tabla_ordinal = pd.DataFrame({
        'Frecuencia_Abs': freq_abs,
        'Frecuencia_Rel_%': freq_rel.round(2),
        'Frecuencia_Acum': freq_acum,
        'Frecuencia_Rel_Acum_%': freq_rel_acum.round(2)
    })
    
    print("📊 Tabla de Frecuencias Ordinales:")
    print(tabla_ordinal)
    
    # Calcular percentiles para variables ordinales
    print(f"\n📏 Análisis de Posición:")
    
    # Mediana (percentil 50)
    valores_ordenados = []
    for categoria in orden_categorias:
        count = freq_abs[categoria] if categoria in freq_abs.index else 0
        valores_ordenados.extend([categoria] * count)
    
    n_total = len(valores_ordenados)
    pos_mediana = n_total // 2
    
    if n_total % 2 == 1:
        mediana = valores_ordenados[pos_mediana]
    else:
        # Para datos categóricos, tomamos el valor en la posición inferior
        mediana = valores_ordenados[pos_mediana - 1]
    
    print(f"   Mediana: {mediana}")
    
    # Cuartiles aproximados
    pos_q1 = n_total // 4
    pos_q3 = 3 * n_total // 4
    
    q1 = valores_ordenados[pos_q1]
    q3 = valores_ordenados[pos_q3]
    
    print(f"   Q1 (aprox): {q1}")
    print(f"   Q3 (aprox): {q3}")
    
    # Moda
    moda = freq_abs.index[freq_abs.argmax()]
    print(f"   Moda: {moda}")
    
    # Visualización específica para ordinales
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Gráfico de barras ordenado
    freq_abs.plot(kind='bar', ax=axes[0], color='lightgreen', edgecolor='black')
    axes[0].set_title(f'Distribución Ordinal - {columna}', fontweight='bold')
    axes[0].set_xlabel(columna)
    axes[0].set_ylabel('Frecuencia')
    axes[0].tick_params(axis='x', rotation=45)
    axes[0].grid(True, alpha=0.3)
    
    # Gráfico de frecuencias acumuladas
    freq_rel_acum.plot(kind='line', marker='o', ax=axes[1], color='red', linewidth=2)
    axes[1].set_title(f'Frecuencias Relativas Acumuladas - {columna}', fontweight='bold')
    axes[1].set_xlabel(columna)
    axes[1].set_ylabel('Porcentaje Acumulado (%)')
    axes[1].set_ylim(0, 100)
    axes[1].grid(True, alpha=0.3)
    axes[1].tick_params(axis='x', rotation=45)
    
    # Gráfico de barras apiladas mostrando distribución
    bottom_values = np.zeros(len(freq_rel))
    colors = plt.cm.RdYlBu(np.linspace(0, 1, len(freq_rel)))
    
    for i, (categoria, porcentaje) in enumerate(freq_rel.items()):
        axes[2].bar('Distribución', porcentaje, bottom=bottom_values[i], 
                   color=colors[i], label=categoria, width=0.5)
        bottom_values[i+1:] += porcentaje
    
    axes[2].set_title(f'Distribución Apilada - {columna}', fontweight='bold')
    axes[2].set_ylabel('Porcentaje (%)')
    axes[2].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
    return tabla_ordinal, {
        'mediana': mediana,
        'q1': q1,
        'q3': q3,
        'moda': moda,
        'orden_categorias': orden_categorias
    }

# Analizar variables ordinales específicas
print("📊 Analizando SATISFACCIÓN GENERAL (variable ordinal clave):")

orden_satisfaccion = ['Muy Insatisfecho', 'Insatisfecho', 'Neutral', 'Satisfecho', 'Muy Satisfecho']
tabla_satisfaccion, stats_satisfaccion = analisis_variable_ordinal(
    df_satisfaccion, 'satisfaccion_general', orden_satisfaccion
)

print("\n📊 Analizando NIVEL EDUCATIVO (variable ordinal):")
orden_educacion = ['Primaria', 'Secundaria', 'Técnico', 'Universitario', 'Posgrado']
tabla_educacion, stats_educacion = analisis_variable_ordinal(
    df_satisfaccion, 'nivel_educativo', orden_educacion
)
````

## 3. MEDIDAS DE DIVERSIDAD Y CONCENTRACIÓN

````python
print("\n" + "="*70)
print("📊 3. MEDIDAS DE DIVERSIDAD Y CONCENTRACIÓN")
print("="*70)

def calcular_medidas_diversidad(df, columna):
    """
    Calcular índices de diversidad y concentración para una variable categórica
    """
    print(f"\n🌈 MEDIDAS DE DIVERSIDAD - {columna.upper()}")
    print("="*50)
    
    # Obtener proporciones
    freq_abs = df[columna].value_counts()
    proporciones = freq_abs / len(df)
    n_categorias = len(freq_abs)
    
    print(f"📊 Información básica:")
    print(f"   Número de categorías: {n_categorias}")
    print(f"   Total de observaciones: {len(df)}")
    
    # 1. Índice de Diversidad de Shannon (Entropía)
    shannon = -np.sum(proporciones * np.log(proporciones))
    shannon_max = np.log(n_categorias)  # Máximo teórico (distribución uniforme)
    shannon_normalizado = shannon / shannon_max if shannon_max > 0 else 0
    
    print(f"\n📈 Índice de Shannon (Entropía):")
    print(f"   Valor observado: {shannon:.4f}")
    print(f"   Máximo teórico: {shannon_max:.4f}")
    print(f"   Normalizado (0-1): {shannon_normalizado:.4f}")
    
    if shannon_normalizado > 0.8:
        print("   → Interpretación: ALTA diversidad (distribución dispersa)")
    elif shannon_normalizado > 0.6:
        print("   → Interpretación: MODERADA diversidad")
    else:
        print("   → Interpretación: BAJA diversidad (distribución concentrada)")
    
    # 2. Índice de Gini
    gini = 1 - np.sum(proporciones**2)
    gini_max = 1 - (1/n_categorias)  # Máximo para distribución uniforme
    gini_normalizado = gini / gini_max if gini_max > 0 else 0
    
    print(f"\n📊 Índice de Gini:")
    print(f"   Valor observado: {gini:.4f}")
    print(f"   Máximo teórico: {gini_max:.4f}")
    print(f"   Normalizado (0-1): {gini_normalizado:.4f}")
    
    # 3. Coeficiente de Variación Cualitativo
    cvq = (n_categorias * np.sum(proporciones**2) - 1) / (n_categorias - 1)
    
    print(f"\n📏 Coeficiente de Variación Cualitativo:")
    print(f"   CVQ: {cvq:.4f}")
    print("   (0 = uniforme, 1 = máxima concentración)")
    
    # 4. Medidas de Concentración
    prop_moda = proporciones.iloc[0]  # Proporción de la categoría más frecuente
    
    print(f"\n🎯 Medidas de Concentración:")
    print(f"   Proporción de la moda: {prop_moda:.4f} ({prop_moda*100:.1f}%)")
    
    # Razón de concentración para las top 3 categorías
    top_3_prop = proporciones.head(3).sum() if len(proporciones) >= 3 else proporciones.sum()
    print(f"   CR3 (top 3 categorías): {top_3_prop:.4f} ({top_3_prop*100:.1f}%)")
    
    # Visualización de diversidad
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Comparación con distribución uniforme
    uniform_props = [1/n_categorias] * n_categorias
    
    x_pos = np.arange(len(proporciones))
    width = 0.35
    
    axes[0].bar(x_pos - width/2, proporciones.values, width, 
               label='Observado', alpha=0.8, color='skyblue')
    axes[0].bar(x_pos + width/2, uniform_props, width, 
               label='Uniforme', alpha=0.8, color='orange')
    
    axes[0].set_title(f'Comparación con Distribución Uniforme - {columna}', fontweight='bold')
    axes[0].set_xlabel('Categorías')
    axes[0].set_ylabel('Proporción')
    axes[0].set_xticks(x_pos)
    axes[0].set_xticklabels(proporciones.index, rotation=45)
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Curva de Lorenz (concentración)
    proporciones_ordenadas = np.sort(proporciones.values)
    lorenz_y = np.cumsum(proporciones_ordenadas)
    lorenz_x = np.linspace(0, 1, len(lorenz_y))
    
    axes[1].plot([0] + list(lorenz_x), [0] + list(lorenz_y), 'o-', label='Curva de Lorenz')
    axes[1].plot([0, 1], [0, 1], '--', label='Línea de Equidad', color='red')
    axes[1].fill_between([0] + list(lorenz_x), [0] + list(lorenz_y), 
                        [0] + list(lorenz_x), alpha=0.3)
    
    axes[1].set_title(f'Curva de Lorenz - {columna}', fontweight='bold')
    axes[1].set_xlabel('Proporción Acumulada de Categorías')
    axes[1].set_ylabel('Proporción Acumulada de Frecuencias')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return {
        'shannon': shannon,
        'shannon_normalizado': shannon_normalizado,
        'gini': gini,
        'gini_normalizado': gini_normalizado,
        'cvq': cvq,
        'proporcion_moda': prop_moda,
        'cr3': top_3_prop,
        'n_categorias': n_categorias
    }

# Calcular medidas de diversidad para variables clave
diversidad_satisfaccion = calcular_medidas_diversidad(df_satisfaccion, 'satisfaccion_general')
diversidad_canal = calcular_medidas_diversidad(df_satisfaccion, 'canal_servicio')
diversidad_region = calcular_medidas_diversidad(df_satisfaccion, 'region')
````

## 4. ANÁLISIS COMPARATIVO ENTRE GRUPOS

````python
print("\n" + "="*70)
print("📊 4. ANÁLISIS COMPARATIVO ENTRE GRUPOS")
print("="*70)

def analizar_categorica_por_grupos(df, variable_categorica, variable_grupo):
    """
    Analizar una variable categórica segmentada por grupos
    """
    print(f"\n🔍 ANÁLISIS COMPARATIVO: {variable_categorica.upper()} por {variable_grupo.upper()}")
    print("="*60)
    
    # Crear tabla de contingencia
    tabla_contingencia = pd.crosstab(df[variable_grupo], df[variable_categorica], margins=True)
    
    print("📊 Tabla de Contingencia (Frecuencias Absolutas):")
    print(tabla_contingencia)
    
    # Porcentajes por fila (distribución dentro de cada grupo)
    tabla_pct_fila = pd.crosstab(df[variable_grupo], df[variable_categorica], normalize='index') * 100
    
    print(f"\n📈 Distribución Porcentual por {variable_grupo.upper()}:")
    print(tabla_pct_fila.round(2))
    
    # Porcentajes por columna (composición de cada categoría)
    tabla_pct_columna = pd.crosstab(df[variable_grupo], df[variable_categorica], normalize='columns') * 100
    
    print(f"\n📊 Composición por {variable_categorica.upper()}:")
    print(tabla_pct_columna.round(2))
    
    # Prueba de Chi-cuadrado
    chi2_stat, p_valor, grados_libertad, esperadas = chi2_contingency(tabla_contingencia.iloc[:-1, :-1])
    
    print(f"\n🧪 Prueba de Chi-cuadrado de Independencia:")
    print(f"   Estadístico χ²: {chi2_stat:.4f}")
    print(f"   P-valor: {p_valor:.4f}")
    print(f"   Grados de libertad: {grados_libertad}")
    
    if p_valor < 0.05:
        print("   ✅ Conclusión: Las variables SÍ están asociadas (p < 0.05)")
    else:
        print("   ❌ Conclusión: No hay evidencia de asociación (p ≥ 0.05)")
    
    # V de Cramér (fuerza de la asociación)
    n = tabla_contingencia.iloc[-1, -1]  # Total general
    v_cramer = np.sqrt(chi2_stat / (n * (min(tabla_contingencia.shape) - 2)))
    
    print(f"\n📏 V de Cramér (fuerza de asociación):")
    print(f"   V: {v_cramer:.4f}")
    if v_cramer < 0.1:
        print("   → Asociación: DÉBIL")
    elif v_cramer < 0.3:
        print("   → Asociación: MODERADA")
    else:
        print("   → Asociación: FUERTE")
    
    # Análisis de diversidad por grupos
    print(f"\n🌈 Diversidad por {variable_grupo.upper()}:")
    diversidades_por_grupo = {}
    
    for grupo in df[variable_grupo].unique():
        if pd.notna(grupo):  # Excluir valores faltantes
            df_grupo = df[df[variable_grupo] == grupo]
            if len(df_grupo) > 0:
                proporciones = df_grupo[variable_categorica].value_counts(normalize=True)
                shannon = -np.sum(proporciones * np.log(proporciones))
                diversidades_por_grupo[grupo] = shannon
                print(f"   {grupo}: Shannon = {shannon:.4f}")
    
    # Visualización comparativa
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # 1. Heatmap de la tabla de contingencia
    sns.heatmap(tabla_contingencia.iloc[:-1, :-1], annot=True, fmt='d', cmap='Blues', ax=axes[0,0])
    axes[0,0].set_title(f'Heatmap: {variable_categorica} por {variable_grupo}', fontweight='bold')
    
    # 2. Gráfico de barras agrupadas
    tabla_pct_fila.plot(kind='bar', ax=axes[0,1], rot=45)
    axes[0,1].set_title(f'Distribución % por {variable_grupo}', fontweight='bold')
    axes[0,1].set_ylabel('Porcentaje (%)')
    axes[0,1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 3. Gráfico de barras apiladas
    tabla_pct_fila.plot(kind='bar', stacked=True, ax=axes[1,0], rot=45)
    axes[1,0].set_title(f'Distribución Apilada por {variable_grupo}', fontweight='bold')
    axes[1,0].set_ylabel('Porcentaje (%)')
    axes[1,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 4. Comparación de diversidades
    if diversidades_por_grupo:
        grupos_div = list(diversidades_por_grupo.keys())
        valores_div = list(diversidades_por_grupo.values())
        
        axes[1,1].bar(grupos_div, valores_div, color='lightcoral')
        axes[1,1].set_title(f'Diversidad Shannon por {variable_grupo}', fontweight='bold')
        axes[1,1].set_ylabel('Índice de Shannon')
        axes[1,1].tick_params(axis='x', rotation=45)
        axes[1,1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return {
        'tabla_contingencia': tabla_contingencia,
        'tabla_pct_fila': tabla_pct_fila,
        'chi2_stat': chi2_stat,
        'p_valor': p_valor,
        'v_cramer': v_cramer,
        'diversidades_por_grupo': diversidades_por_grupo
    }

# Análisis comparativos clave
print("📊 Analizando SATISFACCIÓN por TIPO DE CLIENTE:")
analisis_satisfaccion_tipo = analizar_categorica_por_grupos(
    df_satisfaccion, 'satisfaccion_general', 'tipo_cliente'
)

print("\n📊 Analizando CANAL DE SERVICIO por REGIÓN:")
analisis_canal_region = analizar_categorica_por_grupos(
    df_satisfaccion, 'canal_servicio', 'region'
)
````

## 5. ANÁLISIS DE PATRONES COMPLEJOS

````python
print("\n" + "="*70)
print("📊 5. ANÁLISIS DE PATRONES COMPLEJOS")
print("="*70)

def analizar_patrones_multiples(df, variables_categoricas):
    """
    Análisis de patrones entre múltiples variables categóricas
    """
    print(f"\n🔍 ANÁLISIS DE PATRONES MÚLTIPLES")
    print("="*50)
    print(f"Variables analizadas: {', '.join(variables_categoricas)}")
    
    # Crear combinaciones de categorías
    df_temp = df[variables_categoricas].copy()
    
    # Combinar todas las variables en una sola columna
    combinaciones = df_temp.apply(lambda x: ' | '.join(x.astype(str)), axis=1)
    
    # Análisis de combinaciones más frecuentes
    freq_combinaciones = combinaciones.value_counts()
    
    print(f"\n📊 Top 10 combinaciones más frecuentes:")
    print("="*60)
    for i, (combinacion, freq) in enumerate(freq_combinaciones.head(10).items(), 1):
        pct = (freq / len(df)) * 100
        print(f"{i:2d}. {combinacion:<40} | {freq:4d} ({pct:5.1f}%)")
    
    # Análisis de diversidad de combinaciones
    prop_combinaciones = freq_combinaciones / len(df)
    shannon_combinaciones = -np.sum(prop_combinaciones * np.log(prop_combinaciones))
    
    print(f"\n🌈 Diversidad de combinaciones:")
    print(f"   Número de combinaciones únicas: {len(freq_combinaciones)}")
    print(f"   Índice de Shannon: {shannon_combinaciones:.4f}")
    
    # Matriz de correlación categórica usando V de Cramér
    print(f"\n📊 Matriz de Asociación (V de Cramér):")
    n_vars = len(variables_categoricas)
    matriz_cramer = np.zeros((n_vars, n_vars))
    
    for i, var1 in enumerate(variables_categoricas):
        for j, var2 in enumerate(variables_categoricas):
            if i == j:
                matriz_cramer[i, j] = 1.0
            else:
                tabla_cont = pd.crosstab(df[var1], df[var2])
                chi2_val, _, _, _ = chi2_contingency(tabla_cont)
                n = tabla_cont.sum().sum()
                cramer_v = np.sqrt(chi2_val / (n * (min(tabla_cont.shape) - 1)))
                matriz_cramer[i, j] = cramer_v
    
    # Crear DataFrame de la matriz
    df_cramer = pd.DataFrame(matriz_cramer, 
                           index=variables_categoricas, 
                           columns=variables_categoricas)
    print(df_cramer.round(3))
    
    # Visualización de patrones
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # 1. Top combinaciones
    top_10_comb = freq_combinaciones.head(10)
    top_10_comb.plot(kind='barh', ax=axes[0,0])
    axes[0,0].set_title('Top 10 Combinaciones de Categorías', fontweight='bold')
    axes[0,0].set_xlabel('Frecuencia')
    
    # 2. Distribución de frecuencias de combinaciones
    axes[0,1].hist(freq_combinaciones.values, bins=20, edgecolor='black', alpha=0.7)
    axes[0,1].set_title('Distribución de Frecuencias de Combinaciones', fontweight='bold')
    axes[0,1].set_xlabel('Frecuencia de Combinación')
    axes[0,1].set_ylabel('Número de Combinaciones')
    axes[0,1].grid(True, alpha=0.3)
    
    # 3. Heatmap de correlaciones categóricas
    sns.heatmap(df_cramer, annot=True, cmap='YlOrRd', square=True, ax=axes[1,0])
    axes[1,0].set_title('Matriz de Asociación (V de Cramér)', fontweight='bold')
    
    # 4. Proporción de combinaciones raras vs frecuentes
    umbral_raro = 0.01  # 1% como umbral para "raro"
    combinaciones_raras = (prop_combinaciones < umbral_raro).sum()
    combinaciones_frecuentes = (prop_combinaciones >= umbral_raro).sum()
    
    labels = ['Combinaciones Raras\n(<1%)', 'Combinaciones Frecuentes\n(≥1%)']
    sizes = [combinaciones_raras, combinaciones_frecuentes]
    colors = ['lightcoral', 'lightblue']
    
    axes[1,1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    axes[1,1].set_title('Distribución: Combinaciones Raras vs Frecuentes', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    return {
        'freq_combinaciones': freq_combinaciones,
        'shannon_combinaciones': shannon_combinaciones,
        'matriz_cramer': df_cramer,
        'combinaciones_raras': combinaciones_raras,
        'combinaciones_frecuentes': combinaciones_frecuentes
    }

# Analizar patrones entre variables clave
variables_patron = ['satisfaccion_general', 'tipo_cliente', 'canal_servicio', 'region']
patrones_multiples = analizar_patrones_multiples(df_satisfaccion, variables_patron)
````
