## üìñ **PARTE 1: ¬øQu√© son las Medidas de Tendencia Central?**

### **üéØ Definici√≥n**

> Las **medidas de tendencia central** son valores que representan el "centro" o "valor t√≠pico" de un conjunto de datos. Nos ayudan a resumir toda una distribuci√≥n con un solo n√∫mero representativo.

### **ü§î ¬øPor qu√© son Importantes?**

1. **Simplificaci√≥n**: Resumen miles de datos en un solo valor
2. **Comparaci√≥n**: Permiten comparar diferentes grupos
3. **Comunicaci√≥n**: Facilitan explicar resultados a otros
4. **Toma de Decisiones**: Base para decisiones informadas
5. **Punto de Referencia**: Centro alrededor del cual se distribuyen los datos

### **üìä Las Tres Medidas Principales**

```
MEDIDAS DE TENDENCIA CENTRAL
‚îú‚îÄ‚îÄ MEDIA (Promedio)
‚îÇ   ‚îî‚îÄ‚îÄ Suma todos los valores y divide entre n
‚îú‚îÄ‚îÄ MEDIANA (Valor central)
‚îÇ   ‚îî‚îÄ‚îÄ Valor que divide los datos en dos mitades iguales
‚îî‚îÄ‚îÄ MODA (M√°s frecuente)
    ‚îî‚îÄ‚îÄ Valor que aparece m√°s veces
```

## üìä **PARTE 2: La Media Aritm√©tica**

### **üî¢ Definici√≥n y F√≥rmula**

La **media aritm√©tica** (o simplemente "media") es el promedio de todos los valores en un conjunto de datos.

#### **Para una Muestra:**
```
xÃÑ = (x‚ÇÅ + x‚ÇÇ + x‚ÇÉ + ... + x‚Çô) / n = Œ£x / n
```

#### **Para una Poblaci√≥n:**
```
Œº = (x‚ÇÅ + x‚ÇÇ + x‚ÇÉ + ... + x‚Çô) / N = Œ£x / N
```

**Donde:**
- xÃÑ (x-barra) = media muestral
- Œº (mu) = media poblacional
- Œ£x = suma de todos los valores
- n = tama√±o de la muestra
- N = tama√±o de la poblaci√≥n

### **üí° Ejemplo Paso a Paso**

**Situaci√≥n**: Las edades de 7 empleados en una startup son: 25, 28, 22, 30, 26, 24, 29

**C√°lculo:**
1. Sumar todos los valores: 25 + 28 + 22 + 30 + 26 + 24 + 29 = 184
2. Dividir entre el n√∫mero de empleados: 184 √∑ 7 = 26.29 a√±os
3. **Interpretaci√≥n**: La edad promedio de los empleados es 26.3 a√±os

### **‚úÖ Ventajas de la Media**

1. **Utiliza todos los datos**: Considera cada valor del conjunto
2. **Matem√°ticamente √∫til**: Permite c√°lculos adicionales
3. **F√°cil de calcular**: F√≥rmula simple y directa
4. **Ampliamente conocida**: Todos entienden qu√© es un promedio
5. **Base para otros c√°lculos**: Varianza, desviaci√≥n est√°ndar, etc.

### **‚ö†Ô∏è Desventajas de la Media**

1. **Sensible a valores extremos**: Los outliers la afectan mucho
2. **No siempre representativa**: Puede no coincidir con ning√∫n valor real
3. **Solo para datos num√©ricos**: No se puede calcular para datos categ√≥ricos
4. **Puede ser enga√±osa**: En distribuciones muy asim√©tricas

### **üéØ Ejemplo del Problema con Outliers**

**Caso 1 - Sin outliers**: Salarios de 5 empleados
- Datos: $30,000, $32,000, $31,000, $33,000, $29,000
- Media: $31,000 ‚úÖ *Representativa*

**Caso 2 - Con outlier**: Mismo grupo + el CEO
- Datos: $30,000, $32,000, $31,000, $33,000, $29,000, $500,000
- Media: $109,167 ‚ùå *¬°No representa a ning√∫n empleado!*

## üìä **PARTE 3: La Mediana**

### **üî¢ Definici√≥n**

> La **mediana** es el valor que divide un conjunto ordenado de datos en dos mitades iguales. El 50% de los datos est√° por debajo de la mediana y el 50% por encima.

### **üìã Pasos para Calcular la Mediana**

1. **Ordenar** los datos de menor a mayor
2. **Contar** el total de observaciones (n)
3. **Encontrar la posici√≥n**:
   - Si n es impar: posici√≥n = (n + 1) √∑ 2
   - Si n es par: promedio de las posiciones n√∑2 y (n√∑2) + 1

### **üí° Ejemplos Detallados**

#### **Ejemplo 1 - N√∫mero impar de datos**
**Datos**: Tiempo de espera en minutos: 5, 8, 12, 15, 20, 25, 30

1. Ya est√°n ordenados ‚úÖ
2. n = 7 (impar)
3. Posici√≥n = (7 + 1) √∑ 2 = 4
4. **Mediana = 15 minutos**

```
Posici√≥n: 1   2   3   4   5   6   7
Valores:  5   8  12  15  20  25  30
                    ‚Üë
                 Mediana
```

#### **Ejemplo 2 - N√∫mero par de datos**
**Datos**: Calificaciones: 85, 88, 90, 92, 95, 98

1. Ya est√°n ordenados ‚úÖ
2. n = 6 (par)
3. Posiciones centrales: 6√∑2 = 3 y 4
4. Valores en posiciones 3 y 4: 90 y 92
5. **Mediana = (90 + 92) √∑ 2 = 91**

```
Posici√≥n: 1   2   3   4   5   6
Valores: 85  88  90  92  95  98
                ‚Üë   ‚Üë
            Mediana = 91
```

### **‚úÖ Ventajas de la Mediana**

1. **Resistente a outliers**: Los valores extremos no la afectan
2. **Siempre representativa**: Divide los datos exactamente por la mitad
3. **F√°cil de interpretar**: "La mitad est√° por debajo, la mitad por encima"
4. **√ötil para datos ordinales**: Se puede calcular para datos ordenables
5. **Mejor para distribuciones asim√©tricas**: M√°s representativa que la media

### **‚ö†Ô∏è Desventajas de la Mediana**

1. **No usa toda la informaci√≥n**: Solo considera la posici√≥n, no los valores exactos
2. **Menos √∫til matem√°ticamente**: No permite tantos c√°lculos derivados
3. **Puede ser menos precisa**: Con pocos datos, puede no ser muy informativa

### **üéØ La Mediana vs Media con Outliers**

Retomemos el ejemplo de los salarios:

**Con outlier**: $29,000, $30,000, $31,000, $32,000, $33,000, $500,000

- **Media**: $109,167 ‚ùå *Distorsionada*
- **Mediana**: $(31,000 + 32,000) √∑ 2 = $31,500$ ‚úÖ *Representativa*

La mediana nos da una mejor idea del salario "t√≠pico" del empleado.

## üìä **PARTE 4: La Moda**

### **üî¢ Definici√≥n**

> La **moda** es el valor o categor√≠a que aparece con mayor frecuencia en un conjunto de datos. Es la √∫nica medida de tendencia central que se puede aplicar a todos los tipos de variables.

### **üè∑Ô∏è Tipos de Distribuciones seg√∫n la Moda**

#### **1. Unimodal** - Una sola moda
```
Frecuencia:  1  3  7  2  1
Valores:     A  B  C  D  E
                    ‚Üë
                  Moda
```

#### **2. Bimodal** - Dos modas
```
Frecuencia:  1  5  2  5  1
Valores:     A  B  C  D  E
                ‚Üë     ‚Üë
              Modas
```

#### **3. Multimodal** - M√∫ltiples modas
```
Frecuencia:  4  2  4  1  4
Valores:     A  B  C  D  E
             ‚Üë     ‚Üë     ‚Üë
               Modas
```

#### **4. Amodal** - Sin moda
```
Frecuencia:  1  1  1  1  1
Valores:     A  B  C  D  E
         (Todos igual frecuencia)
```

### **üí° Ejemplos por Tipo de Variable**

#### **Variables Categ√≥ricas**
**Colores favoritos en una encuesta:**
- Azul: 15 personas
- Rojo: 8 personas  
- Verde: 12 personas
- Amarillo: 5 personas

**Moda: Azul** (mayor frecuencia)

#### **Variables Num√©ricas Discretas**
**N√∫mero de hijos por familia:**
- 0 hijos: 3 familias
- 1 hijo: 8 familias
- 2 hijos: 12 familias ‚Üê **Moda**
- 3 hijos: 5 familias
- 4 hijos: 2 familias

#### **Variables Num√©ricas Continuas**
Para variables continuas, la moda se determina por:
1. **Agrupaci√≥n en intervalos** (como en histogramas)
2. **El intervalo con mayor frecuencia** es modal
3. **Estimaci√≥n del valor modal** dentro del intervalo

### **‚úÖ Ventajas de la Moda**

1. **Aplicable a cualquier tipo de datos**: Categ√≥ricos, ordinales, num√©ricos
2. **F√°cil de identificar**: Solo requiere contar frecuencias
3. **Resistente a outliers**: Los valores extremos no la afectan
4. **√ötil para datos categ√≥ricos**: √önica medida central posible
5. **Pr√°cticamente √∫til**: "¬øQu√© es lo m√°s com√∫n?"

### **‚ö†Ô∏è Desventajas de la Moda**

1. **Puede no existir**: Todos los valores con igual frecuencia
2. **Puede no ser √∫nica**: M√∫ltiples modas posibles
3. **No usa toda la informaci√≥n**: Solo la frecuencia
4. **Menos estable**: Peque√±os cambios pueden cambiar la moda
5. **Dif√≠cil para datos continuos**: Requiere agrupaci√≥n

## ‚öñÔ∏è **PARTE 5: ¬øCu√°ndo Usar Cada Medida?**

### **üéØ Gu√≠a de Decisi√≥n**

| Situaci√≥n | Medida Recomendada | Raz√≥n |
|-----------|-------------------|-------|
| **Datos sim√©tricos, sin outliers** | **Media** | Utiliza toda la informaci√≥n |
| **Datos con outliers** | **Mediana** | No se ve afectada por extremos |
| **Datos categ√≥ricos** | **Moda** | √önica medida aplicable |
| **Distribuciones asim√©tricas** | **Mediana** | M√°s representativa del centro |
| **An√°lisis de tendencias populares** | **Moda** | Muestra lo m√°s com√∫n |
| **C√°lculos matem√°ticos posteriores** | **Media** | Base para otras estad√≠sticas |

### **üìä Relaci√≥n entre las Medidas seg√∫n la Forma de la Distribuci√≥n**

#### **Distribuci√≥n Sim√©trica** (Curva Normal)
```
Media = Mediana = Moda
```
![Sim√©trica: campana perfecta]

#### **Distribuci√≥n Asim√©trica Positiva** (Cola hacia la derecha)
```
Moda < Mediana < Media
```
![Asim√©trica positiva: cola larga a la derecha]

#### **Distribuci√≥n Asim√©trica Negativa** (Cola hacia la izquierda)  
```
Media < Mediana < Moda
```
![Asim√©trica negativa: cola larga a la izquierda]

### **üéØ Casos de Estudio Pr√°cticos**

#### **Caso 1: Ingresos Familiares** üí∞
**Problema**: Los ingresos suelen tener distribuci√≥n asim√©trica (pocos muy ricos, muchos con ingresos moderados)

**Datos ejemplo**: $25K, $30K, $32K, $35K, $38K, $40K, $200K
- Media: $57K ‚ùå *Sobrestima el ingreso t√≠pico*
- Mediana: $35K ‚úÖ *Mejor representaci√≥n*
- **Recomendaci√≥n**: Usar mediana

#### **Caso 2: Calificaciones de Examen** üìö
**Problema**: Calificaciones tienden a ser sim√©tricas alrededor del promedio

**Datos ejemplo**: 65, 70, 75, 78, 80, 82, 85, 88, 90
- Media: 79.2 ‚úÖ *Representativa*
- Mediana: 80 ‚úÖ *Tambi√©n representativa*
- **Recomendaci√≥n**: Cualquiera es v√°lida, media es m√°s com√∫n

#### **Caso 3: Preferencias de Marca** üè™
**Problema**: Queremos saber qu√© marca es m√°s popular

**Datos**: Nike (45%), Adidas (30%), Puma (15%), Otras (10%)
- Moda: Nike ‚úÖ *√önica medida aplicable*
- **Recomendaci√≥n**: Solo la moda tiene sentido

## üß† **PARTE 6: Conceptos Avanzados**

### **üìä Media Ponderada**

Cuando algunos valores tienen m√°s "peso" o importancia que otros:

**F√≥rmula**:
```
xÃÑ‚Çö = Œ£(w·µ¢ √ó x·µ¢) / Œ£w·µ¢
```

**Ejemplo**: Calificaci√≥n final del curso
- Parciales (20% cada uno): 85, 90
- Final (60%): 78

**C√°lculo**:
```
Calificaci√≥n = (0.20√ó85) + (0.20√ó90) + (0.60√ó78)
             = 17 + 18 + 46.8 = 81.8
```

### **üìä Media Geom√©trica**

√ötil para tasas de crecimiento, porcentajes, ratios:

**F√≥rmula**:
```
Media Geom√©trica = ‚Åø‚àö(x‚ÇÅ √ó x‚ÇÇ √ó ... √ó x‚Çô)
```

**Ejemplo**: Crecimiento anual de ventas: +10%, +20%, -5%
```
Media Geom√©trica = ¬≥‚àö(1.10 √ó 1.20 √ó 0.95) = ¬≥‚àö1.254 = 1.078
Crecimiento promedio anual = 7.8%
```

### **üìä Media Arm√≥nica**

√ötil para promediar velocidades, tasas, ratios:

**F√≥rmula**:
```
Media Arm√≥nica = n / (1/x‚ÇÅ + 1/x‚ÇÇ + ... + 1/x‚Çô)
```

**Ejemplo**: Velocidades promedio en un viaje de ida y vuelta
- Ida: 60 km/h
- Vuelta: 40 km/h
- Media arm√≥nica: 2/(1/60 + 1/40) = 48 km/h

````python
# Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
from sklearn.datasets import load_iris
warnings.filterwarnings('ignore')
````

````python
# Dataset Iris
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
````

# C√ÅLCULO MANUAL DE MEDIDAS DE TENDENCIA CENTRAL

````python
print("üßÆ C√ÅLCULO MANUAL VS AUTOM√ÅTICO - COMPRENDIENDO LOS CONCEPTOS")
print("="*70)

# Tomemos una muestra peque√±a para c√°lculo manual
sample_ages = [18, 19, 20, 20, 21, 22, 24, 25, 28, 35]
print(f"üìä Muestra de edades para an√°lisis: {sample_ages}")

# MEDIA ARITM√âTICA - C√°lculo manual
print("\n1Ô∏è‚É£ MEDIA ARITM√âTICA:")
suma_manual = sum(sample_ages)
n_manual = len(sample_ages)
media_manual = suma_manual / n_manual

print(f"   üìã C√°lculo paso a paso:")
print(f"      Suma de valores: {' + '.join(map(str, sample_ages))} = {suma_manual}")
print(f"      N√∫mero de datos: {n_manual}")
print(f"      Media = {suma_manual} √∑ {n_manual} = {media_manual:.2f} a√±os")

# Comparar con c√°lculo autom√°tico
media_numpy = np.mean(sample_ages)
media_pandas = pd.Series(sample_ages).mean()

print(f"   üêç Python - numpy: {media_numpy:.2f}")
print(f"   üêº Python - pandas: {media_pandas:.2f}")
print(f"   ‚úÖ ¬°Los resultados coinciden!")

# MEDIANA - C√°lculo manual
print("\n2Ô∏è‚É£ MEDIANA:")
sample_sorted = sorted(sample_ages)
print(f"   üìã Datos ordenados: {sample_sorted}")
print(f"   üìè N√∫mero de datos: {len(sample_sorted)} (par)")

# Para n√∫mero par de datos
pos1 = len(sample_sorted) // 2 - 1  # Posici√≥n antes del centro (√≠ndice)
pos2 = len(sample_sorted) // 2      # Posici√≥n despu√©s del centro (√≠ndice)
mediana_manual = (sample_sorted[pos1] + sample_sorted[pos2]) / 2

print(f"   üìç Posiciones centrales: {pos1+1} y {pos2+1} (valores: {sample_sorted[pos1]} y {sample_sorted[pos2]})")
print(f"   üéØ Mediana = ({sample_sorted[pos1]} + {sample_sorted[pos2]}) √∑ 2 = {mediana_manual:.1f} a√±os")

# Comparar con c√°lculo autom√°tico
mediana_numpy = np.median(sample_ages)
mediana_pandas = pd.Series(sample_ages).median()

print(f"   üêç Python - numpy: {mediana_numpy:.1f}")
print(f"   üêº Python - pandas: {mediana_pandas:.1f}")
print(f"   ‚úÖ ¬°Los resultados coinciden!")

# MODA - C√°lculo manual
print("\n3Ô∏è‚É£ MODA:")
from collections import Counter
contador = Counter(sample_ages)
print(f"   üìä Frecuencias: {dict(contador)}")

# Encontrar el valor(es) con mayor frecuencia
max_frecuencia = max(contador.values())
modas = [valor for valor, freq in contador.items() if freq == max_frecuencia]

print(f"   üîù Frecuencia m√°xima: {max_frecuencia}")
print(f"   üéØ Moda(s): {modas}")

# Comparar con c√°lculo autom√°tico usando scipy.stats
moda_scipy = stats.mode(sample_ages, keepdims=True)
print(f"   üêç Python - scipy: {moda_scipy.mode[0]} (frecuencia: {moda_scipy.count[0]})")
````

# AN√ÅLISIS DEL IMPACTO DE OUTLIERS

````python
print("\n" + "="*70)
print("üéØ IMPACTO DE VALORES AT√çPICOS (OUTLIERS)")
print("="*70)

# Datos SIN outliers
datos_normales = [30, 32, 31, 33, 29, 35, 28, 34, 30, 32]
print(f"üìä Salarios normales (en miles): {datos_normales}")

media_normal = np.mean(datos_normales)
mediana_normal = np.median(datos_normales)
moda_normal = stats.mode(datos_normales, keepdims=True).mode[0]

print(f"   Media: ${media_normal:.1f}K")
print(f"   Mediana: ${mediana_normal:.1f}K") 
print(f"   Moda: ${moda_normal}K")

# Datos CON outlier (agregar CEO con salario alto)
datos_con_outlier = datos_normales + [500]  # CEO gana $500K
print(f"\nüìä Salarios CON outlier (CEO): {datos_con_outlier}")

media_outlier = np.mean(datos_con_outlier)
mediana_outlier = np.median(datos_con_outlier)
moda_outlier = stats.mode(datos_con_outlier, keepdims=True).mode[0]

print(f"   Media: ${media_outlier:.1f}K")
print(f"   Mediana: ${mediana_outlier:.1f}K")
print(f"   Moda: ${moda_outlier}K")

# An√°lisis del impacto
print(f"\nüìà IMPACTO DEL OUTLIER:")
cambio_media = ((media_outlier - media_normal) / media_normal) * 100
cambio_mediana = ((mediana_outlier - mediana_normal) / mediana_normal) * 100

print(f"   Media cambi√≥: {cambio_media:.1f}% üìà ¬°MUY AFECTADA!")
print(f"   Mediana cambi√≥: {cambio_mediana:.1f}% üìä Pr√°cticamente sin cambio")

# Visualizaci√≥n del impacto de outliers
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Gr√°fico 1: Sin outlier
ax1.hist(datos_normales, bins=8, alpha=0.7, color='skyblue', edgecolor='black')
ax1.axvline(media_normal, color='red', linestyle='--', linewidth=2, label=f'Media: ${media_normal:.1f}K')
ax1.axvline(mediana_normal, color='green', linestyle='--', linewidth=2, label=f'Mediana: ${mediana_normal:.1f}K')
ax1.set_title('Distribuci√≥n SIN Outlier')
ax1.set_xlabel('Salario (miles)')
ax1.set_ylabel('Frecuencia')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Gr√°fico 2: Con outlier
ax2.hist(datos_con_outlier, bins=10, alpha=0.7, color='coral', edgecolor='black')
ax2.axvline(media_outlier, color='red', linestyle='--', linewidth=2, label=f'Media: ${media_outlier:.1f}K')
ax2.axvline(mediana_outlier, color='green', linestyle='--', linewidth=2, label=f'Mediana: ${mediana_outlier:.1f}K')
ax2.set_title('Distribuci√≥n CON Outlier')
ax2.set_xlabel('Salario (miles)')
ax2.set_ylabel('Frecuencia')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"üí° CONCLUSI√ìN: La mediana es m√°s ROBUSTA ante outliers que la media")
````

# VISUALIZACI√ìN AVANZADA DE MEDIDAS DE TENDENCIA CENTRAL

````python
print("\n" + "="*70)
print("üìä VISUALIZACI√ìN AVANZADA - DISTRIBUCIONES Y MEDIDAS CENTRALES")
print("="*70)

# Crear visualizaci√≥n de 3 variables num√©ricas
fig, axes = plt.subplots(2, 4, figsize=(18, 12))
fig.suptitle('Medidas de Tendencia Central en Dataset de Estudiantes', fontsize=16, y=0.98)

variables_plot = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']

for i, var in enumerate(variables_plot):
    datos = iris_df[var]
    
    # Histograma con medidas centrales (fila superior)
    ax1 = axes[0, i]
    ax1.hist(datos, bins=20, alpha=0.7, color='lightblue', edgecolor='black')
    
    media = datos.mean()
    mediana = datos.median()
    moda_vals = datos.mode()
    if len(moda_vals) > 0:
        moda = moda_vals.iloc[0]
        ax1.axvline(moda, color='orange', linestyle=':', linewidth=2, label=f'Moda: {moda:.1f}')
    
    ax1.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.1f}')
    ax1.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.1f}')
    
    ax1.set_title(f'Distribuci√≥n de {var.replace("_", " ").title()}')
    ax1.set_xlabel(var.replace("_", " ").title())
    ax1.set_ylabel('Frecuencia')
    ax1.legend(fontsize=8)
    ax1.grid(True, alpha=0.3)
    
    # Box plot (fila inferior)
    ax2 = axes[1, i]
    box_data = ax2.boxplot([datos], labels=[var.replace("_", " ").title()], patch_artist=True)
    box_data['boxes'][0].set_facecolor('lightgreen')
    box_data['boxes'][0].set_alpha(0.7)
    
    # Agregar media al boxplot
    ax2.plot(1, media, 'ro', markersize=8, label=f'Media: {media:.1f}')
    ax2.set_title(f'Box Plot - {var.replace("_", " ").title()}')
    ax2.legend(fontsize=8)
    ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt

print("\n" + "="*70)
print("üè∑Ô∏è AN√ÅLISIS DE VARIABLES CATEG√ìRICAS - SOLO LA MODA APLICA")
print("="*70)

variables_categoricas = ['species']

for var in variables_categoricas:
    print(f"\nüìä {var.upper()}:")
    
    # Calcular frecuencias
    frecuencias = iris_df[var].value_counts()
    porcentajes = iris_df[var].value_counts(normalize=True) * 100
    
    # La moda es la categor√≠a con mayor frecuencia
    moda_cat = frecuencias.index[0]
    freq_moda = frecuencias.iloc[0]
    pct_moda = porcentajes.iloc[0]
    
    print(f"   üèÜ MODA: {moda_cat}")
    print(f"   üìà Frecuencia: {freq_moda} ({pct_moda:.1f}%)")
    print(f"\n   üìã Distribuci√≥n completa:")
    
    for categoria, freq in frecuencias.items():
        pct = (freq / len(iris_df)) * 100
        barra = "‚ñà" * int(pct / 2)  # Barra visual
        print(f"      {categoria:12}: {freq:3d} ({pct:5.1f}%) {barra}")

# ==========================================
# üîç Visualizaci√≥n de la variable categ√≥rica
# ==========================================

# Crear figura
fig, ax = plt.subplots(figsize=(7, 6))

# Datos
genero_counts = iris_df['species'].value_counts()
colors1 = ['lightcoral', 'lightblue', 'lightgreen']

# Gr√°fico de pastel
ax.pie(genero_counts.values,
       labels=genero_counts.index,
       autopct='%1.1f%%',
       colors=colors1,
       startangle=90,
       textprops={'fontsize': 12})

# T√≠tulo con la moda
ax.set_title('Distribuci√≥n por Especie\n(Moda: ' + genero_counts.index[0] + ')', fontsize=14)

plt.tight_layout()
plt.show()
````
