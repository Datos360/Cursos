## 📚 Contenido Teórico

### 🎨 Principios de Visualización Efectiva

#### La Gramática de los Gráficos
Los gráficos estadísticos siguen una "gramática" que incluye elementos fundamentales:

1. **Datos (Data):** La información que queremos representar  
2. **Estética (Aesthetics):** Cómo mapeamos los datos a propiedades visuales  
3. **Geometrías (Geometries):** Las formas gráficas que usamos  
4. **Escalas (Scales):** Cómo traducimos datos a propiedades visuales  
5. **Facetas (Facets):** Divisiones o paneles del gráfico  
6. **Estadísticas (Statistics):** Transformaciones de los datos  
7. **Coordenadas (Coordinates):** El sistema de coordenadas usado  

#### Principios de Diseño Fundamentales

- **🎯 Claridad:** El mensaje debe ser inmediatamente comprensible  
  - Eliminar elementos innecesarios (*principio de simplicidad*)  
  - Usar etiquetas claras y descriptivas  
  - Mantener proporciones adecuadas  

- **📊 Precisión:** Representar los datos de forma exacta  
  - Evitar distorsiones en las escalas  
  - No comenzar ejes en valores arbitrarios (salvo justificación)  
  - Usar escalas apropiadas para el tipo de datos  

- **⚡ Eficiencia:** Maximizar la información por "tinta" utilizada  
  - Eliminar redundancias innecesarias  
  - Optimizar el ratio datos/tinta  
  - Usar colores y formas con propósito  

- **🎨 Atractivo:** Generar interés visual sin distraer del mensaje  
  - Usar paletas de colores apropiadas  
  - Mantener consistencia visual  
  - Equilibrar elementos estéticos y funcionales  

---

## 📊 Tipos de Gráficos y Cuándo Usarlos

### Para Variables Categóricas

**📊 Gráficos de Barras**  
- **Uso:** Comparar cantidades entre categorías  
- **Ventajas:** Fácil comparación visual  
- **Variantes:** Simples, agrupadas, apiladas, horizontales  

**🥧 Gráficos de Torta/Pie**  
- **Uso:** Mostrar partes de un todo (máximo 5-6 categorías)  
- **Ventajas:** Intuición inmediata de proporciones  
- **Limitaciones:** Difícil comparar segmentos similares  

**📈 Barras Horizontales**  
- **Uso:** Cuando las etiquetas son largas o hay muchas categorías  
- **Ventajas:** Mejor legibilidad  

---

### Para Variables Numéricas

**📊 Histogramas**  
- **Uso:** Mostrar distribución de una variable continua  
- **Ventajas:** Revelan forma, centro y dispersión  
- **Consideraciones:** Número de *bins* afecta la interpretación  

**📦 Box Plots (Diagramas de Caja)**  
- **Uso:** Comparar distribuciones entre grupos  
- **Ventajas:** Muestran mediana, cuartiles y *outliers*  

**🎻 Violin Plots**  
- **Uso:** Similar a box plots pero mostrando densidad  
- **Ventajas:** Revelan múltiples modas  

**📈 Gráficos de Densidad**  
- **Uso:** Versión suavizada del histograma  
- **Ventajas:** Mejor para comparar distribuciones  

---

### Para Relaciones entre Variables

**🔵 Scatter Plots (Diagramas de Dispersión)**  
- **Uso:** Relaciones entre dos variables numéricas  
- **Información:** Correlación, *outliers*, patrones  

**📊 Gráficos de Líneas**  
- **Uso:** Series temporales o tendencias  
- **Ventajas:** Cambios a lo largo del tiempo  

**🔥 Mapas de Calor (Heatmaps)**  
- **Uso:** Matrices de correlación, tablas de contingencia  
- **Ventajas:** Patrones en matrices de datos  

---

### Gráficos Especializados

- **📊 Pair Plots:** Relaciones entre múltiples variables  
- **📈 Regresión:** Relación con línea de tendencia  
- **🎨 Multivariados:** Más de 2-3 variables simultáneamente (color, tamaño, forma, facetas)

## 🎨 Teoría del Color en Visualización

### Tipos de Escalas de Color

- **🌈 Escalas Cualitativas** → Variables categóricas sin orden  
  Ej.: `Set1`, `Set2`, `tab10`  

- **📊 Escalas Secuenciales** → Variables numéricas ordenadas  
  Ej.: `Blues`, `Reds`, `viridis`  

- **🔄 Escalas Divergentes** → Variables con punto central significativo  
  Ej.: `RdBu`, `coolwarm`, `seismic`  

**Accesibilidad:**  
- Evitar rojo-verde para daltónicos  
- Asegurar contraste suficiente  
- Verificar en escala de grises si será impreso

## 📂 Datasets para Práctica Avanzada

````python
# Cargar múltiples datasets para diferentes tipos de análisis

# Dataset 1: Iris (ya conocido)
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species_name'] = iris_df['species'].map({
    0: 'setosa', 1: 'versicolor', 2: 'virginica'
})

# Dataset 2: Wine (clasificación de vinos)
wine = load_wine()
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['wine_class'] = wine.target
wine_df['wine_type'] = wine_df['wine_class'].map({
    0: 'Class_0', 1: 'Class_1', 2: 'Class_2'
})

# Dataset 3: Datos sintéticos de empleados para práctica
np.random.seed(42)
n_employees = 500

employees_df = pd.DataFrame({
    'employee_id': range(1, n_employees + 1),
    'department': np.random.choice(['IT', 'Sales', 'Marketing', 'HR', 'Finance'], n_employees),
    'experience_years': np.random.exponential(scale=3, size=n_employees),
    'salary': np.random.normal(50000, 15000, n_employees),
    'performance_score': np.random.normal(75, 12, n_employees),
    'age': np.random.normal(35, 8, n_employees),
    'education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], n_employees),
    'remote_work': np.random.choice(['Yes', 'No'], n_employees, p=[0.3, 0.7])
})

# Ajustes realistas
employees_df['salary'] = np.where(
    employees_df['department'] == 'IT', 
    employees_df['salary'] * 1.2, 
    employees_df['salary']
)
employees_df['salary'] = np.maximum(employees_df['salary'], 25000)  # Salario mínimo
employees_df['age'] = np.clip(employees_df['age'], 22, 65)  # Rango de edad realista
employees_df['experience_years'] = np.clip(employees_df['experience_years'], 0, 40)

print("📊 Datasets cargados:")
print(f"• Iris: {iris_df.shape} - Clasificación de flores")
print(f"• Wine: {wine_df.shape} - Clasificación de vinos") 
print(f"• Employees: {employees_df.shape} - Datos de empleados")
print("\n🔍 Primeras filas del dataset de empleados:")
print(employees_df.head())
````

````python
print("🎨 PERSONALIZACIÓN AVANZADA DE MATPLOTLIB")
print("=" * 50)

# Función para crear estilo personalizado
def setup_custom_style():
    """Configura un estilo personalizado para gráficos profesionales"""
    
    # Colores personalizados
    colors = {
        'primary': '#2E86AB',
        'secondary': '#A23B72', 
        'accent': '#F18F01',
        'success': '#C73E1D',
        'neutral': '#6C757D',
        'light': '#F8F9FA',
        'dark': '#212529'
    }
    
    # Configuración del estilo
    plt.rcParams.update({
        # Figura
        'figure.facecolor': colors['light'],
        'figure.edgecolor': 'none',
        
        # Ejes
        'axes.facecolor': 'white',
        'axes.edgecolor': colors['neutral'],
        'axes.linewidth': 0.8,
        'axes.grid': True,
        'axes.axisbelow': True,
        
        # Grid
        'grid.color': colors['neutral'],
        'grid.alpha': 0.3,
        'grid.linewidth': 0.5,
        
        # Colores por defecto
        'axes.prop_cycle': plt.cycler('color', [
            colors['primary'], colors['secondary'], colors['accent'], 
            colors['success'], colors['neutral']
        ]),
        
        # Texto
        'text.color': colors['dark'],
        'axes.labelcolor': colors['dark'],
        'xtick.color': colors['dark'],
        'ytick.color': colors['dark'],
        
        # Spines
        'axes.spines.left': True,
        'axes.spines.bottom': True,
        'axes.spines.top': False,
        'axes.spines.right': False,
    })
    
    return colors

# Aplicar estilo personalizado
custom_colors = setup_custom_style()

# Ejemplo de gráfico con estilo personalizado
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Ejemplos de Visualización con Estilo Personalizado', 
             fontsize=16, fontweight='bold', y=0.95)

# Gráfico 1: Barras con estilo
dept_counts = employees_df['department'].value_counts()
axes[0, 0].bar(dept_counts.index, dept_counts.values, 
               color=custom_colors['primary'], alpha=0.8)
axes[0, 0].set_title('Empleados por Departamento')
axes[0, 0].set_ylabel('Número de Empleados')
axes[0, 0].tick_params(axis='x', rotation=45)

# Gráfico 2: Histograma con múltiples elementos
axes[0, 1].hist(employees_df['salary'], bins=30, 
                color=custom_colors['secondary'], alpha=0.7, edgecolor='white')
axes[0, 1].axvline(employees_df['salary'].mean(), 
                   color=custom_colors['accent'], linestyle='--', linewidth=2,
                   label=f'Media: ${employees_df["salary"].mean():.0f}')
axes[0, 1].set_title('Distribución de Salarios')
axes[0, 1].set_xlabel('Salario ($)')
axes[0, 1].set_ylabel('Frecuencia')
axes[0, 1].legend()

# Gráfico 3: Scatter plot avanzado
scatter = axes[1, 0].scatter(employees_df['experience_years'], employees_df['salary'],
                            c=employees_df['performance_score'], 
                            cmap='viridis', alpha=0.6, s=50)
axes[1, 0].set_title('Relación Experiencia-Salario por Performance')
axes[1, 0].set_xlabel('Años de Experiencia')
axes[1, 0].set_ylabel('Salario ($)')
plt.colorbar(scatter, ax=axes[1, 0], label='Score de Performance')

# Gráfico 4: Box plot con personalización
education_order = ['High School', 'Bachelor', 'Master', 'PhD']
edu_data = [employees_df[employees_df['education'] == edu]['salary'] 
            for edu in education_order]

box_plot = axes[1, 1].boxplot(edu_data, labels=education_order, patch_artist=True)
for patch, color in zip(box_plot['boxes'], 
                       [custom_colors['primary'], custom_colors['secondary'], 
                        custom_colors['accent'], custom_colors['success']]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

axes[1, 1].set_title('Salarios por Nivel Educativo')
axes[1, 1].set_xlabel('Educación')
axes[1, 1].set_ylabel('Salario ($)')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("✅ Estilo personalizado aplicado exitosamente")
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# Helpers: crea dataframes si no existen en tu entorno
# ============================================================
if "iris_df" not in globals():
    from sklearn.datasets import load_iris
    iris = load_iris()
    iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
    iris_df["species"] = iris.target
    iris_df["species_name"] = iris_df["species"].map(dict(enumerate(iris.target_names)))

if "employees_df" not in globals():
    rng = np.random.default_rng(42)
    n = 300
    departments = np.array(["Ventas", "Marketing", "TI", "Finanzas", "RRHH"])
    employees_df = pd.DataFrame({
        "department": rng.choice(departments, size=n, replace=True),
        "salary": rng.normal(65000, 15000, size=n).clip(25000, 120000),
        "experience_years": rng.normal(7, 4, size=n).clip(0, 35),
        "performance_score": rng.integers(50, 100, size=n)
    })

print("\n🎯 GRÁFICOS ESPECIALIZADOS PARA INSIGHTS")
print("=" * 50)

# ============================================================
# 1) Ridge Plot (densidades apiladas) por departamento
# ============================================================
def create_ridge_plot(df_employees):
    """Crear un ridge plot para comparar distribuciones de salarios por departamento."""
    from scipy.stats import gaussian_kde

    depts = sorted(df_employees["department"].unique().tolist())
    fig, ax = plt.subplots(figsize=(12, 8))

    # Colormap consistente pero sin forzar colores específicos si no quieres
    cmap = plt.cm.Set3(np.linspace(0, 1, len(depts)))

    y_step = 1.2
    for i, (dept, color) in enumerate(zip(depts, cmap)):
        dept_data = df_employees.loc[df_employees["department"] == dept, "salary"].dropna().values

        if len(dept_data) < 2:
            # KDE necesita al menos 2 puntos
            y_offset = i * y_step
            ax.text(dept_data.min() if len(dept_data) else 0, y_offset + 0.1,
                    f"{dept} (sin datos suficientes)", fontsize=9)
            continue

        kde = gaussian_kde(dept_data)
        x_min, x_max = np.percentile(dept_data, [1, 99])
        x = np.linspace(x_min, x_max, 200)
        d = kde(x)

        d_norm = d / d.max() * 0.8
        y_offset = i * y_step

        ax.fill_between(x, y_offset, y_offset + d_norm, alpha=0.7, color=color, label=dept)
        ax.plot(x, y_offset + d_norm, linewidth=1, color="black")
        ax.text(x_min - 0.05*(x_max-x_min), y_offset + 0.4, dept, fontweight="bold", va="center")

    ax.set_xlabel("Salario ($)")
    ax.set_title("Distribución de Salarios por Departamento (Ridge Plot)", fontsize=14, fontweight="bold")
    ax.set_yticks([])
    for spine in ["left", "top", "right"]:
        ax.spines[spine].set_visible(False)
    plt.tight_layout()
    return fig

ridge_fig = create_ridge_plot(employees_df)
plt.show()

# ============================================================
# 2) Gráfico de Coordenadas Paralelas (Iris)
# ============================================================
print("\n🔗 Creando Gráfico de Coordenadas Paralelas...")

from pandas.plotting import parallel_coordinates

# Garantizar columna species_name
if "species_name" not in iris_df.columns:
    if "species" in iris_df.columns:
        # si es numérica, mapear a nombres básicos
        map_default = {0: "setosa", 1: "versicolor", 2: "virginica"}
        iris_df["species_name"] = iris_df["species"].map(map_default).fillna("species")
    else:
        iris_df["species_name"] = "species"

# Seleccionar muestra para mantener claridad visual
num_rows = min(60, len(iris_df))
iris_sample = iris_df.sample(n=num_rows, random_state=42)

cols_four = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
# Estandarizar columnas para que las escalas sean comparables en el gráfico
iris_std = iris_sample.copy()
for c in cols_four:
    iris_std[c] = (iris_std[c] - iris_std[c].mean()) / (iris_std[c].std(ddof=0) + 1e-9)

fig, ax = plt.subplots(figsize=(12, 6))
parallel_coordinates(iris_std, 'species_name', cols=cols_four, ax=ax, alpha=0.7)
ax.set_title('Coordenadas Paralelas - Características de Iris')
ax.set_ylabel('Valores Estandarizados')
ax.legend(title='Especie', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ============================================================
# 3) Gráfico de Radar/Spider (promedios por especie)
# ============================================================
def create_radar_chart(df_iris, feature_cols, label_col="species_name"):
    """Radar chart comparando perfiles (normalizados por columna)."""
    # Promedios por especie
    prof = df_iris.groupby(label_col)[feature_cols].mean()

    # Normalizar cada columna a [0,1] para comparabilidad
    prof_norm = (prof - prof.min()) / (prof.max() - prof.min() + 1e-9)

    categories = [c.replace("(cm)", "").title() for c in feature_cols]
    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    fig, axes = plt.subplots(1, len(prof_norm), figsize=(16, 6), subplot_kw=dict(projection='polar'))
    if len(prof_norm) == 1:
        axes = [axes]  # asegurar iterable

    fig.suptitle('Perfiles de Especies - Radar Chart', fontsize=16, fontweight='bold')

    for ax, (species, row) in zip(axes, prof_norm.iterrows()):
        vals = row.values.tolist()
        vals += vals[:1]  # cerrar círculo

        ax.plot(angles, vals, 'o-', linewidth=2)
        ax.fill(angles, vals, alpha=0.25)
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(categories)
        ax.set_yticklabels([])  # quitar etiquetas radiales para limpiar
        ax.set_title(f'Perfil: {species}', fontweight='bold')
        ax.grid(True)

    plt.tight_layout()
    return fig

radar_fig = create_radar_chart(iris_df, ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
plt.show()

# ============================================================
# 4) Gráfico de Burbujas Multidimensional (experiencia vs salario vs performance)
# ============================================================
print("\n💭 Creando Gráfico de Burbujas Multidimensional...")

fig, ax = plt.subplots(figsize=(12, 8))

dept_unique = sorted(employees_df['department'].unique().tolist())
cmap = plt.cm.Set1(np.linspace(0, 1, len(dept_unique)))
handles = []

for color, dept in zip(cmap, dept_unique):
    sub = employees_df[employees_df['department'] == dept]
    h = ax.scatter(
        sub['experience_years'],
        sub['salary'],
        s=sub['performance_score'] * 3,  # tamaño ~ performance
        alpha=0.6,
        color=color,
        edgecolors='black',
        linewidth=0.5,
        label=dept
    )
    handles.append(h)

ax.set_xlabel('Años de Experiencia')
ax.set_ylabel('Salario ($)')
ax.set_title('Experiencia vs Salario vs Performance por Departamento\n(Tamaño de burbuja = Performance Score)',
             fontweight='bold')

# Leyenda de categorías (departamentos)
legend1 = ax.legend(handles=handles, title='Departamento', loc='upper left', frameon=True)
ax.add_artist(legend1)

# Leyenda para tamaños (performance)
sizes_demo = [50, 70, 90]
labels_demo = [f'Score: {s}' for s in sizes_demo]
legend_bubbles = [
    plt.scatter([], [], s=s*3, alpha=0.6, color='gray', edgecolors='black', linewidth=0.5)
    for s in sizes_demo
]
legend2 = ax.legend(legend_bubbles, labels_demo, title='Performance Score', loc='upper right', frameon=True)

plt.tight_layout()
plt.show()
````

### 📊 Dashboards y Paneles Informativos

````python
print("\n📊 CREACIÓN DE DASHBOARDS INFORMATIVOS")
print("=" * 50)

def create_comprehensive_dashboard():
    """Crear un dashboard completo con múltiples visualizaciones"""
    
    # Configurar el layout del dashboard
    fig = plt.figure(figsize=(20, 16))
    gs = gridspec.GridSpec(4, 4, figure=fig, hspace=0.3, wspace=0.3)
    
    # Título principal del dashboard
    fig.suptitle('Dashboard Ejecutivo - Análisis de Recursos Humanos', 
                 fontsize=20, fontweight='bold', y=0.95)
    
    # Panel 1: KPIs principales (texto)
    ax1 = fig.add_subplot(gs[0, :2])
    ax1.axis('off')
    
    # Calcular KPIs
    total_employees = len(employees_df)
    avg_salary = employees_df['salary'].mean()
    avg_performance = employees_df['performance_score'].mean()
    remote_percentage = (employees_df['remote_work'] == 'Yes').mean() * 100
    
    kpi_text = f"""
    📊 INDICADORES CLAVE DE RENDIMIENTO
    
    👥 Total de Empleados: {total_employees:,}
    💰 Salario Promedio: ${avg_salary:,.0f}
    ⭐ Performance Promedio: {avg_performance:.1f}/100
    🏠 Trabajo Remoto: {remote_percentage:.1f}%
    """
    
    ax1.text(0.05, 0.5, kpi_text, fontsize=12, fontweight='bold',
             verticalalignment='center', transform=ax1.transAxes,
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.3))
    
    # Panel 2: Distribución por departamento
    ax2 = fig.add_subplot(gs[0, 2:])
    dept_counts = employees_df['department'].value_counts()
    wedges, texts, autotexts = ax2.pie(dept_counts.values, labels=dept_counts.index, 
                                      autopct='%1.1f%%', startangle=90)
    ax2.set_title('Distribución de Empleados por Departamento', fontweight='bold')
    
    # Panel 3: Histograma de salarios
    ax3 = fig.add_subplot(gs[1, :2])
    ax3.hist(employees_df['salary'], bins=25, alpha=0.7, color=custom_colors['primary'])
    ax3.axvline(avg_salary, color='red', linestyle='--', linewidth=2, 
                label=f'Media: ${avg_salary:.0f}')
    ax3.set_title('Distribución de Salarios', fontweight='bold')
    ax3.set_xlabel('Salario ($)')
    ax3.set_ylabel('Frecuencia')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Panel 4: Box plot salarios por departamento
    ax4 = fig.add_subplot(gs[1, 2:])
    employees_df.boxplot(column='salary', by='department', ax=ax4)
    ax4.set_title('Salarios por Departamento')
    ax4.set_xlabel('Departamento')
    ax4.set_ylabel('Salario ($)')
    ax4.tick_params(axis='x', rotation=45)
    plt.setp(ax4.xaxis.get_majorticklabels(), rotation=45)
    
    # Panel 5: Scatter plot experiencia vs salario
    ax5 = fig.add_subplot(gs[2, :2])
    scatter = ax5.scatter(employees_df['experience_years'], employees_df['salary'],
                         c=employees_df['performance_score'], cmap='viridis',
                         alpha=0.6, s=30)
    ax5.set_title('Experiencia vs. Salario (color = Performance)', fontweight='bold')
    ax5.set_xlabel('Años de Experiencia')
    ax5.set_ylabel('Salario ($)')
    plt.colorbar(scatter, ax=ax5, label='Performance Score')
    
    # Panel 6: Heatmap de correlaciones
    ax6 = fig.add_subplot(gs[2, 2:])
    corr_matrix = employees_df[['experience_years', 'salary', 'performance_score', 'age']].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                square=True, fmt='.2f', ax=ax6, cbar_kws={'shrink': 0.8})
    ax6.set_title('Matriz de Correlaciones', fontweight='bold')
    
    # Panel 7: Performance por educación
    ax7 = fig.add_subplot(gs[3, :2])
    education_order = ['High School', 'Bachelor', 'Master', 'PhD']
    sns.boxplot(data=employees_df, x='education', y='performance_score', 
                order=education_order, ax=ax7)
    ax7.set_title('Performance por Nivel Educativo', fontweight='bold')
    ax7.set_xlabel('Educación')
    ax7.set_ylabel('Performance Score')
    ax7.tick_params(axis='x', rotation=45)
    
    # Panel 8: Trabajo remoto por departamento
    ax8 = fig.add_subplot(gs[3, 2:])
    remote_by_dept = pd.crosstab(employees_df['department'], employees_df['remote_work'], 
                                normalize='index') * 100
    remote_by_dept.plot(kind='bar', stacked=True, ax=ax8, 
                       color=[custom_colors['accent'], custom_colors['primary']])
    ax8.set_title('Trabajo Remoto por Departamento (%)', fontweight='bold')
    ax8.set_xlabel('Departamento')
    ax8.set_ylabel('Porcentaje')
    ax8.legend(title='Trabajo Remoto')
    ax8.tick_params(axis='x', rotation=45)
    
    return fig

# Crear el dashboard
dashboard_fig = create_comprehensive_dashboard()
plt.show()

print("✅ Dashboard completo creado exitosamente")
````

### 🔧 Herramientas y Templates Reutilizables

````python
print("\n🔧 HERRAMIENTAS Y TEMPLATES REUTILIZABLES")
print("=" * 55)

class VisualizationToolkit:
    """
    Clase con herramientas avanzadas para visualización de datos
    """
    
    def __init__(self):
        self.setup_style()
        
    def setup_style(self):
        """Configurar estilo predeterminado"""
        self.colors = {
            'primary': '#2E86AB',
            'secondary': '#A23B72',
            'accent': '#F18F01',
            'success': '#C73E1D',
            'neutral': '#6C757D'
        }
        
        plt.rcParams.update({
            'axes.prop_cycle': plt.cycler('color', list(self.colors.values())),
            'font.size': 10,
            'axes.titlesize': 12,
            'axes.labelsize': 11,
            'figure.titlesize': 14
        })
    
    def correlation_heatmap(self, df, figsize=(10, 8), annot=True):
        """Crear heatmap de correlaciones optimizado"""
        
        # Seleccionar solo columnas numéricas
        numeric_df = df.select_dtypes(include=[np.number])
        
        if numeric_df.empty:
            print("❌ No se encontraron columnas numéricas")
            return None
        
        # Calcular correlaciones
        corr_matrix = numeric_df.corr()
        
        # Crear máscara para triángulo superior
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
        
        # Crear figura
        fig, ax = plt.subplots(figsize=figsize)
        
        # Heatmap
        sns.heatmap(corr_matrix, mask=mask, annot=annot, cmap='coolwarm',
                   center=0, square=True, fmt='.2f',
                   cbar_kws={'shrink': 0.8}, ax=ax)
        
        ax.set_title('Matriz de Correlaciones', fontweight='bold', pad=20)
        
        return fig
    
    def distribution_comparison(self, df, column, group_by, figsize=(15, 5)):
        """Comparar distribuciones entre grupos"""
        
        fig, axes = plt.subplots(1, 3, figsize=figsize)
        fig.suptitle(f'Análisis de Distribución: {column} por {group_by}', 
                    fontweight='bold')
        
        # 1. Histogramas superpuestos
        groups = df[group_by].unique()
        for group in groups:
            group_data = df[df[group_by] == group][column]
            axes[0].hist(group_data, alpha=0.7, label=str(group), bins=20)
        
        axes[0].set_title('Histogramas Superpuestos')
        axes[0].set_xlabel(column)
        axes[0].set_ylabel('Frecuencia')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        # 2. Box plots
        df.boxplot(column=column, by=group_by, ax=axes[1])
        axes[1].set_title('Box Plots Comparativos')
        axes[1].set_xlabel(group_by)
        axes[1].set_ylabel(column)
        
        # 3. Violin plots
        sns.violinplot(data=df, x=group_by, y=column, ax=axes[2])
        axes[2].set_title('Violin Plots')
        axes[2].set_xlabel(group_by)
        axes[2].set_ylabel(column)
        axes[2].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        return fig
    
    def create_summary_stats_viz(self, df, columns=None):
        """Crear visualización de estadísticas resumidas"""
        
        if columns is None:
            columns = df.select_dtypes(include=[np.number]).columns.tolist()
        
        if not columns:
            print("❌ No se encontraron columnas numéricas")
            return None
        
        # Calcular estadísticas
        stats_df = df[columns].describe().T
        stats_df['cv'] = stats_df['std'] / stats_df['mean']  # Coeficiente de variación
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Resumen Estadístico de Variables Numéricas', fontweight='bold')
        
        # 1. Medidas de tendencia central
        stats_df[['mean', '50%']].plot(kind='bar', ax=axes[0, 0])
        axes[0, 0].set_title('Medidas de Tendencia Central')
        axes[0, 0].set_ylabel('Valor')
        axes[0, 0].legend(['Media', 'Mediana'])
        axes[0, 0].tick_params(axis='x', rotation=45)
        
        # 2. Medidas de dispersión
        stats_df[['std', 'cv']].plot(kind='bar', ax=axes[0, 1], secondary_y=True)
        axes[0, 1].set_title('Medidas de Dispersión')
        axes[0, 1].set_ylabel('Desviación Estándar')
        axes[0, 1].right_ax.set_ylabel('Coeficiente de Variación')
        axes[0, 1].tick_params(axis='x', rotation=45)
        
        # 3. Rangos
        ranges = stats_df['max'] - stats_df['min']
        iqr = stats_df['75%'] - stats_df['25%']
        pd.DataFrame({'Rango Total': ranges, 'IQR': iqr}).plot(kind='bar', ax=axes[1, 0])
        axes[1, 0].set_title('Análisis de Rangos')
        axes[1, 0].set_ylabel('Valor')
        axes[1, 0].tick_params(axis='x', rotation=45)
        
        # 4. Tabla de estadísticas
        axes[1, 1].axis('off')
        table_data = stats_df.round(2)
        table = axes[1, 1].table(cellText=table_data.values,
                                rowLabels=table_data.index,
                                colLabels=table_data.columns,
                                cellLoc='center',
                                loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(8)
        table.scale(1, 1.5)
        axes[1, 1].set_title('Tabla Resumen', fontweight='bold')
        
        plt.tight_layout()
        return fig

# Instanciar y usar el toolkit
viz_toolkit = VisualizationToolkit()

# Ejemplo 1: Heatmap de correlaciones
print("\n🔥 Creando Heatmap de Correlaciones...")
corr_fig = viz_toolkit.correlation_heatmap(employees_df)
plt.show()

# Ejemplo 2: Comparación de distribuciones
print("\n📊 Comparando Distribuciones...")
dist_fig = viz_toolkit.distribution_comparison(employees_df, 'salary', 'department')
plt.show()

# Ejemplo 3: Resumen estadístico visual
print("\n📈 Resumen Estadístico Visual...")
summary_fig = viz_toolkit.create_summary_stats_viz(employees_df)
plt.show()

print("\n✅ Toolkit de visualización implementado exitosamente")
````
