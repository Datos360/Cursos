### üé® Principios de Visualizaci√≥n Efectiva

#### La Gram√°tica de los Gr√°ficos
Los gr√°ficos estad√≠sticos siguen una "gram√°tica" que incluye elementos fundamentales:

1. **Datos (Data):** La informaci√≥n que queremos representar  
2. **Est√©tica (Aesthetics):** C√≥mo mapeamos los datos a propiedades visuales  
3. **Geometr√≠as (Geometries):** Las formas gr√°ficas que usamos  
4. **Escalas (Scales):** C√≥mo traducimos datos a propiedades visuales  
5. **Facetas (Facets):** Divisiones o paneles del gr√°fico  
6. **Estad√≠sticas (Statistics):** Transformaciones de los datos  
7. **Coordenadas (Coordinates):** El sistema de coordenadas usado  

#### Principios de Dise√±o Fundamentales

- **üéØ Claridad:** El mensaje debe ser inmediatamente comprensible  
  - Eliminar elementos innecesarios (*principio de simplicidad*)  
  - Usar etiquetas claras y descriptivas  
  - Mantener proporciones adecuadas  

- **üìä Precisi√≥n:** Representar los datos de forma exacta  
  - Evitar distorsiones en las escalas  
  - No comenzar ejes en valores arbitrarios (salvo justificaci√≥n)  
  - Usar escalas apropiadas para el tipo de datos  

- **‚ö° Eficiencia:** Maximizar la informaci√≥n por "tinta" utilizada  
  - Eliminar redundancias innecesarias  
  - Optimizar el ratio datos/tinta  
  - Usar colores y formas con prop√≥sito  

- **üé® Atractivo:** Generar inter√©s visual sin distraer del mensaje  
  - Usar paletas de colores apropiadas  
  - Mantener consistencia visual  
  - Equilibrar elementos est√©ticos y funcionales

## üìä Tipos de Gr√°ficos y Cu√°ndo Usarlos

### Para Variables Categ√≥ricas

**üìä Gr√°ficos de Barras**  
- **Uso:** Comparar cantidades entre categor√≠as  
- **Ventajas:** F√°cil comparaci√≥n visual  
- **Variantes:** Simples, agrupadas, apiladas, horizontales  

**ü•ß Gr√°ficos de Torta/Pie**  
- **Uso:** Mostrar partes de un todo (m√°ximo 5-6 categor√≠as)  
- **Ventajas:** Intuici√≥n inmediata de proporciones  
- **Limitaciones:** Dif√≠cil comparar segmentos similares  

**üìà Barras Horizontales**  
- **Uso:** Cuando las etiquetas son largas o hay muchas categor√≠as  
- **Ventajas:** Mejor legibilidad  

---

### Para Variables Num√©ricas

**üìä Histogramas**  
- **Uso:** Mostrar distribuci√≥n de una variable continua  
- **Ventajas:** Revelan forma, centro y dispersi√≥n  
- **Consideraciones:** N√∫mero de *bins* afecta la interpretaci√≥n  

**üì¶ Box Plots (Diagramas de Caja)**  
- **Uso:** Comparar distribuciones entre grupos  
- **Ventajas:** Muestran mediana, cuartiles y *outliers*  

**üéª Violin Plots**  
- **Uso:** Similar a box plots pero mostrando densidad  
- **Ventajas:** Revelan m√∫ltiples modas  

**üìà Gr√°ficos de Densidad**  
- **Uso:** Versi√≥n suavizada del histograma  
- **Ventajas:** Mejor para comparar distribuciones  

---

### Para Relaciones entre Variables

**üîµ Scatter Plots (Diagramas de Dispersi√≥n)**  
- **Uso:** Relaciones entre dos variables num√©ricas  
- **Informaci√≥n:** Correlaci√≥n, *outliers*, patrones  

**üìä Gr√°ficos de L√≠neas**  
- **Uso:** Series temporales o tendencias  
- **Ventajas:** Cambios a lo largo del tiempo  

**üî• Mapas de Calor (Heatmaps)**  
- **Uso:** Matrices de correlaci√≥n, tablas de contingencia  
- **Ventajas:** Patrones en matrices de datos  

---

### Gr√°ficos Especializados

- **üìä Pair Plots:** Relaciones entre m√∫ltiples variables  
- **üìà Regresi√≥n:** Relaci√≥n con l√≠nea de tendencia  
- **üé® Multivariados:** M√°s de 2-3 variables simult√°neamente (color, tama√±o, forma, facetas)

- **üìä Escalas Secuenciales** ‚Üí Variables num√©ricas ordenadas

## üìÇ Datasets para Pr√°ctica Avanzada

````python
# Cargar m√∫ltiples datasets para diferentes tipos de an√°lisis

# Dataset 1: Iris (ya conocido)
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species_name'] = iris_df['species'].map({
    0: 'setosa', 1: 'versicolor', 2: 'virginica'
})

# Dataset 2: Wine (clasificaci√≥n de vinos)
wine = load_wine()
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['wine_class'] = wine.target
wine_df['wine_type'] = wine_df['wine_class'].map({
    0: 'Class_0', 1: 'Class_1', 2: 'Class_2'
})

# Dataset 3: Datos sint√©ticos de empleados para pr√°ctica
np.random.seed(42)
n_employees = 500

employees_df = pd.DataFrame({
    'employee_id': range(1, n_employees + 1),
    'department': np.random.choice(['IT', 'Sales', 'Marketing', 'HR', 'Finance'], n_employees),
    'experience_years': np.random.exponential(scale=3, size=n_employees),
    'salary': np.random.normal(50000, 15000, n_employees),
    'performance_score': np.random.normal(75, 12, n_employees),
    'age': np.random.normal(35, 8, n_employees),
    'education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], n_employees),
    'remote_work': np.random.choice(['Yes', 'No'], n_employees, p=[0.3, 0.7])
})

# Ajustes realistas
employees_df['salary'] = np.where(
    employees_df['department'] == 'IT', 
    employees_df['salary'] * 1.2, 
    employees_df['salary']
)
employees_df['salary'] = np.maximum(employees_df['salary'], 25000)  # Salario m√≠nimo
employees_df['age'] = np.clip(employees_df['age'], 22, 65)  # Rango de edad realista
employees_df['experience_years'] = np.clip(employees_df['experience_years'], 0, 40)

print("üìä Datasets cargados:")
print(f"‚Ä¢ Iris: {iris_df.shape} - Clasificaci√≥n de flores")
print(f"‚Ä¢ Wine: {wine_df.shape} - Clasificaci√≥n de vinos") 
print(f"‚Ä¢ Employees: {employees_df.shape} - Datos de empleados")
print("\nüîç Primeras filas del dataset de empleados:")
print(employees_df.head())
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# Helpers: crea dataframes si no existen en tu entorno
# ============================================================
if "iris_df" not in globals():
    from sklearn.datasets import load_iris
    iris = load_iris()
    iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
    iris_df["species"] = iris.target
    iris_df["species_name"] = iris_df["species"].map(dict(enumerate(iris.target_names)))

if "employees_df" not in globals():
    rng = np.random.default_rng(42)
    n = 300
    departments = np.array(["Ventas", "Marketing", "TI", "Finanzas", "RRHH"])
    employees_df = pd.DataFrame({
        "department": rng.choice(departments, size=n, replace=True),
        "salary": rng.normal(65000, 15000, size=n).clip(25000, 120000),
        "experience_years": rng.normal(7, 4, size=n).clip(0, 35),
        "performance_score": rng.integers(50, 100, size=n)
    })

print("\nüéØ GR√ÅFICOS ESPECIALIZADOS PARA INSIGHTS")
print("=" * 50)

# ============================================================
# 1) Ridge Plot (densidades apiladas) por departamento
# ============================================================
def create_ridge_plot(df_employees):
    """Crear un ridge plot para comparar distribuciones de salarios por departamento."""
    from scipy.stats import gaussian_kde

    depts = sorted(df_employees["department"].unique().tolist())
    fig, ax = plt.subplots(figsize=(12, 8))

    # Colormap consistente pero sin forzar colores espec√≠ficos si no quieres
    cmap = plt.cm.Set3(np.linspace(0, 1, len(depts)))

    y_step = 1.2
    for i, (dept, color) in enumerate(zip(depts, cmap)):
        dept_data = df_employees.loc[df_employees["department"] == dept, "salary"].dropna().values

        if len(dept_data) < 2:
            # KDE necesita al menos 2 puntos
            y_offset = i * y_step
            ax.text(dept_data.min() if len(dept_data) else 0, y_offset + 0.1,
                    f"{dept} (sin datos suficientes)", fontsize=9)
            continue

        kde = gaussian_kde(dept_data)
        x_min, x_max = np.percentile(dept_data, [1, 99])
        x = np.linspace(x_min, x_max, 200)
        d = kde(x)

        d_norm = d / d.max() * 0.8
        y_offset = i * y_step

        ax.fill_between(x, y_offset, y_offset + d_norm, alpha=0.7, color=color, label=dept)
        ax.plot(x, y_offset + d_norm, linewidth=1, color="black")
        ax.text(x_min - 0.05*(x_max-x_min), y_offset + 0.4, dept, fontweight="bold", va="center")

    ax.set_xlabel("Salario ($)")
    ax.set_title("Distribuci√≥n de Salarios por Departamento (Ridge Plot)", fontsize=14, fontweight="bold")
    ax.set_yticks([])
    for spine in ["left", "top", "right"]:
        ax.spines[spine].set_visible(False)
    plt.tight_layout()
    return fig

ridge_fig = create_ridge_plot(employees_df)
plt.show()

# ============================================================
# 2) Gr√°fico de Coordenadas Paralelas (Iris)
# ============================================================
print("\nüîó Creando Gr√°fico de Coordenadas Paralelas...")

from pandas.plotting import parallel_coordinates

# Garantizar columna species_name
if "species_name" not in iris_df.columns:
    if "species" in iris_df.columns:
        # si es num√©rica, mapear a nombres b√°sicos
        map_default = {0: "setosa", 1: "versicolor", 2: "virginica"}
        iris_df["species_name"] = iris_df["species"].map(map_default).fillna("species")
    else:
        iris_df["species_name"] = "species"

# Seleccionar muestra para mantener claridad visual
num_rows = min(60, len(iris_df))
iris_sample = iris_df.sample(n=num_rows, random_state=42)

cols_four = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
# Estandarizar columnas para que las escalas sean comparables en el gr√°fico
iris_std = iris_sample.copy()
for c in cols_four:
    iris_std[c] = (iris_std[c] - iris_std[c].mean()) / (iris_std[c].std(ddof=0) + 1e-9)

fig, ax = plt.subplots(figsize=(12, 6))
parallel_coordinates(iris_std, 'species_name', cols=cols_four, ax=ax, alpha=0.7)
ax.set_title('Coordenadas Paralelas - Caracter√≠sticas de Iris')
ax.set_ylabel('Valores Estandarizados')
ax.legend(title='Especie', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ============================================================
# 3) Gr√°fico de Radar/Spider (promedios por especie)
# ============================================================
def create_radar_chart(df_iris, feature_cols, label_col="species_name"):
    """Radar chart comparando perfiles (normalizados por columna)."""
    # Promedios por especie
    prof = df_iris.groupby(label_col)[feature_cols].mean()

    # Normalizar cada columna a [0,1] para comparabilidad
    prof_norm = (prof - prof.min()) / (prof.max() - prof.min() + 1e-9)

    categories = [c.replace("(cm)", "").title() for c in feature_cols]
    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    fig, axes = plt.subplots(1, len(prof_norm), figsize=(16, 6), subplot_kw=dict(projection='polar'))
    if len(prof_norm) == 1:
        axes = [axes]  # asegurar iterable

    fig.suptitle('Perfiles de Especies - Radar Chart', fontsize=16, fontweight='bold')

    for ax, (species, row) in zip(axes, prof_norm.iterrows()):
        vals = row.values.tolist()
        vals += vals[:1]  # cerrar c√≠rculo

        ax.plot(angles, vals, 'o-', linewidth=2)
        ax.fill(angles, vals, alpha=0.25)
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(categories)
        ax.set_yticklabels([])  # quitar etiquetas radiales para limpiar
        ax.set_title(f'Perfil: {species}', fontweight='bold')
        ax.grid(True)

    plt.tight_layout()
    return fig

radar_fig = create_radar_chart(iris_df, ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
plt.show()

# ============================================================
# 4) Gr√°fico de Burbujas Multidimensional (experiencia vs salario vs performance)
# ============================================================
print("\nüí≠ Creando Gr√°fico de Burbujas Multidimensional...")

fig, ax = plt.subplots(figsize=(12, 8))

dept_unique = sorted(employees_df['department'].unique().tolist())
cmap = plt.cm.Set1(np.linspace(0, 1, len(dept_unique)))
handles = []

for color, dept in zip(cmap, dept_unique):
    sub = employees_df[employees_df['department'] == dept]
    h = ax.scatter(
        sub['experience_years'],
        sub['salary'],
        s=sub['performance_score'] * 3,  # tama√±o ~ performance
        alpha=0.6,
        color=color,
        edgecolors='black',
        linewidth=0.5,
        label=dept
    )
    handles.append(h)

ax.set_xlabel('A√±os de Experiencia')
ax.set_ylabel('Salario ($)')
ax.set_title('Experiencia vs Salario vs Performance por Departamento\n(Tama√±o de burbuja = Performance Score)',
             fontweight='bold')

# Leyenda de categor√≠as (departamentos)
legend1 = ax.legend(handles=handles, title='Departamento', loc='upper left', frameon=True)
ax.add_artist(legend1)

# Leyenda para tama√±os (performance)
sizes_demo = [50, 70, 90]
labels_demo = [f'Score: {s}' for s in sizes_demo]
legend_bubbles = [
    plt.scatter([], [], s=s*3, alpha=0.6, color='gray', edgecolors='black', linewidth=0.5)
    for s in sizes_demo
]
legend2 = ax.legend(legend_bubbles, labels_demo, title='Performance Score', loc='upper right', frameon=True)

plt.tight_layout()
plt.show()
````

````python
print("\nüé® VISUALIZACIONES INTERACTIVAS CON PLOTLY")
print("=" * 50)

# Verificar si plotly est√° disponible
try:
    import plotly.express as px
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    plotly_available = True
    print("‚úÖ Plotly disponible - Creando visualizaciones interactivas")
except ImportError:
    plotly_available = False
    print("‚ö†Ô∏è Plotly no disponible - Saltando visualizaciones interactivas")

if plotly_available:
    
    # 1. Scatter plot interactivo
    print("\nüîµ Scatter Plot Interactivo...")
    
    fig_scatter = px.scatter(employees_df, 
                            x='experience_years', 
                            y='salary',
                            color='department',
                            size='performance_score',
                            hover_data=['age', 'education'],
                            title='An√°lisis Interactivo: Experiencia vs. Salario',
                            labels={
                                'experience_years': 'A√±os de Experiencia',
                                'salary': 'Salario ($)',
                                'department': 'Departamento'
                            })
    
    fig_scatter.update_layout(
        width=900,
        height=600,
        title_x=0.5,
        font=dict(size=12)
    )
    
    # Mostrar (en Jupyter se ver√≠a interactivo)
    fig_scatter.show()
    
    # 2. Dashboard interactivo con subplots
    print("\nüìä Dashboard Interactivo...")
    
    fig_dashboard = make_subplots(
        rows=2, cols=2,
        subplot_titles=('Distribuci√≥n de Salarios', 'Salarios por Departamento',
                       'Performance vs. Experiencia', 'Educaci√≥n vs. Salario'),
        specs=[[{"type": "histogram"}, {"type": "box"}],
               [{"type": "scatter"}, {"type": "violin"}]]
    )
    
    # Histograma
    fig_dashboard.add_trace(
        go.Histogram(x=employees_df['salary'], name='Salarios', nbinsx=25),
        row=1, col=1
    )
    
    # Box plot por departamento
    for dept in employees_df['department'].unique():
        dept_data = employees_df[employees_df['department'] == dept]
        fig_dashboard.add_trace(
            go.Box(y=dept_data['salary'], name=dept),
            row=1, col=2
        )
    
    # Scatter plot
    fig_dashboard.add_trace(
        go.Scatter(x=employees_df['experience_years'],
                  y=employees_df['performance_score'],
                  mode='markers',
                  marker=dict(color=employees_df['salary'],
                            colorscale='viridis',
                            size=8,
                            colorbar=dict(title='Salario')),
                  name='Empleados'),
        row=2, col=1
    )
    
    # Violin plot por educaci√≥n
    for edu in employees_df['education'].unique():
        edu_data = employees_df[employees_df['education'] == edu]
        fig_dashboard.add_trace(
            go.Violin(y=edu_data['salary'], name=edu, box_visible=True),
            row=2, col=2
        )
    
    fig_dashboard.update_layout(
        height=800,
        title_text="Dashboard Interactivo - An√°lisis de Empleados",
        title_x=0.5,
        showlegend=False
    )
    
    fig_dashboard.show()
    
    # 3. Gr√°fico de correlaciones interactivo
    print("\nüî• Heatmap de Correlaciones Interactivo...")
    
    numeric_cols = ['experience_years', 'salary', 'performance_score', 'age']
    corr_matrix = employees_df[numeric_cols].corr()
    
    fig_corr = px.imshow(corr_matrix,
                        text_auto=True,
                        aspect="auto",
                        title="Matriz de Correlaciones Interactiva",
                        color_continuous_scale='RdBu_r')
    
    fig_corr.update_layout(
        width=600,
        height=500,
        title_x=0.5
    )
    
    fig_corr.show()
    
    print("‚úÖ Visualizaciones interactivas creadas")

else:
    print("üí° Para visualizaciones interactivas, instala plotly: pip install plotly")
````

### üîß Herramientas y Templates Reutilizables

````python
print("\nüîß HERRAMIENTAS Y TEMPLATES REUTILIZABLES")
print("=" * 55)

class VisualizationToolkit:
    """
    Clase con herramientas avanzadas para visualizaci√≥n de datos
    """
    
    def __init__(self):
        self.setup_style()
        
    def setup_style(self):
        """Configurar estilo predeterminado"""
        self.colors = {
            'primary': '#2E86AB',
            'secondary': '#A23B72',
            'accent': '#F18F01',
            'success': '#C73E1D',
            'neutral': '#6C757D'
        }
        
        plt.rcParams.update({
            'axes.prop_cycle': plt.cycler('color', list(self.colors.values())),
            'font.size': 10,
            'axes.titlesize': 12,
            'axes.labelsize': 11,
            'figure.titlesize': 14
        })
    
    def correlation_heatmap(self, df, figsize=(10, 8), annot=True):
        """Crear heatmap de correlaciones optimizado"""
        
        # Seleccionar solo columnas num√©ricas
        numeric_df = df.select_dtypes(include=[np.number])
        
        if numeric_df.empty:
            print("‚ùå No se encontraron columnas num√©ricas")
            return None
        
        # Calcular correlaciones
        corr_matrix = numeric_df.corr()
        
        # Crear m√°scara para tri√°ngulo superior
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
        
        # Crear figura
        fig, ax = plt.subplots(figsize=figsize)
        
        # Heatmap
        sns.heatmap(corr_matrix, mask=mask, annot=annot, cmap='coolwarm',
                   center=0, square=True, fmt='.2f',
                   cbar_kws={'shrink': 0.8}, ax=ax)
        
        ax.set_title('Matriz de Correlaciones', fontweight='bold', pad=20)
        
        return fig
    
    def distribution_comparison(self, df, column, group_by, figsize=(15, 5)):
        """Comparar distribuciones entre grupos"""
        
        fig, axes = plt.subplots(1, 3, figsize=figsize)
        fig.suptitle(f'An√°lisis de Distribuci√≥n: {column} por {group_by}', 
                    fontweight='bold')
        
        # 1. Histogramas superpuestos
        groups = df[group_by].unique()
        for group in groups:
            group_data = df[df[group_by] == group][column]
            axes[0].hist(group_data, alpha=0.7, label=str(group), bins=20)
        
        axes[0].set_title('Histogramas Superpuestos')
        axes[0].set_xlabel(column)
        axes[0].set_ylabel('Frecuencia')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        # 2. Box plots
        df.boxplot(column=column, by=group_by, ax=axes[1])
        axes[1].set_title('Box Plots Comparativos')
        axes[1].set_xlabel(group_by)
        axes[1].set_ylabel(column)
        
        # 3. Violin plots
        sns.violinplot(data=df, x=group_by, y=column, ax=axes[2])
        axes[2].set_title('Violin Plots')
        axes[2].set_xlabel(group_by)
        axes[2].set_ylabel(column)
        axes[2].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        return fig
    
    def create_summary_stats_viz(self, df, columns=None):
        """Crear visualizaci√≥n de estad√≠sticas resumidas"""
        
        if columns is None:
            columns = df.select_dtypes(include=[np.number]).columns.tolist()
        
        if not columns:
            print("‚ùå No se encontraron columnas num√©ricas")
            return None
        
        # Calcular estad√≠sticas
        stats_df = df[columns].describe().T
        stats_df['cv'] = stats_df['std'] / stats_df['mean']  # Coeficiente de variaci√≥n
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Resumen Estad√≠stico de Variables Num√©ricas', fontweight='bold')
        
        # 1. Medidas de tendencia central
        stats_df[['mean', '50%']].plot(kind='bar', ax=axes[0, 0])
        axes[0, 0].set_title('Medidas de Tendencia Central')
        axes[0, 0].set_ylabel('Valor')
        axes[0, 0].legend(['Media', 'Mediana'])
        axes[0, 0].tick_params(axis='x', rotation=45)
        
        # 2. Medidas de dispersi√≥n
        stats_df[['std', 'cv']].plot(kind='bar', ax=axes[0, 1], secondary_y=True)
        axes[0, 1].set_title('Medidas de Dispersi√≥n')
        axes[0, 1].set_ylabel('Desviaci√≥n Est√°ndar')
        axes[0, 1].right_ax.set_ylabel('Coeficiente de Variaci√≥n')
        axes[0, 1].tick_params(axis='x', rotation=45)
        
        # 3. Rangos
        ranges = stats_df['max'] - stats_df['min']
        iqr = stats_df['75%'] - stats_df['25%']
        pd.DataFrame({'Rango Total': ranges, 'IQR': iqr}).plot(kind='bar', ax=axes[1, 0])
        axes[1, 0].set_title('An√°lisis de Rangos')
        axes[1, 0].set_ylabel('Valor')
        axes[1, 0].tick_params(axis='x', rotation=45)
        
        # 4. Tabla de estad√≠sticas
        axes[1, 1].axis('off')
        table_data = stats_df.round(2)
        table = axes[1, 1].table(cellText=table_data.values,
                                rowLabels=table_data.index,
                                colLabels=table_data.columns,
                                cellLoc='center',
                                loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(8)
        table.scale(1, 1.5)
        axes[1, 1].set_title('Tabla Resumen', fontweight='bold')
        
        plt.tight_layout()
        return fig

# Instanciar y usar el toolkit
viz_toolkit = VisualizationToolkit()

# Ejemplo 1: Heatmap de correlaciones
print("\nüî• Creando Heatmap de Correlaciones...")
corr_fig = viz_toolkit.correlation_heatmap(employees_df)
plt.show()

# Ejemplo 2: Comparaci√≥n de distribuciones
print("\nüìä Comparando Distribuciones...")
dist_fig = viz_toolkit.distribution_comparison(employees_df, 'salary', 'department')
plt.show()

# Ejemplo 3: Resumen estad√≠stico visual
print("\nüìà Resumen Estad√≠stico Visual...")
summary_fig = viz_toolkit.create_summary_stats_viz(employees_df)
plt.show()

print("\n‚úÖ Toolkit de visualizaci√≥n implementado exitosamente")
````
