## üìñ **PARTE 1: ¬øQu√© es la Estad√≠stica Descriptiva?**

### **Definici√≥n**

> **La estad√≠stica descriptiva** es la rama de la estad√≠stica que se encarga de **recopilar, organizar, resumir y presentar datos** de manera que podamos entender las caracter√≠sticas principales de un conjunto de informaci√≥n.

### **üîç Diferencias Clave**

| Estad√≠stica Descriptiva | Estad√≠stica Inferencial |
|------------------------|------------------------|
| **Describe** los datos que tenemos | **Infiere** conclusiones sobre poblaciones |
| **Resume** la informaci√≥n disponible | **Predice** y generaliza |
| **No hace predicciones** | **Hace estimaciones y prueba hip√≥tesis** |
| Ejemplo: "El 60% de nuestros clientes son mujeres" | Ejemplo: "Con 95% de confianza, entre 55% y 65% de todos los clientes son mujeres" |

### **üåç Aplicaciones en la Vida Real**

- An√°lisis de ventas mensuales
- Perfil demogr√°fico de clientes
- Rendimiento de campa√±as publicitarias

#### **2. Salud P√∫blica**
- Estad√≠sticas de enfermedades por regi√≥n
- Efectividad de tratamientos m√©dicos
- An√°lisis de factores de riesgo

#### **3. Educaci√≥n**
- Rendimiento acad√©mico por grado

- Evaluaci√≥n de m√©todos de ense√±anza

#### **4. Deportes**
- Estad√≠sticas de jugadores
- An√°lisis de rendimiento de equipos
- Comparaci√≥n de temporadas

#### **5. Gobierno y Pol√≠tica**
- Censos poblacionales
- √çndices econ√≥micos
- An√°lisis de resultados electorales

### **üí° ¬øPor qu√© es Importante?**

1. **Toma de Decisiones Informadas**: Los datos nos ayudan a decidir bas√°ndonos en hechos, no en intuiciones
2. **Comunicaci√≥n Efectiva**: Los n√∫meros bien presentados comunican mejor que miles de palabras

## üìä **PARTE 2: Tipos de Variables**

### **üî¢ Clasificaci√≥n Principal**

```
VARIABLES
‚îú‚îÄ‚îÄ CUALITATIVAS (Categ√≥ricas)
‚îÇ   ‚îú‚îÄ‚îÄ Nominales
‚îÇ   ‚îî‚îÄ‚îÄ Ordinales
‚îî‚îÄ‚îÄ CUANTITATIVAS (Num√©ricas)
    ‚îú‚îÄ‚îÄ Discretas
    ‚îî‚îÄ‚îÄ Continuas
```

### **Variables Cualitativas (Categ√≥ricas)**

#### **Variables Nominales**
- **Definici√≥n**: Categor√≠as sin orden natural
- **Caracter√≠sticas**: Solo se pueden contar, no tienen jerarqu√≠a
- **Ejemplos**:
  - G√©nero: Masculino, Femenino, Otro
  - Color de ojos: Azul, Verde, Caf√©, Negro
  - Estado civil: Soltero, Casado, Divorciado, Viudo
  - Marca de autom√≥vil: Toyota, Ford, BMW, etc.

#### **Variables Ordinales**
- **Definici√≥n**: Categor√≠as con orden natural o jerarqu√≠a
- **Caracter√≠sticas**: Se pueden ordenar de menor a mayor (o viceversa)
- **Ejemplos**:
  - Nivel educativo: Primaria < Secundaria < Universidad < Posgrado
  - Satisfacci√≥n: Muy Insatisfecho < Insatisfecho < Neutral < Satisfecho < Muy Satisfecho
  - Tama√±o: Peque√±o < Mediano < Grande
  - Calificaciones: A > B > C > D > F

### **Variables Cuantitativas (Num√©ricas)**

#### **Variables Discretas**
- **Definici√≥n**: Solo pueden tomar valores enteros espec√≠ficos
- **Caracter√≠sticas**: Se pueden contar, hay "saltos" entre valores
- **Ejemplos**:
  - N√∫mero de hijos: 0, 1, 2, 3, 4...
  - Cantidad de productos vendidos: 10, 25, 100...
  - N√∫mero de empleados: 5, 50, 500...
  - Calificaci√≥n en examen: 0-100 (enteros)

#### **Variables Continuas**
- **Definici√≥n**: Pueden tomar cualquier valor dentro de un rango
- **Caracter√≠sticas**: Se pueden medir, no hay "saltos"
- **Ejemplos**:
  - Altura: 1.75m, 1.758m, 1.7581m...
  - Peso: 65.3kg, 65.35kg, 65.351kg...
  - Temperatura: 23.4¬∞C, 23.45¬∞C...
  - Tiempo: 5.25 segundos, 5.251 segundos...

### **üìè Escalas de Medici√≥n**

#### **Escala Nominal**
- Solo categor√≠as diferentes
- Operaciones: Contar frecuencias
- Ejemplo: Colores, marcas

#### **Escala Ordinal**
- Categor√≠as ordenadas
- Operaciones: Contar, ordenar
- Ejemplo: Ranking, calificaciones con letras

#### **Escala de Intervalo**
- N√∫meros con intervalos iguales, pero sin cero absoluto
- Operaciones: Sumar, restar, promedios
- Ejemplo: Temperatura en Celsius (0¬∞C no significa "ausencia de temperatura")

#### **Escala de Raz√≥n**
- N√∫meros con intervalos iguales y cero absoluto
- Operaciones: Todas las operaciones matem√°ticas
- Ejemplo: Altura, peso, ingresos (0 significa "ausencia total")

### **üìä Poblaci√≥n vs. Muestra**

## Paso 1: Configuraci√≥n del Entorno y Primeros Pasos con Datos

````python
# Librer√≠as fundamentales para estad√≠stica descriptiva
import pandas as pd              # Para manipulaci√≥n de datos
import numpy as np               # Para c√°lculos matem√°ticos y estad√≠sticos
import matplotlib.pyplot as plt  # Para gr√°ficos b√°sicos
import seaborn as sns           # Para gr√°ficos estad√≠sticos avanzados
````

## PASO 2: CARGA DE NUESTRO PRIMER DATASET - IRIS

````python
# Cargamos el famoso dataset Iris
from sklearn.datasets import load_iris
iris_data = load_iris()

# Convertimos a DataFrame de pandas para facilitar el an√°lisis
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print("‚úÖ Dataset Iris cargado exitosamente")
print(f"üìè Dimensiones: {iris_df.shape[0]} filas √ó {iris_df.shape[1]} columnas")
````

## PASO 3: EXPLORACI√ìN INICIAL DE LOS DATOS

````python
print("\n" + "="*70)
print("üîç EXPLORACI√ìN INICIAL - CONOCIENDO NUESTROS DATOS")
print("="*70)

print("\n1Ô∏è‚É£ PRIMERAS 5 FILAS (m√©todo .head()):")
print(iris_df.head())

print("\n2Ô∏è‚É£ INFORMACI√ìN GENERAL DEL DATASET (m√©todo .info()):")
iris_df.info()

print("\n3Ô∏è‚É£ ESTAD√çSTICAS DESCRIPTIVAS B√ÅSICAS (m√©todo .describe()):")
print(iris_df.describe())

print("\n4Ô∏è‚É£ NOMBRES DE TODAS LAS COLUMNAS:")
print("Columnas:", list(iris_df.columns))

print("\n5Ô∏è‚É£ TIPOS DE DATOS POR COLUMNA:")
print(iris_df.dtypes)
````

## PASO 4: IDENTIFICACI√ìN DE TIPOS DE VARIABLES

````python
print("\n" + "="*70)
print("üè∑Ô∏è IDENTIFICACI√ìN DE TIPOS DE VARIABLES")
print("="*70)

# An√°lisis manual de cada variable
variables_info = {
    'sepal length (cm)': {
        'tipo': 'Cuantitativa Continua',
        'razon': 'Se puede medir con decimales, representa longitud',
        'escala': 'Raz√≥n (tiene cero absoluto)'
    },
    'sepal width (cm)': {
        'tipo': 'Cuantitativa Continua',
        'razon': 'Se puede medir con decimales, representa ancho',
        'escala': 'Raz√≥n (tiene cero absoluto)'
    },
    'petal length (cm)': {
        'tipo': 'Cuantitativa Continua',
        'razon': 'Se puede medir con decimales, representa longitud',
        'escala': 'Raz√≥n (tiene cero absoluto)'
    },
    'petal width (cm)': {
        'tipo': 'Cuantitativa Continua',
        'razon': 'Se puede medir con decimales, representa ancho',
        'escala': 'Raz√≥n (tiene cero absoluto)'
    },
    'species': {
        'tipo': 'Cualitativa Nominal',
        'razon': 'Categor√≠as de especies sin orden natural',
        'escala': 'Nominal (solo categor√≠as diferentes)'
    }
}

for variable, info in variables_info.items():
    print(f"\nüè∑Ô∏è {variable}:")
    print(f"   Tipo: {info['tipo']}")
    print(f"   Raz√≥n: {info['razon']}")
    print(f"   Escala: {info['escala']}")
````

## PASO 5: AN√ÅLISIS ESPEC√çFICO POR TIPO DE VARIABLE

````python
print("\n" + "="*70)
print("üìà AN√ÅLISIS ESPEC√çFICO POR TIPO DE VARIABLE")
print("="*70)

# AN√ÅLISIS DE VARIABLES CUANTITATIVAS
print("\nüî¢ VARIABLES CUANTITATIVAS - Medidas b√°sicas:")
numeric_vars = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

for var in numeric_vars:
    print(f"\nüìä {var}:")
    print(f"   Promedio: {iris_df[var].mean():.2f}")
    print(f"   Mediana: {iris_df[var].median():.2f}")
    print(f"   M√≠nimo: {iris_df[var].min():.2f}")
    print(f"   M√°ximo: {iris_df[var].max():.2f}")
    print(f"   Desviaci√≥n est√°ndar: {iris_df[var].std():.2f}")

# AN√ÅLISIS DE VARIABLES CUALITATIVAS
print("\nüè∑Ô∏è VARIABLE CUALITATIVA - Frecuencias:")
print("\nüìä Distribuci√≥n de especies:")
especies_frecuencia = iris_df['species'].value_counts()
print(especies_frecuencia)

print("\nüìä Porcentajes por especie:")
especies_porcentaje = iris_df['species'].value_counts(normalize=True) * 100
for especie, porcentaje in especies_porcentaje.items():
    print(f"   {especie}: {porcentaje:.1f}%")
````

## PASO 6: PRIMERAS VISUALIZACIONES

````python
print("\n" + "="*70)
print("üìä CREANDO NUESTRAS PRIMERAS VISUALIZACIONES")
print("="*70)

# Configuramos el estilo de los gr√°ficos
plt.style.use('default')
sns.set_palette("husl")

# GR√ÅFICO 1: Histograma de una variable cuantitativa
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(iris_df['sepal length (cm)'], bins=15, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribuci√≥n de Longitud del S√©palo')
plt.xlabel('Longitud (cm)')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.3)

# GR√ÅFICO 2: Gr√°fico de barras para variable categ√≥rica
plt.subplot(1, 3, 2)
especies_count = iris_df['species'].value_counts()
plt.bar(especies_count.index, especies_count.values, color=['coral', 'lightgreen', 'gold'])
plt.title('Cantidad por Especie')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# GR√ÅFICO 3: Gr√°fico de dispersi√≥n (dos variables cuantitativas)
plt.subplot(1, 3, 3)
for especie in iris_df['species'].unique():
    datos_especie = iris_df[iris_df['species'] == especie]
    plt.scatter(datos_especie['sepal length (cm)'], datos_especie['sepal width (cm)'], 
               label=especie, alpha=0.7, s=50)

plt.title('Longitud vs Ancho del S√©palo')
plt.xlabel('Longitud (cm)')
plt.ylabel('Ancho (cm)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("‚úÖ Gr√°ficos creados exitosamente")
````
