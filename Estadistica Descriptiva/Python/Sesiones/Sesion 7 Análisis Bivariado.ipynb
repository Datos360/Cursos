## üìñ FUNDAMENTOS TE√ìRICOS

### ¬øQu√© es el An√°lisis Bivariado?

El an√°lisis bivariado es el estudio estad√≠stico de **dos variables simult√°neamente** para determinar si existe una relaci√≥n entre ellas. Mientras que el an√°lisis univariado se centra en describir cada variable por separado, el an√°lisis bivariado nos permite:

- **Identificar patrones de relaci√≥n** entre variables
- **Cuantificar la fuerza** de estas relaciones  
- **Determinar la direcci√≥n** (positiva o negativa) de la relaci√≥n
- **Evaluar la independencia** estad√≠stica entre variables

### Tipos de Relaciones Bivariadas

Dependiendo del tipo de variables que analicemos, tenemos diferentes enfoques:

| Variable 1 | Variable 2 | An√°lisis Apropiado |
|------------|------------|-------------------|
| Cuantitativa | Cuantitativa | **Correlaci√≥n** (Pearson, Spearman) |
| Categ√≥rica | Categ√≥rica | **Tablas de contingencia**, Chi-cuadrado |
| Categ√≥rica | Cuantitativa | **Comparaci√≥n de grupos**, ANOVA |

## 1. AN√ÅLISIS DE CORRELACI√ìN

### 1.1 Conceptos Fundamentales

La **correlaci√≥n** mide la fuerza y direcci√≥n de la relaci√≥n lineal entre dos variables cuantitativas.

#### Caracter√≠sticas importantes:
- **Rango**: Los coeficientes de correlaci√≥n van de -1 a +1
- **Interpretaci√≥n**:
  - **r = +1**: Correlaci√≥n positiva perfecta
  - **r = 0**: No hay correlaci√≥n lineal
  - **r = -1**: Correlaci√≥n negativa perfecta

#### Interpretaci√≥n pr√°ctica:
- **|r| > 0.7**: Correlaci√≥n fuerte
- **0.3 < |r| ‚â§ 0.7**: Correlaci√≥n moderada  
- **|r| ‚â§ 0.3**: Correlaci√≥n d√©bil

### 1.2 Correlaci√≥n de Pearson

**Cu√°ndo usarla:**
- Ambas variables son cuantitativas
- La relaci√≥n es aproximadamente lineal
- Los datos siguen distribuci√≥n normal (aproximadamente)

**F√≥rmula conceptual:**
```
r = Œ£[(xi - xÃÑ)(yi - »≥)] / ‚àö[Œ£(xi - xÃÑ)¬≤ √ó Œ£(yi - »≥)¬≤]
```

**Ejemplo conceptual:**
Imagina que queremos estudiar la relaci√≥n entre:
- **A√±os de educaci√≥n** (Variable X)
- **Ingresos anuales** (Variable Y)

Si r = 0.65, interpretar√≠amos que existe una correlaci√≥n positiva moderada: a mayor educaci√≥n, tiende a haber mayores ingresos.

### 1.3 Correlaci√≥n de Spearman

**Cu√°ndo usarla:**
- Variables ordinales o cuantitativas que no siguen distribuci√≥n normal
- La relaci√≥n es mon√≥tona pero no necesariamente lineal
- Presencia de valores at√≠picos

**Ventaja principal:** Es m√°s robusta ante outliers porque trabaja con rangos.

### 1.4 ‚ö†Ô∏è Correlaci√≥n vs. Causalidad

**IMPORTANTE:** La correlaci√≥n NO implica causalidad.

- **Correlaci√≥n alta** significa que las variables var√≠an juntas
- **Causalidad** significa que una variable causa cambios en otra
- Pueden existir **variables confusoras** que expliquen ambas relaciones

**Ejemplo:** 
- Correlaci√≥n: Ventas de helados y ahogamientos (r = 0.8)
- Variable confusora: Temperatura (a mayor temperatura, m√°s ventas de helados Y m√°s gente en piscinas)

## 2. TABLAS DE CONTINGENCIA

### 2.1 Concepto y Construcci√≥n

Una **tabla de contingencia** (o tabla cruzada) muestra la distribuci√≥n de frecuencias de dos variables categ√≥ricas simult√°neamente.

#### Estructura b√°sica:
```
                      Variable Y
                  Cat1  Cat2  Cat3  Total
Variable X  Cat1   n11   n12   n13   n1+
            Cat2   n21   n22   n23   n2+
            Total  n+1   n+2   n+3    n
```

Donde:
- **nij**: frecuencia de la combinaci√≥n categor√≠a i de X con categor√≠a j de Y
- **ni+**: total marginal de la fila i
- **n+j**: total marginal de la columna j

### 2.2 Tipos de Frecuencias

1. **Frecuencias absolutas**: Conteos directos
2. **Frecuencias relativas**: Proporciones del total general
3. **Frecuencias condicionales**:
   - Por filas: proporci√≥n dentro de cada fila
   - Por columnas: proporci√≥n dentro de cada columna

### 2.3 An√°lisis de Independencia

Dos variables categ√≥ricas son **independientes** si el conocimiento de una no nos proporciona informaci√≥n sobre la otra.

**Hip√≥tesis:**
- **H‚ÇÄ**: Las variables son independientes
- **H‚ÇÅ**: Las variables est√°n asociadas

---

## 3. MEDIDAS DE ASOCIACI√ìN

### 3.1 Chi-cuadrado de Pearson

**Prop√≥sito:** Evaluar si existe asociaci√≥n entre dos variables categ√≥ricas.

**F√≥rmula:**
```
œá¬≤ = Œ£[(Observado - Esperado)¬≤ / Esperado]
```

**Interpretaci√≥n:**
- **œá¬≤ = 0**: Independencia perfecta
- **œá¬≤ > 0**: Indica grado de asociaci√≥n (mayor valor = mayor asociaci√≥n)

### 3.2 Proceso de la Prueba Chi-cuadrado

1. **Plantear hip√≥tesis:**
   - H‚ÇÄ: Las variables son independientes
   - H‚ÇÅ: Las variables est√°n asociadas

2. **Calcular frecuencias esperadas:**
   - E_ij = (Total_fila_i √ó Total_columna_j) / Total_general

3. **Calcular estad√≠stico œá¬≤**

4. **Comparar con valor cr√≠tico o p-valor**

5. **Tomar decisi√≥n estad√≠stica**

### 3.3 Supuestos del Chi-cuadrado

- **Observaciones independientes**
- **Frecuencias esperadas ‚â• 5** en al menos 80% de las celdas
- **Ninguna frecuencia esperada < 1**

### 3.4 Limitaciones del Chi-cuadrado

- Es sensible al tama√±o de la muestra
- No indica la fuerza de la asociaci√≥n directamente
- Solo detecta asociaci√≥n, no la naturaleza de la relaci√≥n

## IMPORTACI√ìN DE LIBRER√çAS

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, chi2_contingency
from scipy.stats import chi2
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaciones
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 11
````

## CREACI√ìN DEL DATASET SOCIOECON√ìMICO

````python
# Configurar semilla para reproducibilidad
np.random.seed(42)
n_samples = 500

print("üèóÔ∏è Generando dataset socioecon√≥mico...")

# Generar a√±os de educaci√≥n (distribuci√≥n normal)
anos_educacion = np.random.normal(13, 3, n_samples)
anos_educacion = np.clip(anos_educacion, 8, 20)  # Entre 8 y 20 a√±os

# Generar ingresos correlacionados con educaci√≥n
ingresos_base = 25000 + 3500 * anos_educacion + np.random.normal(0, 8000, n_samples)
ingresos_anuales = np.clip(ingresos_base, 20000, 120000)

# Generar edad
edad = np.random.normal(35, 10, n_samples)
edad = np.clip(edad, 22, 65)

# Generar experiencia laboral (basada en edad y educaci√≥n)
experiencia = edad - anos_educacion - 18 + np.random.normal(0, 2, n_samples)
experiencia = np.clip(experiencia, 0, 40)

# Generar ocupaciones (variable categ√≥rica)
ocupaciones = ['T√©cnico', 'Profesional', 'Gerente', 'Operario', 'Administrativo']
prob_ocupacion = [0.25, 0.30, 0.15, 0.20, 0.10]
ocupacion = np.random.choice(ocupaciones, n_samples, p=prob_ocupacion)

# Crear nivel educativo categ√≥rico
def categorizar_educacion(anos):
    if anos < 12:
        return 'B√°sica'
    elif anos < 16:
        return 'Media/T√©cnica'
    else:
        return 'Superior'

nivel_educativo = [categorizar_educacion(anos) for anos in anos_educacion]

# Generar g√©nero con ligera brecha salarial
genero = np.random.choice(['Masculino', 'Femenino'], n_samples, p=[0.52, 0.48])
ajuste_genero = np.where(genero == 'Femenino', 0.92, 1.0)
ingresos_anuales = ingresos_anuales * ajuste_genero

# Crear DataFrame
df_socioeconomico = pd.DataFrame({
    'edad': edad.round(0).astype(int),
    'anos_educacion': anos_educacion.round(1),
    'experiencia_laboral': experiencia.round(1),
    'ingresos_anuales': ingresos_anuales.round(0).astype(int),
    'ocupacion': ocupacion,
    'nivel_educativo': nivel_educativo,
    'genero': genero
})

print("‚úÖ Dataset creado exitosamente")
print("üìä Informaci√≥n del dataset:")
print("="*50)
print(f"Tama√±o: {df_socioeconomico.shape}")
print(f"Variables: {list(df_socioeconomico.columns)}")
print("\nüîç Primeras 5 observaciones:")
print(df_socioeconomico.head())

print("\nüìà Estad√≠sticas descriptivas:")
print(df_socioeconomico.describe())
````

````python
print("\n" + "="*70)
print("üìä 1. AN√ÅLISIS DE CORRELACI√ìN")
print("="*70)

# Seleccionar variables num√©ricas
variables_numericas = ['edad', 'anos_educacion', 'experiencia_laboral', 'ingresos_anuales']
df_numerico = df_socioeconomico[variables_numericas]

# Calcular matriz de correlaci√≥n de Pearson
matriz_correlacion = df_numerico.corr()

print("üìä MATRIZ DE CORRELACI√ìN (Pearson):")
print(matriz_correlacion.round(3))

# Visualizar matriz de correlaci√≥n
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Heatmap b√°sico
sns.heatmap(matriz_correlacion, 
            annot=True, 
            cmap='RdBu_r', 
            center=0,
            square=True,
            fmt='.3f',
            ax=ax[0])
ax[0].set_title('Matriz de Correlaci√≥n - Completa', fontsize=14, fontweight='bold')

# Heatmap triangular
mask = np.triu(np.ones_like(matriz_correlacion, dtype=bool))
sns.heatmap(matriz_correlacion, 
            annot=True, 
            mask=mask,
            cmap='RdBu_r', 
            center=0,
            square=True,
            fmt='.3f',
            ax=ax[1])
ax[1].set_title('Matriz de Correlaci√≥n - Triangular', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
````

1.1 FUNCI√ìN PARA AN√ÅLISIS DETALLADO DE CORRELACIONES

````python
def analizar_correlacion(df, var1, var2, metodo='pearson'):
    """
    Funci√≥n para analizar la correlaci√≥n entre dos variables
    """
    # Calcular correlaci√≥n seg√∫n el m√©todo
    if metodo == 'pearson':
        coef, p_valor = pearsonr(df[var1], df[var2])
        tipo_correlacion = "Pearson"
    elif metodo == 'spearman':
        coef, p_valor = spearmanr(df[var1], df[var2])
        tipo_correlacion = "Spearman"
    
    # Interpretaci√≥n de la fuerza
    if abs(coef) > 0.7:
        fuerza = "FUERTE"
    elif abs(coef) > 0.3:
        fuerza = "MODERADA"
    else:
        fuerza = "D√âBIL"
    
    # Interpretaci√≥n de la direcci√≥n
    if coef > 0:
        direccion = "POSITIVA"
        explicacion = f"A mayor {var1}, mayor {var2}"
    else:
        direccion = "NEGATIVA"
        explicacion = f"A mayor {var1}, menor {var2}"
    
    # Significancia estad√≠stica
    significativa = "S√ç" if p_valor < 0.05 else "NO"
    
    print(f"\nüîç AN√ÅLISIS: {var1.upper()} vs {var2.upper()}")
    print("="*60)
    print(f"M√©todo: {tipo_correlacion}")
    print(f"Coeficiente: {coef:.4f}")
    print(f"P-valor: {p_valor:.4f}")
    print(f"Fuerza: {fuerza}")
    print(f"Direcci√≥n: {direccion}")
    print(f"Significativa (Œ±=0.05): {significativa}")
    print(f"Interpretaci√≥n: {explicacion}")
    
    return coef, p_valor

# Analizar correlaciones espec√≠ficas m√°s interesantes
correlaciones_interes = [
    ('anos_educacion', 'ingresos_anuales'),
    ('experiencia_laboral', 'ingresos_anuales'), 
    ('edad', 'experiencia_laboral'),
    ('edad', 'ingresos_anuales')
]

for var1, var2 in correlaciones_interes:
    coef_pearson, _ = analizar_correlacion(df_socioeconomico, var1, var2, 'pearson')
    coef_spearman, _ = analizar_correlacion(df_socioeconomico, var1, var2, 'spearman')
    print(f"Diferencia Pearson-Spearman: {abs(coef_pearson - coef_spearman):.4f}")
    print("-"*60)
````

## 2. AN√ÅLISIS DE TABLAS DE CONTINGENCIA

````python
print("\n" + "="*70)
print("üìä 2. AN√ÅLISIS DE TABLAS DE CONTINGENCIA")
print("="*70)

def crear_tabla_contingencia(df, var1, var2, mostrar_porcentajes=True):
    """
    Funci√≥n completa para crear y analizar tablas de contingencia
    """
    print(f"\nüìä TABLA DE CONTINGENCIA: {var1.upper()} vs {var2.upper()}")
    print("="*70)
    
    # Tabla de frecuencias absolutas
    tabla_abs = pd.crosstab(df[var1], df[var2], margins=True)
    print("üî¢ FRECUENCIAS ABSOLUTAS:")
    print(tabla_abs)
    
    if mostrar_porcentajes:
        print("\nüìà PORCENTAJES POR FILA:")
        tabla_fila = pd.crosstab(df[var1], df[var2], normalize='index') * 100
        print(tabla_fila.round(2))
        
        print("\nüìâ PORCENTAJES POR COLUMNA:")
        tabla_col = pd.crosstab(df[var1], df[var2], normalize='columns') * 100
        print(tabla_col.round(2))
        
        print("\nüéØ PORCENTAJES DEL TOTAL:")
        tabla_total = pd.crosstab(df[var1], df[var2], normalize='all') * 100
        print(tabla_total.round(2))
    
    return tabla_abs

# Analizar relaci√≥n entre Nivel Educativo y Ocupaci√≥n
tabla1 = crear_tabla_contingencia(df_socioeconomico, 'nivel_educativo', 'ocupacion')
````

2.1 PRUEBA DE CHI-CUADRADO

````python
def test_chi_cuadrado(df, var1, var2, alpha=0.05):
    """
    Realizar prueba de chi-cuadrado para independencia
    """
    print(f"\nüß™ PRUEBA DE CHI-CUADRADO: {var1.upper()} vs {var2.upper()}")
    print("="*70)
    
    # Crear tabla de contingencia (sin totales marginales)
    tabla_obs = pd.crosstab(df[var1], df[var2])
    
    # Realizar prueba de chi-cuadrado
    chi2_stat, p_valor, grados_libertad, tabla_esperada = chi2_contingency(tabla_obs)
    
    # Valor cr√≠tico
    valor_critico = chi2.ppf(1 - alpha, grados_libertad)
    
    # Resultados
    print(f"Estad√≠stico Chi-cuadrado: {chi2_stat:.4f}")
    print(f"P-valor: {p_valor:.4f}")
    print(f"Grados de libertad: {grados_libertad}")
    print(f"Valor cr√≠tico (Œ±={alpha}): {valor_critico:.4f}")
    
    # Interpretaci√≥n
    if p_valor < alpha:
        decision = "RECHAZAR H‚ÇÄ"
        conclusion = "Las variables S√ç est√°n asociadas (dependientes)"
    else:
        decision = "NO RECHAZAR H‚ÇÄ"
        conclusion = "Las variables NO est√°n asociadas (independientes)"
    
    print(f"\nüìã DECISI√ìN: {decision}")
    print(f"üéØ CONCLUSI√ìN: {conclusion}")
    
    # Mostrar tabla de frecuencias esperadas
    print(f"\nüìä FRECUENCIAS ESPERADAS:")
    tabla_esperada_df = pd.DataFrame(tabla_esperada, 
                                    index=tabla_obs.index, 
                                    columns=tabla_obs.columns)
    print(tabla_esperada_df.round(2))
    
    # Calcular residuos estandarizados
    residuos = (tabla_obs - tabla_esperada) / np.sqrt(tabla_esperada)
    print(f"\nüìê RESIDUOS ESTANDARIZADOS:")
    print(residuos.round(2))
    
    return chi2_stat, p_valor, tabla_obs, tabla_esperada

# Probar independencia entre variables categ√≥ricas
chi2_stat, p_val, tabla_obs, tabla_esp = test_chi_cuadrado(df_socioeconomico, 'nivel_educativo', 'ocupacion')

# Analizar relaci√≥n entre G√©nero y Ocupaci√≥n
print("\n" + "="*80)
tabla2 = crear_tabla_contingencia(df_socioeconomico, 'genero', 'ocupacion')
chi2_stat2, p_val2, tabla_obs2, tabla_esp2 = test_chi_cuadrado(df_socioeconomico, 'genero', 'ocupacion')
````

## 3. AN√ÅLISIS MIXTO: CATEG√ìRICA vs CUANTITATIVA

````python
print("\n" + "="*70)
print("üìä 3. AN√ÅLISIS MIXTO: CATEG√ìRICA vs CUANTITATIVA")
print("="*70)

def analizar_categorica_vs_numerica(df, var_cat, var_num):
    """
    Analizar relaci√≥n entre variable categ√≥rica y num√©rica
    """
    print(f"\nüìä AN√ÅLISIS: {var_cat.upper()} vs {var_num.upper()}")
    print("="*60)
    
    # Estad√≠sticas descriptivas por grupo
    stats_por_grupo = df.groupby(var_cat)[var_num].agg([
        'count', 'mean', 'median', 'std', 'min', 'max'
    ]).round(2)
    
    print("üìà ESTAD√çSTICAS POR GRUPO:")
    print(stats_por_grupo)
    
    # Calcular diferencias entre grupos
    medias = df.groupby(var_cat)[var_num].mean()
    print(f"\nüìä MEDIAS POR GRUPO:")
    for grupo, media in medias.items():
        print(f"  {grupo}: ${media:,.0f}")
    
    # Diferencia m√°xima entre grupos
    diff_max = medias.max() - medias.min()
    print(f"\nüí∞ DIFERENCIA M√ÅXIMA: ${diff_max:,.0f}")
    
    return stats_por_grupo

# Analizar ingresos por g√©nero
stats_genero = analizar_categorica_vs_numerica(df_socioeconomico, 'genero', 'ingresos_anuales')

# Analizar ingresos por ocupaci√≥n
stats_ocupacion = analizar_categorica_vs_numerica(df_socioeconomico, 'ocupacion', 'ingresos_anuales')

# Analizar ingresos por nivel educativo
stats_educacion = analizar_categorica_vs_numerica(df_socioeconomico, 'nivel_educativo', 'ingresos_anuales')
````

## 4. AN√ÅLISIS AVANZADO: CORRELACI√ìN POR GRUPOS

````python
print("\n" + "="*70)
print("üìä 4. AN√ÅLISIS AVANZADO: CORRELACI√ìN POR GRUPOS")
print("="*70)

def correlacion_por_grupos(df, var1, var2, grupo):
    """
    Calcular correlaci√≥n entre dos variables para cada grupo
    """
    print(f"\nüîç CORRELACI√ìN {var1.upper()} vs {var2.upper()} POR {grupo.upper()}:")
    print("="*60)
    
    correlaciones = {}
    for categoria in df[grupo].unique():
        df_subset = df[df[grupo] == categoria]
        corr, p_val = pearsonr(df_subset[var1], df_subset[var2])
        correlaciones[categoria] = {'correlacion': corr, 'p_valor': p_val}
        
        significativa = "S√ç" if p_val < 0.05 else "NO"
        print(f"{categoria}: r = {corr:.4f}, p = {p_val:.4f}, Significativa: {significativa}")
    
    return correlaciones

# Analizar correlaci√≥n educaci√≥n-ingresos por g√©nero
corr_edu_ing_genero = correlacion_por_grupos(df_socioeconomico, 
                                            'anos_educacion', 
                                            'ingresos_anuales', 
                                            'genero')

# Visualizar correlaciones por grupos
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Scatter plot por g√©nero
for i, genero in enumerate(['Masculino', 'Femenino']):
    df_genero = df_socioeconomico[df_socioeconomico['genero'] == genero]
    
    sns.scatterplot(data=df_genero, x='anos_educacion', y='ingresos_anuales', 
                   ax=axes[i], alpha=0.6)
    sns.regplot(data=df_genero, x='anos_educacion', y='ingresos_anuales', 
               ax=axes[i], scatter=False, color='red')
    
    corr = corr_edu_ing_genero[genero]['correlacion']
    p_val = corr_edu_ing_genero[genero]['p_valor']
    
    axes[i].set_title(f'Educaci√≥n vs Ingresos - {genero}\nr = {corr:.3f}, p = {p_val:.4f}', 
                      fontweight='bold')
    axes[i].set_xlabel('A√±os de Educaci√≥n', fontweight='bold')
    axes[i].set_ylabel('Ingresos Anuales ($)', fontweight='bold')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
