## üìñ FUNDAMENTOS TE√ìRICOS

### ¬øQu√© son los Valores At√≠picos (Outliers)?

Los **valores at√≠picos** o **outliers** son observaciones que se desv√≠an significativamente del patr√≥n general de los datos. Son puntos de datos que est√°n "muy alejados" del resto de las observaciones.

#### **Caracter√≠sticas de los Outliers:**
- üîç **Desviaci√≥n extrema**: Se alejan considerablemente del centro de la distribuci√≥n
- üìä **Impacto estad√≠stico**: Afectan significativamente las medidas de tendencia central y dispersi√≥n
- ‚ö†Ô∏è **Se√±ales de alarma**: Pueden indicar errores de medici√≥n, problemas en la recolecci√≥n o fen√≥menos excepcionales

### Importancia de Detectar Outliers

#### **¬øPor qu√© es crucial identificarlos?**

1. **üìà Sesgan las estad√≠sticas descriptivas**
   - La media es muy sensible a valores extremos
   - Inflan artificialmente la desviaci√≥n est√°ndar
   - Pueden hacer que datos normales parezcan anormales

2. **üî¨ Afectan modelos estad√≠sticos**
   - Reducen la precisi√≥n de modelos predictivos
   - Violan supuestos de normalidad
   - Pueden generar conclusiones err√≥neas

3. **üí° Revelan informaci√≥n valiosa**
   - Pueden indicar errores en los datos
   - Sugieren poblaciones especiales
   - Se√±alan oportunidades o riesgos √∫nicos

## 1. TIPOS DE VALORES AT√çPICOS

### 1.1 Seg√∫n su Origen

#### **üîß Outliers por Errores**
- **Errores de medici√≥n**: Instrumentos defectuosos, calibraci√≥n incorrecta
- **Errores de captura**: Digitaci√≥n incorrecta, problemas de transmisi√≥n
- **Errores de procesamiento**: Conversiones de unidades err√≥neas

**Ejemplo:** Un empleado con salario registrado como $500,000,000 en lugar de $50,000

#### **üåü Outliers Naturales**
- **Valores leg√≠timos** pero extremos dentro del contexto
- **Representan casos reales** de la poblaci√≥n
- **No deben eliminarse** autom√°ticamente

**Ejemplo:** Un CEO con salario de $2,000,000 en una empresa donde el promedio es $60,000

### 1.2 Seg√∫n su Dimensionalidad

#### **üìè Outliers Univariados**
- Extremos en **una sola variable**
- F√°ciles de detectar visualmente
- M√©todos: IQR, Z-score, box plots

#### **üîó Outliers Multivariados**
- Combinaciones extremas de **m√∫ltiples variables**
- M√°s dif√≠ciles de detectar
- Una observaci√≥n puede ser normal en cada variable individual pero at√≠pica en su combinaci√≥n

**Ejemplo:** Persona de 25 a√±os con 40 a√±os de experiencia laboral (ambos valores pueden ser normales por separado)

## 2. M√âTODOS DE DETECCI√ìN

### 2.1 M√©todo del Rango Intercuart√≠lico (IQR)

#### **üìä Principio Fundamental**
El m√©todo IQR utiliza los cuartiles para definir "valores normales" y establecer l√≠mites para identificar outliers.

#### **üî¢ F√≥rmulas Clave**
```
IQR = Q3 - Q1
L√≠mite inferior = Q1 - 1.5 √ó IQR
L√≠mite superior = Q3 + 1.5 √ó IQR
```

#### **‚öñÔ∏è Criterio de Clasificaci√≥n**
- **Outlier moderado**: Fuera de [Q1 - 1.5√óIQR, Q3 + 1.5√óIQR]
- **Outlier extremo**: Fuera de [Q1 - 3√óIQR, Q3 + 3√óIQR]

#### **‚úÖ Ventajas**
- No asume distribuci√≥n normal
- Robusto ante outliers existentes
- F√°cil de calcular e interpretar
- Visualizaci√≥n clara con box plots

#### **‚ùå Limitaciones**
- El factor 1.5 es arbitrario
- Puede ser muy conservador o liberal seg√∫n la distribuci√≥n
- No funciona bien con distribuciones muy asim√©tricas

### 2.2 M√©todo Z-Score (Puntuaci√≥n Est√°ndar)

#### **üìä Principio Fundamental**
Mide cu√°ntas desviaciones est√°ndar se aleja una observaci√≥n de la media.

#### **üî¢ F√≥rmula**
```
Z = (x - Œº) / œÉ
```
Donde:
- x = valor observado
- Œº = media de la poblaci√≥n (o muestra)
- œÉ = desviaci√≥n est√°ndar

#### **‚öñÔ∏è Criterios Comunes**
- **|Z| > 2**: Posible outlier (‚âà 5% de datos normales)
- **|Z| > 2.5**: Probable outlier (‚âà 1.2% de datos normales)
- **|Z| > 3**: Outlier muy probable (‚âà 0.3% de datos normales)

#### **‚úÖ Ventajas**
- Basado en propiedades estad√≠sticas conocidas
- Permite establecer niveles de confianza
- √ötil para distribuciones aproximadamente normales

#### **‚ùå Limitaciones**
- **Asume distribuci√≥n normal**
- **Sensible a outliers existentes** (la media y desviaci√≥n est√°ndar se ven afectadas por outliers)
- No funciona bien con distribuciones asim√©tricas

### 2.3 Z-Score Modificado (M√©todo Robusto)

#### **üìä Mejora del Z-Score Tradicional**
Utiliza la **mediana** y la **desviaci√≥n absoluta mediana (MAD)** en lugar de la media y desviaci√≥n est√°ndar.

#### **üî¢ F√≥rmulas**
```
Z_modificado = 0.6745 √ó (x - mediana) / MAD
MAD = mediana(|xi - mediana|)
```

#### **‚öñÔ∏è Criterio**
- **|Z_modificado| > 3.5**: Outlier probable

#### **‚úÖ Ventajas**
- **Robusto ante outliers existentes**
- No requiere distribuci√≥n normal
- M√°s confiable que el Z-score tradicional

#### **‚ùå Limitaciones**
- Menos conocido y usado
- El factor 0.6745 puede no ser √≥ptimo para todas las distribuciones

## 3. ESTRATEGIAS DE TRATAMIENTO

### 3.1 Mantener los Outliers

#### **ü§î ¬øCu√°ndo mantenerlos?**
- **Son valores leg√≠timos** de la poblaci√≥n de inter√©s
- **Representan casos extremos importantes** (ej: clientes VIP)
- **El an√°lisis se beneficia** de incluir la variabilidad completa
- **M√©todos robustos** se usar√°n en an√°lisis posteriores

#### **üìà Consideraciones**
- Usar estad√≠sticas robustas (mediana, IQR)
- Reportar resultados con y sin outliers
- Mencionar su presencia en conclusiones

### 3.2 Transformar los Datos

#### **üîÑ Tipos de Transformaciones**

**1. Transformaci√≥n Logar√≠tmica**
```
x_transformado = log(x)
```
- Reduce el impacto de valores muy grandes
- √ötil para datos con distribuci√≥n log-normal
- ‚ö†Ô∏è Requiere valores positivos

**2. Transformaci√≥n de Ra√≠z Cuadrada**
```
x_transformado = ‚àöx
```
- Menos dr√°stica que el logaritmo
- √ötil para datos de conteo
- ‚ö†Ô∏è Requiere valores no negativos

**3. Transformaci√≥n Box-Cox**
- Encuentra la transformaci√≥n √≥ptima autom√°ticamente
- Muy flexible pero m√°s compleja de interpretar

**4. Winsorizaci√≥n**
```
Reemplazar valores extremos por percentiles espec√≠ficos
```
- Ejemplo: Reemplazar valores > P95 por el valor del P95
- Mantiene el tama√±o de la muestra
- Reduce el impacto sin eliminar observaciones

#### **‚úÖ Ventajas de Transformar**
- Mantiene todas las observaciones
- Puede normalizar la distribuci√≥n
- Reduce el impacto sin perder informaci√≥n

#### **‚ùå Limitaciones**
- Cambia la interpretaci√≥n de los datos
- No siempre es apropiado seg√∫n el contexto
- Puede complicar la comunicaci√≥n de resultados

### 3.3 Eliminar Outliers

#### **‚ö†Ô∏è ¬øCu√°ndo es apropiado?**
- **Errores de medici√≥n confirmados**
- **Datos de poblaciones diferentes** a la de inter√©s
- **An√°lisis exploratorio inicial** para entender patrones
- **M√©todos estad√≠sticos que lo requieren** expl√≠citamente

#### **üìã Proceso Recomendado**
1. **Documentar la decisi√≥n** y criterios utilizados
2. **Reportar cu√°ntas observaciones** se eliminaron
3. **Presentar resultados** con y sin outliers
4. **Justificar por qu√©** es apropiado eliminarlos

#### **‚ö†Ô∏è PRECAUCIONES**
- **Nunca eliminar autom√°ticamente** sin investigar
- **Considerar el contexto** del problema
- **Evaluar si represent an informaci√≥n valiosa**
- **Documentar todas las decisiones**

### 3.4 An√°lisis Separado

#### **üîç Enfoque Dual**
- Realizar an√°lisis **principal sin outliers**
- An√°lisis **secundario solo con outliers**
- **Comparar resultados** y extraer insights

#### **üí° Beneficios**
- Entiende ambas poblaciones
- No pierde informaci√≥n valiosa
- Proporciona an√°lisis m√°s completo

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaciones
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 11
````

### CREACI√ìN DEL DATASET INMOBILIARIO CON OUTLIERS

````python
# Configurar semilla para reproducibilidad
np.random.seed(42)
n_samples = 1000

print("üè† Generando dataset inmobiliario con outliers naturales...")

# Dataset base con propiedades normales
area_m2 = np.random.normal(120, 40, n_samples)
area_m2 = np.clip(area_m2, 50, 300)

# Precio base correlacionado con √°rea
precio_base = 2000 * area_m2 + np.random.normal(50000, 30000, n_samples)
precio_millones = np.clip(precio_base, 100000, 800000)

# A√±os de construcci√≥n
anos_construccion = np.random.randint(1950, 2024, n_samples)
antiguedad = 2024 - anos_construccion

# N√∫mero de habitaciones (correlacionado con √°rea)
habitaciones = np.round(area_m2 / 40).astype(int)
habitaciones = np.clip(habitaciones, 1, 6)

# Ubicaci√≥n (afecta precio)
ubicaciones = ['Centro', 'Norte', 'Sur', 'Este', 'Oeste']
ubicacion = np.random.choice(ubicaciones, n_samples)

# Ajuste por ubicaci√≥n
factor_ubicacion = {'Centro': 1.5, 'Norte': 1.2, 'Sur': 0.8, 'Este': 0.9, 'Oeste': 1.0}
for i, loc in enumerate(ubicacion):
    precio_millones[i] *= factor_ubicacion[loc]

# ‚ö†Ô∏è INTRODUCIR OUTLIERS CONTROLADOS ‚ö†Ô∏è

# 1. Mansiones de lujo (outliers naturales por √°rea y precio)
n_mansiones = 15
indices_mansiones = np.random.choice(n_samples, n_mansiones, replace=False)
area_m2[indices_mansiones] = np.random.uniform(400, 1000, n_mansiones)
precio_millones[indices_mansiones] = np.random.uniform(1500000, 5000000, n_mansiones)
habitaciones[indices_mansiones] = np.random.randint(6, 12, n_mansiones)

# 2. Propiedades hist√≥ricas muy antiguas
n_historicas = 8
indices_historicas = np.random.choice(
    [i for i in range(n_samples) if i not in indices_mansiones], 
    n_historicas, replace=False
)
anos_construccion[indices_historicas] = np.random.randint(1800, 1900, n_historicas)
antiguedad[indices_historicas] = 2024 - anos_construccion[indices_historicas]
# Precio alto por valor hist√≥rico
precio_millones[indices_historicas] *= np.random.uniform(2.0, 3.5, n_historicas)

# 3. Errores de captura de datos (outliers por error)
n_errores = 5
indices_errores = np.random.choice(
    [i for i in range(n_samples) if i not in indices_mansiones and i not in indices_historicas], 
    n_errores, replace=False
)
# Precios imposibles (errores de digitaci√≥n)
precio_millones[indices_errores] = np.random.uniform(50000000, 100000000, n_errores)

# Crear DataFrame
df_inmobiliario = pd.DataFrame({
    'precio_usd': precio_millones.round(0).astype(int),
    'area_m2': area_m2.round(1),
    'habitaciones': habitaciones,
    'anos_construccion': anos_construccion,
    'antiguedad_anos': antiguedad,
    'ubicacion': ubicacion
})

# Crear variables categ√≥ricas adicionales
def categorizar_precio(precio):
    if precio < 200000:
        return 'Econ√≥mico'
    elif precio < 500000:
        return 'Medio'
    elif precio < 1000000:
        return 'Alto'
    else:
        return 'Lujo'

def categorizar_area(area):
    if area < 80:
        return 'Peque√±a'
    elif area < 150:
        return 'Mediana'
    elif area < 250:
        return 'Grande'
    else:
        return 'Mansi√≥n'

df_inmobiliario['categoria_precio'] = df_inmobiliario['precio_usd'].apply(categorizar_precio)
df_inmobiliario['categoria_area'] = df_inmobiliario['area_m2'].apply(categorizar_area)

print("‚úÖ Dataset inmobiliario creado exitosamente")
print("üìä Informaci√≥n del dataset:")
print("="*50)
print(f"Tama√±o: {df_inmobiliario.shape}")
print(f"Variables: {list(df_inmobiliario.columns)}")

print("\nüîç Primeras observaciones:")
print(df_inmobiliario.head())

print("\nüìà Estad√≠sticas descriptivas iniciales:")
print(df_inmobiliario.describe())

# Identificar outliers introducidos para verificaci√≥n
print(f"\nüè∞ Mansiones introducidas: {len(indices_mansiones)}")
print(f"üèõÔ∏è Propiedades hist√≥ricas: {len(indices_historicas)}")
print(f"‚ùå Errores de captura: {len(indices_errores)}")
````

## 1. VISUALIZACI√ìN INICIAL PARA IDENTIFICAR OUTLIERS

## 2. M√âTODO IQR (RANGO INTERCUART√çLICO)

````python
print("\n" + "="*70)
print("üìä 2. DETECCI√ìN DE OUTLIERS - M√âTODO IQR")
print("="*70)

def detectar_outliers_iqr(df, columna, factor=1.5):
    """
    Detectar outliers usando el m√©todo del Rango Intercuart√≠lico
    
    Args:
        df: DataFrame
        columna: nombre de la columna
        factor: multiplicador para el IQR (1.5 = moderado, 3.0 = extremo)
    
    Returns:
        dict con informaci√≥n de outliers
    """
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    
    limite_inferior = Q1 - factor * IQR
    limite_superior = Q3 + factor * IQR
    
    # Identificar outliers
    outliers_mask = (df[columna] < limite_inferior) | (df[columna] > limite_superior)
    outliers_df = df[outliers_mask]
    
    # Estad√≠sticas
    n_outliers = len(outliers_df)
    porcentaje_outliers = (n_outliers / len(df)) * 100
    
    resultado = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'limite_inferior': limite_inferior,
        'limite_superior': limite_superior,
        'n_outliers': n_outliers,
        'porcentaje_outliers': porcentaje_outliers,
        'outliers_indices': outliers_df.index.tolist(),
        'outliers_valores': outliers_df[columna].tolist(),
        'mask': outliers_mask
    }
    
    return resultado

def reporte_outliers_iqr(df, columna, factor=1.5):
    """
    Generar reporte completo de outliers usando IQR
    """
    print(f"\nüîç AN√ÅLISIS IQR - {columna.upper()} (factor = {factor})")
    print("="*60)
    
    resultado = detectar_outliers_iqr(df, columna, factor)
    
    print(f"üìä Estad√≠sticas de la distribuci√≥n:")
    print(f"   Q1 (Percentil 25): {resultado['Q1']:,.2f}")
    print(f"   Q3 (Percentil 75): {resultado['Q3']:,.2f}")
    print(f"   IQR: {resultado['IQR']:,.2f}")
    
    print(f"\nüìè L√≠mites para outliers:")
    print(f"   L√≠mite inferior: {resultado['limite_inferior']:,.2f}")
    print(f"   L√≠mite superior: {resultado['limite_superior']:,.2f}")
    
    print(f"\n‚ö†Ô∏è Outliers detectados:")
    print(f"   N√∫mero de outliers: {resultado['n_outliers']}")
    print(f"   Porcentaje del total: {resultado['porcentaje_outliers']:.2f}%")
    
    if resultado['n_outliers'] > 0:
        print(f"\nüéØ Valores outliers m√°s extremos:")
        valores_outliers = sorted(resultado['outliers_valores'])
        print(f"   M√≠nimo outlier: {min(valores_outliers):,.2f}")
        print(f"   M√°ximo outlier: {max(valores_outliers):,.2f}")
        
        # Mostrar algunos outliers espec√≠ficos
        if len(valores_outliers) <= 10:
            print(f"   Todos los outliers: {[round(x, 2) for x in valores_outliers]}")
        else:
            print(f"   Primeros 5: {[round(x, 2) for x in valores_outliers[:5]]}")
            print(f"   √öltimos 5: {[round(x, 2) for x in valores_outliers[-5:]]}")
    
    return resultado

# Analizar outliers en precio usando IQR
resultado_precio_iqr = reporte_outliers_iqr(df_inmobiliario, 'precio_usd', factor=1.5)
resultado_precio_iqr_extremo = reporte_outliers_iqr(df_inmobiliario, 'precio_usd', factor=3.0)

# Analizar outliers en √°rea
resultado_area_iqr = reporte_outliers_iqr(df_inmobiliario, 'area_m2', factor=1.5)
````

## 3. M√âTODO Z-SCORE (PUNTUACI√ìN EST√ÅNDAR)

````python
print("\n" + "="*70)
print("üìä 3. DETECCI√ìN DE OUTLIERS - M√âTODO Z-SCORE")
print("="*70)

def detectar_outliers_zscore(df, columna, umbral=3):
    """
    Detectar outliers usando Z-score (puntuaci√≥n est√°ndar)
    """
    media = df[columna].mean()
    std_dev = df[columna].std()
    
    # Calcular Z-scores
    z_scores = np.abs((df[columna] - media) / std_dev)
    
    # Identificar outliers
    outliers_mask = z_scores > umbral
    outliers_df = df[outliers_mask]
    
    # Estad√≠sticas
    n_outliers = len(outliers_df)
    porcentaje_outliers = (n_outliers / len(df)) * 100
    
    resultado = {
        'media': media,
        'std_dev': std_dev,
        'umbral': umbral,
        'z_scores': z_scores,
        'n_outliers': n_outliers,
        'porcentaje_outliers': porcentaje_outliers,
        'outliers_indices': outliers_df.index.tolist(),
        'outliers_valores': outliers_df[columna].tolist(),
        'outliers_z_scores': z_scores[outliers_mask].tolist(),
        'mask': outliers_mask
    }
    
    return resultado

def reporte_outliers_zscore(df, columna, umbral=3):
    """
    Generar reporte completo de outliers usando Z-score
    """
    print(f"\nüîç AN√ÅLISIS Z-SCORE - {columna.upper()} (umbral = {umbral})")
    print("="*60)
    
    resultado = detectar_outliers_zscore(df, columna, umbral)
    
    print(f"üìä Estad√≠sticas de la distribuci√≥n:")
    print(f"   Media: {resultado['media']:,.2f}")
    print(f"   Desviaci√≥n est√°ndar: {resultado['std_dev']:,.2f}")
    print(f"   Umbral Z-score: ¬±{resultado['umbral']}")
    
    print(f"\n‚ö†Ô∏è Outliers detectados:")
    print(f"   N√∫mero de outliers: {resultado['n_outliers']}")
    print(f"   Porcentaje del total: {resultado['porcentaje_outliers']:.2f}%")
    
    if resultado['n_outliers'] > 0:
        print(f"\nüéØ Outliers m√°s extremos (por Z-score):")
        # Ordenar por Z-score absoluto
        outliers_data = list(zip(resultado['outliers_valores'], 
                                resultado['outliers_z_scores'],
                                resultado['outliers_indices']))
        outliers_data.sort(key=lambda x: abs(x[1]), reverse=True)
        
        # Mostrar top 5 outliers
        print("   Top 5 outliers m√°s extremos:")
        for i, (valor, z_score, indice) in enumerate(outliers_data[:5]):
            print(f"   {i+1}. Valor: {valor:,.0f}, Z-score: {z_score:.2f}, √çndice: {indice}")
    
    return resultado

# Analizar outliers en precio usando Z-score
resultado_precio_zscore = reporte_outliers_zscore(df_inmobiliario, 'precio_usd', umbral=2)
resultado_precio_zscore_estricto = reporte_outliers_zscore(df_inmobiliario, 'precio_usd', umbral=3)

# Analizar outliers en √°rea usando Z-score
resultado_area_zscore = reporte_outliers_zscore(df_inmobiliario, 'area_m2', umbral=2)
````

## 4. M√âTODO Z-SCORE MODIFICADO (ROBUSTO)

````python
print("\n" + "="*70)
print("üìä 4. DETECCI√ìN DE OUTLIERS - Z-SCORE MODIFICADO")
print("="*70)

def detectar_outliers_zscore_modificado(df, columna, umbral=3.5):
    """
    Detectar outliers usando Z-score modificado (m√©todo robusto)
    Utiliza mediana y MAD en lugar de media y desviaci√≥n est√°ndar
    """
    mediana = df[columna].median()
    
    # Calcular MAD (Median Absolute Deviation)
    mad = np.median(np.abs(df[columna] - mediana))
    
    # Factor de consistencia para distribuci√≥n normal
    factor_consistencia = 1.4826  # Equivalente a 0.6745 en denominador
    
    # Calcular Z-scores modificados
    z_scores_mod = 0.6745 * (df[columna] - mediana) / mad
    z_scores_mod_abs = np.abs(z_scores_mod)
    
    # Identificar outliers
    outliers_mask = z_scores_mod_abs > umbral
    outliers_df = df[outliers_mask]
    
    # Estad√≠sticas
    n_outliers = len(outliers_df)
    porcentaje_outliers = (n_outliers / len(df)) * 100
    
    resultado = {
        'mediana': mediana,
        'mad': mad,
        'umbral': umbral,
        'z_scores_modificados': z_scores_mod,
        'z_scores_abs': z_scores_mod_abs,
        'n_outliers': n_outliers,
        'porcentaje_outliers': porcentaje_outliers,
        'outliers_indices': outliers_df.index.tolist(),
        'outliers_valores': outliers_df[columna].tolist(),
        'outliers_z_mod': z_scores_mod[outliers_mask].tolist(),
        'mask': outliers_mask
    }
    
    return resultado

def reporte_outliers_zscore_modificado(df, columna, umbral=3.5):
    """
    Generar reporte completo de outliers usando Z-score modificado
    """
    print(f"\nüîç AN√ÅLISIS Z-SCORE MODIFICADO - {columna.upper()} (umbral = {umbral})")
    print("="*60)
    
    resultado = detectar_outliers_zscore_modificado(df, columna, umbral)
    
    print(f"üìä Estad√≠sticas robustas de la distribuci√≥n:")
    print(f"   Mediana: {resultado['mediana']:,.2f}")
    print(f"   MAD (Desviaci√≥n Absoluta Mediana): {resultado['mad']:,.2f}")
    print(f"   Umbral Z-score modificado: ¬±{resultado['umbral']}")
    
    print(f"\n‚ö†Ô∏è Outliers detectados:")
    print(f"   N√∫mero de outliers: {resultado['n_outliers']}")
    print(f"   Porcentaje del total: {resultado['porcentaje_outliers']:.2f}%")
    
    if resultado['n_outliers'] > 0:
        print(f"\nüéØ Outliers m√°s extremos (por Z-score modificado):")
        # Ordenar por Z-score modificado absoluto
        outliers_data = list(zip(resultado['outliers_valores'], 
                                resultado['outliers_z_mod'],
                                resultado['outliers_indices']))
        outliers_data.sort(key=lambda x: abs(x[1]), reverse=True)
        
        # Mostrar top 5 outliers
        print("   Top 5 outliers m√°s extremos:")
        for i, (valor, z_mod, indice) in enumerate(outliers_data[:5]):
            print(f"   {i+1}. Valor: {valor:,.0f}, Z-mod: {z_mod:.2f}, √çndice: {indice}")
    
    return resultado

# Analizar outliers usando Z-score modificado
resultado_precio_zmod = reporte_outliers_zscore_modificado(df_inmobiliario, 'precio_usd')
resultado_area_zmod = reporte_outliers_zscore_modificado(df_inmobiliario, 'area_m2')
````

## 5. COMPARACI√ìN DE M√âTODOS

````python
print("\n" + "="*70)
print("üìä 5. COMPARACI√ìN DE M√âTODOS DE DETECCI√ìN")
print("="*70)

def comparar_metodos_deteccion(df, columna):
    """
    Comparar todos los m√©todos de detecci√≥n de outliers para una variable
    """
    print(f"\nüìä COMPARACI√ìN DE M√âTODOS - {columna.upper()}")
    print("="*60)
    
    # Aplicar todos los m√©todos
    iqr_15 = detectar_outliers_iqr(df, columna, 1.5)
    iqr_30 = detectar_outliers_iqr(df, columna, 3.0)
    zscore_2 = detectar_outliers_zscore(df, columna, 2)
    zscore_3 = detectar_outliers_zscore(df, columna, 3)
    zmod = detectar_outliers_zscore_modificado(df, columna, 3.5)
    
    # Crear tabla comparativa
    metodos = [
        ('IQR (factor=1.5)', iqr_15['n_outliers'], iqr_15['porcentaje_outliers']),
        ('IQR (factor=3.0)', iqr_30['n_outliers'], iqr_30['porcentaje_outliers']),
        ('Z-score (>2)', zscore_2['n_outliers'], zscore_2['porcentaje_outliers']),
        ('Z-score (>3)', zscore_3['n_outliers'], zscore_3['porcentaje_outliers']),
        ('Z-score Modificado (>3.5)', zmod['n_outliers'], zmod['porcentaje_outliers'])
    ]
    
    print("üìã Resumen comparativo:")
    print("-" * 60)
    print(f"{'M√©todo':<25} {'N¬∞ Outliers':<12} {'Porcentaje':<12}")
    print("-" * 60)
    
    for metodo, n_out, pct_out in metodos:
        print(f"{metodo:<25} {n_out:<12} {pct_out:<12.2f}%")
    
    # Encontrar outliers consistentes (detectados por m√∫ltiples m√©todos)
    sets_outliers = [
        set(iqr_15['outliers_indices']),
        set(zscore_2['outliers_indices']),
        set(zmod['outliers_indices'])
    ]
    
    # Intersecci√≥n de los tres m√©todos principales
    outliers_consistentes = sets_outliers[0] & sets_outliers[1] & sets_outliers[2]
    outliers_cualquier_metodo = sets_outliers[0] | sets_outliers[1] | sets_outliers[2]
    
    print(f"\nüéØ An√°lisis de consistencia:")
    print(f"   Outliers detectados por TODOS los m√©todos: {len(outliers_consistentes)}")
    print(f"   Outliers detectados por ALG√öN m√©todo: {len(outliers_cualquier_metodo)}")
    
    if len(outliers_consistentes) > 0:
        print(f"   √çndices consistentes: {sorted(list(outliers_consistentes))[:10]}...")
    
    return {
        'iqr_15': iqr_15,
        'zscore_2': zscore_2, 
        'zmod': zmod,
        'consistentes': outliers_consistentes,
        'cualquier_metodo': outliers_cualquier_metodo
    }

# Comparar m√©todos para precio
comparacion_precio = comparar_metodos_deteccion(df_inmobiliario, 'precio_usd')

# Comparar m√©todos para √°rea
comparacion_area = comparar_metodos_deteccion(df_inmobiliario, 'area_m2')
````

## 6. VISUALIZACI√ìN DE OUTLIERS DETECTADOS

````python
print("\n" + "="*70)
print("üìä 6. VISUALIZACI√ìN DE OUTLIERS DETECTADOS")
print("="*70)

def visualizar_outliers_detectados(df, columna, resultados_deteccion):
    """
    Crear visualizaciones que muestren los outliers detectados.
    - df: DataFrame con los datos
    - columna: nombre de la columna num√©rica a analizar
    - resultados_deteccion: dict con resultados de detecci√≥n.
        Debe poder contener:
          - 'iqr_15': {'outliers_valores': iterable_de_valores}
          - 'consistentes': iterable de √≠ndices considerados outliers consistentes
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # =========================
    # 1) HISTOGRAMA + L√çNEAS EN OUTLIERS IQR
    # =========================
    axes[0, 0].hist(df[columna], bins=50, alpha=0.7, color='skyblue', edgecolor='black')

    # Obtiene valores outliers por IQR si existen
    outliers_iqr = []
    if 'iqr_15' in resultados_deteccion:
        outliers_iqr = resultados_deteccion['iqr_15'].get('outliers_valores', [])
    # Asegurar que sea iterable tipo lista/array
    outliers_iqr = list(outliers_iqr) if outliers_iqr is not None else []

    for outlier in outliers_iqr:
        axes[0, 0].axvline(outlier, color='red', alpha=0.6, linewidth=1)

    axes[0, 0].set_title(f'Histograma de {columna}\n(L√≠neas rojas = Outliers IQR)', fontweight='bold')
    axes[0, 0].set_xlabel(columna)
    axes[0, 0].set_ylabel('Frecuencia')
    axes[0, 0].grid(True, alpha=0.3)
    
    # =========================
    # 2) BOX PLOT + MARCA DE OUTLIERS CONSISTENTES
    # =========================
    box_data = df[columna]
    bp = axes[0, 1].boxplot(box_data, patch_artist=True, vert=True)
    bp['boxes'][0].set_facecolor('lightblue')
    
    # Outliers consistentes pueden venir como set; convertir a lista para indexar
    outliers_consistentes = resultados_deteccion.get('consistentes', [])
    # Si es set, convi√©rtelo a lista
    if isinstance(outliers_consistentes, set):
        outliers_indices = list(outliers_consistentes)
    else:
        outliers_indices = list(outliers_consistentes)

    if len(outliers_indices) > 0:
        # Filtrar por √≠ndices v√°lidos (que existan en df.index)
        outliers_indices = [i for i in outliers_indices if i in df.index]
        if len(outliers_indices) > 0:
            valores_extremos = df.loc[outliers_indices, columna]
            axes[0, 1].scatter([1]*len(valores_extremos), valores_extremos, 
                               color='red', s=50, alpha=0.8, label='Outliers Consistentes')
            axes[0, 1].legend()
    
    axes[0, 1].set_title(f'Box Plot de {columna}', fontweight='bold')
    axes[0, 1].set_ylabel(columna)
    axes[0, 1].grid(True, alpha=0.3)
    
    # =========================
    # 3) Q-Q PLOT
    # =========================
    stats.probplot(df[columna].dropna(), dist="norm", plot=axes[1, 0])
    axes[1, 0].set_title(f'Q-Q Plot de {columna}\n(Evaluaci√≥n de Normalidad)', fontweight='bold')
    axes[1, 0].grid(True, alpha=0.3)
    
    # =========================
    # 4) SCATTER: √çNDICE vs VALOR + OUTLIERS CONSISTENTES
    # =========================
    axes[1, 1].scatter(df.index, df[columna], alpha=0.6, color='blue', s=20)

    if len(outliers_indices) > 0:
        outliers_indices = [i for i in outliers_indices if i in df.index]
        if len(outliers_indices) > 0:
            outliers_valores = df.loc[outliers_indices, columna]
            axes[1, 1].scatter(outliers_indices, outliers_valores, 
                               color='red', s=50, alpha=0.8, label='Outliers Consistentes')
            axes[1, 1].legend()
    
    axes[1, 1].set_title(f'Valores de {columna} por √çndice', fontweight='bold')
    axes[1, 1].set_xlabel('√çndice')
    axes[1, 1].set_ylabel(columna)
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

print("üìä Visualizando outliers en precio...")
visualizar_outliers_detectados(df_inmobiliario, 'precio_usd', comparacion_precio)

print("üìä Visualizando outliers en √°rea...")
visualizar_outliers_detectados(df_inmobiliario, 'area_m2', comparacion_area)
````

## 7. AN√ÅLISIS DE IMPACTO DE OUTLIERS

````python
print("\n" + "="*70)
print("üìä 7. AN√ÅLISIS DE IMPACTO DE OUTLIERS")
print("="*70)

def analizar_impacto_outliers(df, columna, outliers_indices):
    """
    Analizar el impacto de los outliers en las estad√≠sticas descriptivas
    """
    print(f"\nüîç IMPACTO DE OUTLIERS EN {columna.upper()}")
    print("="*60)
    
    # Estad√≠sticas con outliers
    con_outliers = {
        'media': df[columna].mean(),
        'mediana': df[columna].median(),
        'std_dev': df[columna].std(),
        'q1': df[columna].quantile(0.25),
        'q3': df[columna].quantile(0.75),
        'min': df[columna].min(),
        'max': df[columna].max(),
        'n_obs': len(df)
    }
    
    # Estad√≠sticas sin outliers
    df_sin_outliers = df.drop(index=outliers_indices)
    sin_outliers = {
        'media': df_sin_outliers[columna].mean(),
        'mediana': df_sin_outliers[columna].median(),
        'std_dev': df_sin_outliers[columna].std(),
        'q1': df_sin_outliers[columna].quantile(0.25),
        'q3': df_sin_outliers[columna].quantile(0.75),
        'min': df_sin_outliers[columna].min(),
        'max': df_sin_outliers[columna].max(),
        'n_obs': len(df_sin_outliers)
    }
    
    # Calcular cambios porcentuales
    def cambio_porcentual(antes, despues):
        if antes == 0:
            return float('inf') if despues != 0 else 0
        return ((despues - antes) / abs(antes)) * 100
    
    print("üìä COMPARACI√ìN DE ESTAD√çSTICAS:")
    print("-" * 60)
    print(f"{'Estad√≠stica':<15} {'Con Outliers':<15} {'Sin Outliers':<15} {'Cambio %':<10}")
    print("-" * 60)
    
    estadisticas = ['media', 'mediana', 'std_dev', 'min', 'max']
    for stat in estadisticas:
        con = con_outliers[stat]
        sin = sin_outliers[stat]
        cambio = cambio_porcentual(con, sin)
        
        print(f"{stat.capitalize():<15} {con:<15,.0f} {sin:<15,.0f} {cambio:<10.1f}%")
    
    print(f"N¬∞ Observaciones {con_outliers['n_obs']:<15} {sin_outliers['n_obs']:<15} "
          f"{len(outliers_indices)} eliminadas")
    
    # An√°lisis de sensibilidad
    print(f"\nüéØ AN√ÅLISIS DE SENSIBILIDAD:")
    print("-" * 30)
    
    cambio_media = abs(cambio_porcentual(con_outliers['media'], sin_outliers['media']))
    cambio_std = abs(cambio_porcentual(con_outliers['std_dev'], sin_outliers['std_dev']))
    cambio_mediana = abs(cambio_porcentual(con_outliers['mediana'], sin_outliers['mediana']))
    
    if cambio_media > 10:
        print(f"‚ö†Ô∏è  ALTA SENSIBILIDAD en la media: {cambio_media:.1f}% de cambio")
    elif cambio_media > 5:
        print(f"‚ö†Ô∏è  MEDIA SENSIBILIDAD en la media: {cambio_media:.1f}% de cambio")
    else:
        print(f"‚úÖ BAJA SENSIBILIDAD en la media: {cambio_media:.1f}% de cambio")
    
    if cambio_mediana > 5:
        print(f"‚ö†Ô∏è  La mediana cambi√≥ significativamente: {cambio_mediana:.1f}%")
    else:
        print(f"‚úÖ La mediana es robusta: {cambio_mediana:.1f}% de cambio")
    
    print(f"üìä La desviaci√≥n est√°ndar cambi√≥: {cambio_std:.1f}%")
    
    return {
        'con_outliers': con_outliers,
        'sin_outliers': sin_outliers,
        'cambio_media': cambio_media,
        'cambio_mediana': cambio_mediana,
        'cambio_std': cambio_std
    }

# Analizar impacto usando outliers consistentes
if len(comparacion_precio['consistentes']) > 0:
    impacto_precio = analizar_impacto_outliers(df_inmobiliario, 'precio_usd', 
                                              list(comparacion_precio['consistentes']))

if len(comparacion_area['consistentes']) > 0:
    impacto_area = analizar_impacto_outliers(df_inmobiliario, 'area_m2', 
                                           list(comparacion_area['consistentes']))
````

## 8. ESTRATEGIAS DE TRATAMIENTO

````python
print("\n" + "="*70)
print("üìä 8. ESTRATEGIAS DE TRATAMIENTO DE OUTLIERS")
print("="*70)

def aplicar_transformacion_logaritmica(df, columna):
    """
    Aplicar transformaci√≥n logar√≠tmica para reducir impacto de outliers
    """
    # Asegurarse de que todos los valores son positivos
    valores_positivos = df[columna] > 0
    
    if not valores_positivos.all():
        print(f"‚ö†Ô∏è Advertencia: {columna} contiene valores no positivos")
        return None
    
    # Aplicar transformaci√≥n logar√≠tmica
    columna_log = f"{columna}_log"
    df[columna_log] = np.log(df[columna])
    
    return columna_log

def aplicar_winzorizacion(df, columna, percentiles=(5, 95)):
    """
    Aplicar winsorizaci√≥n (reemplazar outliers por percentiles)
    """
    p_inferior, p_superior = percentiles
    limite_inf = np.percentile(df[columna], p_inferior)
    limite_sup = np.percentile(df[columna], p_superior)
    
    columna_winsorizada = f"{columna}_winsorized"
    df[columna_winsorizada] = df[columna].clip(lower=limite_inf, upper=limite_sup)
    
    n_cambios_inf = (df[columna] < limite_inf).sum()
    n_cambios_sup = (df[columna] > limite_sup).sum()
    
    print(f"üîß Winsorizaci√≥n aplicada a {columna}:")
    print(f"   Percentiles usados: {p_inferior}% - {p_superior}%")
    print(f"   L√≠mites: {limite_inf:,.2f} - {limite_sup:,.2f}")
    print(f"   Valores modificados (inferior): {n_cambios_inf}")
    print(f"   Valores modificados (superior): {n_cambios_sup}")
    
    return columna_winsorizada

def crear_dataset_sin_outliers(df, outliers_indices, nombre_sufijo="_sin_outliers"):
    """
    Crear una versi√≥n del dataset sin los outliers especificados
    """
    df_limpio = df.drop(index=outliers_indices)
    
    print(f"üìä Dataset creado sin outliers:")
    print(f"   Observaciones originales: {len(df)}")
    print(f"   Outliers eliminados: {len(outliers_indices)}")
    print(f"   Observaciones restantes: {len(df_limpio)}")
    print(f"   Porcentaje retenido: {(len(df_limpio)/len(df))*100:.1f}%")
    
    return df_limpio

print("\nüîß APLICANDO ESTRATEGIAS DE TRATAMIENTO:")

# 1. Transformaci√≥n logar√≠tmica
print("\n1. TRANSFORMACI√ìN LOGAR√çTMICA:")
print("-" * 40)
precio_log = aplicar_transformacion_logaritmica(df_inmobiliario, 'precio_usd')
if precio_log:
    print(f"‚úÖ Transformaci√≥n aplicada: {precio_log}")

# 2. Winsorizaci√≥n
print("\n2. WINSORIZACI√ìN:")
print("-" * 40)
precio_winsorizado = aplicar_winzorizacion(df_inmobiliario, 'precio_usd', (5, 95))

# 3. Eliminaci√≥n de outliers
print("\n3. ELIMINACI√ìN DE OUTLIERS:")
print("-" * 40)
if len(comparacion_precio['consistentes']) > 0:
    df_sin_outliers = crear_dataset_sin_outliers(df_inmobiliario, 
                                                list(comparacion_precio['consistentes']))
````

## 10. AN√ÅLISIS DE OUTLIERS MULTIVARIADOS

````python
print("\n" + "="*70)
print("üìä 10. DETECCI√ìN DE OUTLIERS MULTIVARIADOS")
print("="*70)

def detectar_outliers_mahalanobis(df, columnas_numericas):
    """
    Detectar outliers multivariados usando la distancia de Mahalanobis
    """
    from scipy.stats import chi2
    
    # Seleccionar solo las columnas num√©ricas especificadas
    data = df[columnas_numericas].dropna()
    
    # Calcular media y matriz de covarianza
    media = data.mean()
    cov_matrix = data.cov()
    
    # Calcular distancia de Mahalanobis para cada observaci√≥n
    def mahalanobis_distance(x, media, cov_inv):
        diff = x - media
        return np.sqrt(diff.T @ cov_inv @ diff)
    
    try:
        cov_inv = np.linalg.inv(cov_matrix)
        distancias = data.apply(lambda row: mahalanobis_distance(row, media, cov_inv), axis=1)
        
        # Punto de corte basado en distribuci√≥n chi-cuadrado
        p_value = 0.001  # 99.9% de confianza
        grados_libertad = len(columnas_numericas)
        punto_corte = chi2.ppf(1 - p_value, grados_libertad)
        
        # Identificar outliers
        outliers_mask = distancias > punto_corte
        outliers_indices = data[outliers_mask].index.tolist()
        
        resultado = {
            'distancias': distancias,
            'punto_corte': punto_corte,
            'outliers_indices': outliers_indices,
            'n_outliers': len(outliers_indices),
            'porcentaje_outliers': (len(outliers_indices) / len(data)) * 100,
            'grados_libertad': grados_libertad
        }
        
        return resultado
        
    except np.linalg.LinAlgError:
        print("‚ö†Ô∏è Error: Matriz de covarianza singular. No se puede calcular Mahalanobis.")
        return None

def reporte_outliers_multivariados(df, columnas_numericas):
    """
    Generar reporte de outliers multivariados
    """
    print(f"\nüîç AN√ÅLISIS MULTIVARIADO - DISTANCIA DE MAHALANOBIS")
    print("="*60)
    print(f"Variables analizadas: {', '.join(columnas_numericas)}")
    
    resultado = detectar_outliers_mahalanobis(df, columnas_numericas)
    
    if resultado is None:
        return None
    
    print(f"\nüìä Resultados:")
    print(f"   Grados de libertad: {resultado['grados_libertad']}")
    print(f"   Punto de corte (p=0.001): {resultado['punto_corte']:.2f}")
    print(f"   Outliers multivariados detectados: {resultado['n_outliers']}")
    print(f"   Porcentaje del total: {resultado['porcentaje_outliers']:.2f}%")
    
    if resultado['n_outliers'] > 0:
        print(f"\nüéØ Top 5 outliers multivariados:")
        # Ordenar por distancia de Mahalanobis
        distancias_outliers = resultado['distancias'].loc[resultado['outliers_indices']]
        distancias_ordenadas = distancias_outliers.sort_values(ascending=False)
        
        for i, (indice, distancia) in enumerate(distancias_ordenadas.head().items()):
            print(f"   {i+1}. √çndice: {indice}, Distancia: {distancia:.2f}")
            # Mostrar valores de las variables para este outlier
            valores = df.loc[indice, columnas_numericas]
            print(f"      Valores: {dict(valores.round(0))}")
    
    return resultado

# Analizar outliers multivariados
columnas_para_mahalanobis = ['precio_usd', 'area_m2', 'habitaciones', 'antiguedad_anos']
outliers_multivariados = reporte_outliers_multivariados(df_inmobiliario, columnas_para_mahalanobis)
````
