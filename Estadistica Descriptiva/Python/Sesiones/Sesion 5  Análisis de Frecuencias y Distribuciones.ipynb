## 📖 **PARTE 1: Fundamentos del Análisis de Frecuencias**

### **🎯 ¿Qué es el Análisis de Frecuencias?**

> El **análisis de frecuencias** es el proceso de organizar, contar y resumir datos para revelar patrones, tendencias y características distributivas que no son evidentes en los datos sin procesar.

### **🔍 ¿Por qué es Fundamental?**

1. **Revelación de Patrones**: Los datos crudos ocultan patrones que las frecuencias revelan
2. **Compresión de Información**: Miles de datos se resumen en tablas manejables
3. **Base para Visualización**: Fundamento de histogramas y gráficos distributivos
4. **Detección de Anomalías**: Identificación de valores inusuales o gaps en datos
5. **Preparación para Modelado**: Base para selección de distribuciones teóricas

### **📊 Tipos de Análisis de Frecuencias**

```
ANÁLISIS DE FRECUENCIAS
├── VARIABLES CATEGÓRICAS
│   ├── Frecuencias simples
│   └── Tablas de contingencia
└── VARIABLES NUMÉRICAS
    ├── Datos discretos (sin agrupar)
    ├── Datos continuos (agrupados)
    └── Análisis de distribuciones
```

### **💡 El Proceso Completo**

```
PROCESO DE ANÁLISIS DE FRECUENCIAS
1. PREPARACIÓN → 2. AGRUPACIÓN → 3. CONTEO → 4. ANÁLISIS → 5. INTERPRETACIÓN
```

## 📊 **PARTE 2: Tablas de Frecuencias para Variables Categóricas**

### **🏷️ Características de Variables Categóricas**

Las variables categóricas representan **características cualitativas** que dividen a los elementos en categorías:
- **Nominales**: Sin orden natural (género, color, marca)
- **Ordinales**: Con orden específico (nivel educativo, satisfacción)

### **📋 Estructura de Tabla de Frecuencias Categóricas**

| Categoría | Frecuencia Absoluta (fi) | Frecuencia Relativa (hi) | Frecuencia Relativa % | Frecuencia Acumulada |
|-----------|--------------------------|-------------------------|-----------------------|---------------------|
| Cat 1 | n₁ | n₁/n | (n₁/n) × 100 | n₁ |
| Cat 2 | n₂ | n₂/n | (n₂/n) × 100 | n₁ + n₂ |
| ... | ... | ... | ... | ... |
| **Total** | **n** | **1.00** | **100%** | **n** |

### **💡 Ejemplo Práctico: Preferencia de Marcas**

**Datos**: Encuesta a 100 personas sobre marca de smartphone preferida

| Marca | Frecuencia | Porcentaje | Freq. Acumulada | Interpretación |
|-------|------------|------------|-----------------|----------------|
| Samsung | 35 | 35% | 35 | Marca líder |
| Apple | 28 | 28% | 63 | Segunda preferencia |
| Huawei | 15 | 15% | 78 | Tercera opción |
| Xiaomi | 12 | 12% | 90 | Nicho de mercado |
| Otras | 10 | 10% | 100 | Mercado fragmentado |

### **🎯 Interpretaciones Clave**
- **Moda**: Samsung (35%) - La categoría más popular
- **Concentración**: Las 2 primeras marcas capturan 63% del mercado
- **Diversidad**: 10% prefiere marcas alternativas

### **📊 Medidas para Variables Categóricas**

#### **1. Índice de Diversidad de Shannon**
```
H = -Σ(pi × log(pi))
```
- **Interpretación**: Mayor valor = mayor diversidad
- **Aplicación**: Análisis de mercado, biodiversidad, etc.

#### **2. Índice de Concentración de Gini**
```
G = 1 - Σ(pi²)
```
- **Interpretación**: 0 = máxima concentración, cercano a 1 = máxima diversidad
- **Aplicación**: Concentración de mercado, desigualdad

## 📊 **PARTE 4: Construcción Óptima de Intervalos de Clase**

### **📏 Paso 1: Determinar el Número de Intervalos**

#### **Regla de Sturges (La más común)**
```
k = 1 + 3.322 × log₁₀(n)
```
- **Aplicación**: Distribuciones aproximadamente normales
- **Ventaja**: Simple y ampliamente aceptada
- **Limitación**: Puede subestimar para distribuciones asimétricas

#### **Regla de la Raíz Cuadrada**
```
k = √n
```
- **Aplicación**: Alternativa simple
- **Ventaja**: Fácil de calcular mentalmente
- **Uso**: Análisis exploratorio rápido

#### **Regla de Freedman-Diaconis (Robusta)**
```
Amplitud = 2 × IQR × n^(-1/3)
k = (Máx - Mín) / Amplitud
```
- **Aplicación**: Datos con outliers o distribuciones asimétricas
- **Ventaja**: Robusta ante valores extremos
- **Recomendación**: Mejor para datos reales

#### **Regla de Scott (Basada en desviación estándar)**
```
Amplitud = 3.49 × σ × n^(-1/3)
k = (Máx - Mín) / Amplitud
```
- **Aplicación**: Distribuciones cercanas a la normal
- **Ventaja**: Considera la variabilidad de los datos

### **📊 Comparación de Reglas**

| Regla | Fortalezas | Debilidades | Cuándo Usar |
|-------|------------|-------------|-------------|
| **Sturges** | Simple, estándar | Subestima para n grande | Distribuciones normales |
| **Raíz Cuadrada** | Muy simple | Demasiado simple | Exploración rápida |
| **Freedman-Diaconis** | Robusta, adaptativa | Más compleja | Datos con outliers |
| **Scott** | Considera variabilidad | Sensible a outliers | Datos aproximadamente normales |

### **💡 Ejemplo Comparativo**

**Datos**: 1000 observaciones de ingresos mensuales
- **Mínimo**: $15,000, **Máximo**: $180,000
- **IQR**: $45,000, **σ**: $32,000

```
Sturges:    k = 1 + 3.322 × log₁₀(1000) ≈ 11 intervalos
Raíz:       k = √1000 ≈ 32 intervalos
F-D:        Amplitud = 2 × 45,000 × 1000^(-1/3) ≈ $9,000 → k ≈ 18
Scott:      Amplitud = 3.49 × 32,000 × 1000^(-1/3) ≈ $11,200 → k ≈ 15
```

**Recomendación**: Para datos de ingresos (típicamente asimétricos), usar Freedman-Diaconis.

### **📏 Paso 2: Definir Límites de Intervalos**

#### **Principios Fundamentales:**
1. **Cobertura Completa**: Todos los datos deben estar incluidos
2. **Exclusión Mutua**: Cada dato pertenece a un solo intervalo
3. **Amplitud Uniforme**: Preferiblemente intervalos de igual tamaño
4. **Límites Naturales**: Usar números "redondos" cuando sea posible

#### **Notación de Intervalos:**
```
[a, b]  → Cerrado: incluye tanto a como b
(a, b]  → Semicerrado: excluye a, incluye b
[a, b)  → Semicerrado: incluye a, excluye b
(a, b)  → Abierto: excluye tanto a como b
```

#### **Convención Estándar:**
- **Primer intervalo**: [mín, límite₁]
- **Intervalos intermedios**: (límite_{i-1}, límite_i]
- **Interpretación**: El límite inferior se excluye, el superior se incluye

### **💡 Ejemplo Completo: Construcción de Intervalos**

**Datos**: Tiempos de espera en minutos: 2, 3, 5, 7, 8, 10, 12, 15, 18, 22, 25, 28, 30

1. **Aplicar reglas**:
   - n = 13 → Sturges: k = 1 + 3.322 × log₁₀(13) ≈ 5 intervalos
   - Rango = 30 - 2 = 28 minutos
   - Amplitud = 28 ÷ 5 = 5.6 ≈ 6 minutos

2. **Definir intervalos**:
   ```
   [2, 8]    → 2, 3, 5, 7, 8        (5 valores)
   (8, 14]   → 10, 12               (2 valores)
   (14, 20]  → 15, 18               (2 valores)
   (20, 26]  → 22, 25               (2 valores)
   (26, 32]  → 28, 30               (2 valores)
   ```

3. **Tabla resultante**:

| Intervalo | Marca de Clase | Frecuencia | Porcentaje | Freq. Acumulada |
|-----------|----------------|------------|------------|-----------------|
| [2, 8] | 5 | 5 | 38.5% | 5 |
| (8, 14] | 11 | 2 | 15.4% | 7 |
| (14, 20] | 17 | 2 | 15.4% | 9 |
| (20, 26] | 23 | 2 | 15.4% | 11 |
| (26, 32] | 29 | 2 | 15.4% | 13 |

## 📊 **PARTE 5: Histogramas - Visualización de Distribuciones**

### **📈 ¿Qué es un Histograma?**

> Un **histograma** es una representación gráfica de la distribución de frecuencias de una variable continua, usando barras rectangulares cuya área representa la frecuencia de cada intervalo.

### **🔧 Componentes del Histograma**

#### **Ejes:**
- **Eje X (horizontal)**: Valores de la variable (intervalos de clase)
- **Eje Y (vertical)**: Frecuencia (absoluta, relativa, o densidad)

#### **Barras:**
- **Base**: Amplitud del intervalo de clase
- **Altura**: Frecuencia del intervalo
- **Área**: Proporcional a la frecuencia (importante para intervalos desiguales)

### **📊 Tipos de Histogramas**

#### **1. Histograma de Frecuencias Absolutas**
- **Eje Y**: Número de observaciones
- **Uso**: Mostrar conteos reales
- **Limitación**: Dificulta comparación entre datasets de diferente tamaño

#### **2. Histograma de Frecuencias Relativas**
- **Eje Y**: Proporción o porcentaje
- **Uso**: Comparación entre datasets
- **Ventaja**: Suma total = 1.0 (100%)

#### **3. Histograma de Densidad**
- **Eje Y**: Densidad = frecuencia / amplitud
- **Uso**: Intervalos de diferente amplitud
- **Propiedad**: Área total = 1.0

### **🎯 Interpretación de Formas**

#### **Distribución Simétrica (Normal)**
```
      📊
     /|\
    / | \
   /  |  \
  /   |   \
 /_______|_______\
```
- **Características**: Media ≈ Mediana ≈ Moda
- **Ejemplos**: Altura, peso, errores de medición
- **Implicaciones**: Aplicable regla empírica (68-95-99.7)

#### **Distribución Asimétrica Positiva**
```
📊
|\
| \
|  \
|   \____
|_______\
```
- **Características**: Media > Mediana > Moda
- **Ejemplos**: Ingresos, precios de casas, tiempo de vida
- **Implicaciones**: Usar mediana como medida central

#### **Distribución Asimétrica Negativa**
```
    📊
   /|
  / |
 /  |
/   |
\   |
```
- **Características**: Moda > Mediana > Media
- **Ejemplos**: Edad al morir, calificaciones en examen fácil
- **Implicaciones**: Considerar transformaciones

#### **Distribución Bimodal**
```
📊       📊
/|\     /|\
\_|_____/|_/
```
- **Características**: Dos picos claramente diferenciados
- **Ejemplos**: Mezcla de dos poblaciones
- **Implicaciones**: Analizar por separado cada grupo

#### **Distribución Uniforme**
```
📊📊📊📊📊
|  |  |  |  |
|__|__|__|__|
```
- **Características**: Frecuencias aproximadamente iguales
- **Ejemplos**: Números aleatorios, algunos procesos industriales
- **Implicaciones**: Todos los valores igualmente probables

### **⚠️ Problemas Comunes con Histogramas**

#### **1. Número Inadecuado de Intervalos**
- **Muy pocos**: Pérdida de detalle, sobre-suavizado
- **Muy muchos**: Ruido excesivo, dificulta interpretación

#### **2. Límites Inapropiados**
- **Límites artificiales**: Crear picos falsos
- **Intervalos desiguales**: Distorsionar la interpretación

#### **3. Mala Selección de Escala**
- **Escala Y inapropiada**: Exagerar o minimizar diferencias
- **Supresión del cero**: Crear impresiones falsas

## 📊 **PARTE 6: Identificación de Patrones Distributivos**

### **🔍 Análisis Sistemático de Distribuciones**

#### **Paso 1: Evaluación Visual**
1. **Forma general**: ¿Simétrica, asimétrica, multimodal?
2. **Concentración**: ¿Datos concentrados o dispersos?
3. **Colas**: ¿Pesadas, ligeras, una más larga que otra?
4. **Gaps**: ¿Hay espacios sin datos?
5. **Outliers**: ¿Valores extremos aislados?

#### **Paso 2: Métricas Cuantitativas**
1. **Asimetría (Skewness)**: Confirma observaciones visuales
2. **Curtosis**: Evalúa concentración y comportamiento de colas
3. **Coeficiente de variación**: Evalúa dispersión relativa

#### **Paso 3: Comparación con Distribuciones Teóricas**
1. **Normal**: ¿Se ajusta a la campana de Gauss?
2. **Exponencial**: ¿Decrecimiento exponencial?
3. **Uniforme**: ¿Frecuencias relativamente constantes?
4. **Log-normal**: ¿Asimétrica que se normaliza con logaritmo?

### **📚 Catálogo de Distribuciones Comunes**

#### **1. Distribución Normal (Gaussiana)**
- **Forma**: Campana simétrica
- **Parámetros**: μ (media) y σ (desviación estándar)
- **Aplicaciones**: Errores, medidas biológicas, promedios
- **Identificación**: Simetría + curtosis ≈ 0

#### **2. Distribución Exponencial**
- **Forma**: Decrecimiento exponencial desde el origen
- **Parámetro**: λ (tasa)
- **Aplicaciones**: Tiempo entre eventos, duraciones
- **Identificación**: Asimetría muy positiva + inicio en cero

#### **3. Distribución Uniforme**
- **Forma**: Rectangular, frecuencias constantes
- **Parámetros**: a (mínimo) y b (máximo)
- **Aplicaciones**: Números aleatorios, procesos controlados
- **Identificación**: Curtosis muy negativa + forma plana

#### **4. Distribución Log-Normal**
- **Forma**: Asimétrica positiva con cola larga
- **Origen**: Variable cuyo logaritmo es normal
- **Aplicaciones**: Ingresos, tamaños de partículas, precios
- **Identificación**: Fuerte asimetría positiva + log(X) ≈ Normal

#### **5. Distribución Bimodal**
- **Forma**: Dos picos distintivos
- **Origen**: Mezcla de dos distribuciones
- **Aplicaciones**: Mezcla de poblaciones, procesos duales
- **Identificación**: Dos modas claras + valle intermedio

### **🎯 Casos Especiales en Datos Reales**

#### **Distribuciones Truncadas**
- **Causa**: Límites físicos o artificiales en la medición
- **Ejemplos**: Calificaciones (0-5), edades en estudios específicos
- **Identificación**: Corte abrupto en uno o ambos extremos

#### **Distribuciones Contaminadas**
- **Causa**: Mezcla de datos de diferentes fuentes
- **Ejemplos**: Mediciones con errores sistemáticos
- **Identificación**: Forma irregular, múltiples modas

#### **Distribuciones con Gaps**
- **Causa**: Ausencia natural de ciertos valores
- **Ejemplos**: Datos discretos agrupados, mediciones con precisión limitada
- **Identificación**: Espacios vacíos regulares o irregulares

## 🎯 **PARTE 7: Criterios de Calidad para Análisis de Frecuencias**

### **✅ Criterios de una Buena Tabla de Frecuencias**

1. **Completitud**: Todos los datos incluidos, ninguno excluido
2. **Exclusividad**: Cada dato pertenece a exactamente un intervalo
3. **Suficiencia**: Suficiente detalle para análisis, no exceso
4. **Interpretabilidad**: Límites y categorías claros y lógicos
5. **Robustez**: Resultados estables ante pequeños cambios

### **✅ Criterios de un Buen Histograma**

1. **Claridad Visual**: Fácil de leer e interpretar
2. **Proporcionalidad**: Áreas proporcionales a frecuencias
3. **Escala Apropiada**: Rango y graduación adecuados
4. **Etiquetado Completo**: Títulos, ejes y unidades claramente marcados
5. **Contexto**: Información suficiente para interpretación independiente

### **⚠️ Señales de Alerta**

#### **En Datos:**
- **Gaps inexplicables**: Pueden indicar problemas de medición
- **Valores imposibles**: Requieren limpieza de datos
- **Concentración extrema**: Pueden indicar sesgo en recolección

#### **En Histogramas:**
- **Demasiada granularidad**: Dificulta ver patrones
- **Insuficiente detalle**: Oculta características importantes
- **Escalas engañosas**: Pueden crear impresiones falsas

## 🏷️ Análisis de Variable Categórica: Especies

````python
# PASO A PASO: Tabla de frecuencias para variable categórica

print("=" * 60)
print("🌸 ANÁLISIS DE FRECUENCIAS - ESPECIES DE IRIS")
print("=" * 60)

# Paso 1: Seleccionar la variable
especie = iris_df['species_name']
print(f"Variable analizada: {especie.name}")
print(f"Tipo de datos: {especie.dtype}")
print(f"Valores únicos: {list(especie.unique())}")

# Paso 2: Calcular frecuencias absolutas
freq_absoluta = especie.value_counts().sort_index()
print(f"\n📊 Frecuencias Absolutas:")
print(freq_absoluta)

# Paso 3: Calcular frecuencias relativas
total_datos = len(especie)
freq_relativa = freq_absoluta / total_datos
print(f"\n📈 Frecuencias Relativas:")
print(freq_relativa.round(4))

# Paso 4: Calcular porcentajes
freq_porcentaje = freq_relativa * 100
print(f"\n📋 Porcentajes:")
print(freq_porcentaje.round(1))

# Paso 5: Tabla completa
tabla_especies = pd.DataFrame({
    'Especie': freq_absoluta.index,
    'Frecuencia_Absoluta': freq_absoluta.values,
    'Frecuencia_Relativa': freq_relativa.values.round(4),
    'Porcentaje': freq_porcentaje.values.round(1)
})

print(f"\n🗂️ TABLA COMPLETA DE FRECUENCIAS:")
print(tabla_especies)

# Verificación
print(f"\n✅ Verificación:")
print(f"Suma frecuencias absolutas: {tabla_especies['Frecuencia_Absoluta'].sum()}")
print(f"Suma porcentajes: {tabla_especies['Porcentaje'].sum():.1f}%")
````

````python
# Crear visualizaciones para especies
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Análisis de Frecuencias - Especies de Iris', 
             fontsize=16, fontweight='bold')

# Gráfico de barras
axes[0].bar(tabla_especies['Especie'], tabla_especies['Frecuencia_Absoluta'],
           color=['lightcoral', 'lightblue', 'lightgreen'], alpha=0.8)
axes[0].set_title('Frecuencias Absolutas')
axes[0].set_ylabel('Frecuencia')
axes[0].set_xlabel('Especie')

# Añadir valores en las barras
for i, v in enumerate(tabla_especies['Frecuencia_Absoluta']):
    axes[0].text(i, v + 0.5, str(v), ha='center', va='bottom', fontweight='bold')

# Gráfico de barras horizontales
axes[1].barh(tabla_especies['Especie'], tabla_especies['Porcentaje'],
            color=['lightcoral', 'lightblue', 'lightgreen'], alpha=0.8)
axes[1].set_title('Distribución Porcentual')
axes[1].set_xlabel('Porcentaje (%)')
axes[1].set_ylabel('Especie')

# Gráfico de torta
axes[2].pie(tabla_especies['Frecuencia_Absoluta'], 
           labels=tabla_especies['Especie'],
           autopct='%1.1f%%', startangle=90,
           colors=['lightcoral', 'lightblue', 'lightgreen'])
axes[2].set_title('Distribución por Especies')

plt.tight_layout()
plt.show()
````

## 🔢 Análisis de Variable Numérica: Longitud de Sépalo

````python
print("\n" + "=" * 70)
print("📏 ANÁLISIS DE FRECUENCIAS - LONGITUD DE SÉPALO")
print("=" * 70)

# Paso 1: Seleccionar variable numérica
longitud_sepalo = iris_df['sepal length (cm)']
print(f"Variable: {longitud_sepalo.name}")
print(f"Número de datos: {len(longitud_sepalo)}")
print(f"Rango: {longitud_sepalo.min():.2f} - {longitud_sepalo.max():.2f} cm")
print(f"Media: {longitud_sepalo.mean():.2f} cm")

# Paso 2: Determinar número de intervalos (Regla de Sturges)
n = len(longitud_sepalo)
k_sturges = int(1 + 3.322 * np.log10(n))
print(f"\n🔢 Cálculo de intervalos:")
print(f"Número de datos: {n}")
print(f"Intervalos según Sturges: {k_sturges}")

# Paso 3: Calcular amplitud de intervalos
rango = longitud_sepalo.max() - longitud_sepalo.min()
amplitud = rango / k_sturges
print(f"Rango de datos: {rango:.3f} cm")
print(f"Amplitud por intervalo: {amplitud:.3f} cm")

# Paso 4: Crear intervalos
bins = np.linspace(longitud_sepalo.min(), longitud_sepalo.max(), k_sturges + 1)
print(f"\n📊 Límites de intervalos:")
for i, limite in enumerate(bins):
    print(f"  {i}: {limite:.3f} cm")
````

### 📋 Construcción de Tabla de Frecuencias Numérica

````python
# Crear tabla de frecuencias para variable numérica
print(f"\n📋 CONSTRUCCIÓN DE TABLA DE FRECUENCIAS:")

# Calcular frecuencias con numpy
freq_abs, bin_edges = np.histogram(longitud_sepalo, bins=k_sturges)

# Crear etiquetas de intervalos
etiquetas_intervalos = []
for i in range(len(bin_edges) - 1):
    etiqueta = f"[{bin_edges[i]:.2f}, {bin_edges[i+1]:.2f})"
    etiquetas_intervalos.append(etiqueta)

# Calcular estadísticas adicionales
freq_rel = freq_abs / n
freq_porcentaje = freq_rel * 100
freq_acumulada = np.cumsum(freq_abs)
freq_porcentaje_acum = np.cumsum(freq_porcentaje)
puntos_medios = (bin_edges[:-1] + bin_edges[1:]) / 2

# Crear tabla completa
tabla_numerica = pd.DataFrame({
    'Intervalo': etiquetas_intervalos,
    'Límite_Inferior': np.round(bin_edges[:-1], 3),
    'Límite_Superior': np.round(bin_edges[1:], 3),
    'Punto_Medio': np.round(puntos_medios, 3),
    'Frecuencia_Absoluta': freq_abs,
    'Frecuencia_Relativa': np.round(freq_rel, 4),
    'Porcentaje': np.round(freq_porcentaje, 1),
    'Frecuencia_Acumulada': freq_acumulada,
    'Porcentaje_Acumulado': np.round(freq_porcentaje_acum, 1)
})

print("🗂️ TABLA COMPLETA DE FRECUENCIAS - LONGITUD DE SÉPALO:")
print(tabla_numerica)

# Verificaciones
print(f"\n✅ Verificaciones:")
print(f"Suma frecuencias absolutas: {tabla_numerica['Frecuencia_Absoluta'].sum()}")
print(f"Suma frecuencias relativas: {tabla_numerica['Frecuencia_Relativa'].sum():.4f}")
print(f"Último valor acumulado: {tabla_numerica['Frecuencia_Acumulada'].iloc[-1]}")
````

# ⚙️ Función Automatizada para Análisis de Frecuencias

````python
def analizar_frecuencias_completo(serie, n_bins=None, titulo="Análisis de Frecuencias"):
    """
    Función completa para análisis de frecuencias
    
    Parámetros:
    serie: pandas Series con los datos
    n_bins: número de intervalos (None para usar Sturges)
    titulo: título del análisis
    
    Retorna:
    DataFrame con tabla de frecuencias y bins utilizados
    """
    
    print(f"\n{'='*15} {titulo} {'='*15}")
    
    # Información básica
    n = len(serie)
    print(f"📊 Variable: {serie.name}")
    print(f"📈 Datos: {n} observaciones")
    print(f"📏 Rango: {serie.min():.3f} - {serie.max():.3f}")
    print(f"📍 Media: {serie.mean():.3f}")
    
    # Determinar bins
    if n_bins is None:
        n_bins = int(1 + 3.322 * np.log10(n))
    
    print(f"🔢 Intervalos: {n_bins} (Regla de Sturges)")
    
    # Crear tabla de frecuencias
    freq_abs, bin_edges = np.histogram(serie, bins=n_bins)
    
    # Calcular estadísticas
    intervalos = [f"[{bin_edges[i]:.2f}, {bin_edges[i+1]:.2f})" 
                 for i in range(len(bin_edges) - 1)]
    puntos_medios = (bin_edges[:-1] + bin_edges[1:]) / 2
    freq_rel = freq_abs / n
    freq_pct = freq_rel * 100
    freq_acum = np.cumsum(freq_abs)
    freq_pct_acum = np.cumsum(freq_pct)
    
    # Crear DataFrame
    tabla = pd.DataFrame({
        'Intervalo': intervalos,
        'Punto_Medio': np.round(puntos_medios, 3),
        'Frecuencia': freq_abs,
        'Freq_Relativa': np.round(freq_rel, 4),
        'Porcentaje': np.round(freq_pct, 1),
        'Freq_Acumulada': freq_acum,
        'Pct_Acumulado': np.round(freq_pct_acum, 1)
    })
    
    print(f"\n📋 TABLA DE FRECUENCIAS:")
    print(tabla)
    
    # Interpretaciones
    idx_max = tabla['Frecuencia'].idxmax()
    intervalo_max = tabla.loc[idx_max, 'Intervalo']
    freq_max = tabla.loc[idx_max, 'Frecuencia']
    
    print(f"\n🔍 INTERPRETACIONES:")
    print(f"• Intervalo más frecuente: {intervalo_max} ({freq_max} casos)")
    print(f"• Representa el {tabla.loc[idx_max, 'Porcentaje']:.1f}% de los datos")
    
    return tabla, bin_edges

# Usar la función
tabla_automatica, bins_auto = analizar_frecuencias_completo(
    iris_df['petal length (cm)'], 
    titulo="LONGITUD DE PÉTALO"
)
````
