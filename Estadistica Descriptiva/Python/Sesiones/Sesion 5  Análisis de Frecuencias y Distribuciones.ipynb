## üìä **PARTE 4: Construcci√≥n √ìptima de Intervalos de Clase**

### **üìè Paso 1: Determinar el N√∫mero de Intervalos**

#### **Regla de Sturges (La m√°s com√∫n)**
```
k = 1 + 3.322 √ó log‚ÇÅ‚ÇÄ(n)
```
- **Aplicaci√≥n**: Distribuciones aproximadamente normales
- **Ventaja**: Simple y ampliamente aceptada
- **Limitaci√≥n**: Puede subestimar para distribuciones asim√©tricas

#### **Regla de la Ra√≠z Cuadrada**
```
k = ‚àön
```
- **Aplicaci√≥n**: Alternativa simple
- **Ventaja**: F√°cil de calcular mentalmente
- **Uso**: An√°lisis exploratorio r√°pido

#### **Regla de Freedman-Diaconis (Robusta)**
```
Amplitud = 2 √ó IQR √ó n^(-1/3)
k = (M√°x - M√≠n) / Amplitud
```
- **Aplicaci√≥n**: Datos con outliers o distribuciones asim√©tricas
- **Ventaja**: Robusta ante valores extremos
- **Recomendaci√≥n**: Mejor para datos reales

#### **Regla de Scott (Basada en desviaci√≥n est√°ndar)**
```
Amplitud = 3.49 √ó œÉ √ó n^(-1/3)
k = (M√°x - M√≠n) / Amplitud
```
- **Aplicaci√≥n**: Distribuciones cercanas a la normal
- **Ventaja**: Considera la variabilidad de los datos

### **üìä Comparaci√≥n de Reglas**

| Regla | Fortalezas | Debilidades | Cu√°ndo Usar |
|-------|------------|-------------|-------------|
| **Sturges** | Simple, est√°ndar | Subestima para n grande | Distribuciones normales |
| **Ra√≠z Cuadrada** | Muy simple | Demasiado simple | Exploraci√≥n r√°pida |
| **Freedman-Diaconis** | Robusta, adaptativa | M√°s compleja | Datos con outliers |
| **Scott** | Considera variabilidad | Sensible a outliers | Datos aproximadamente normales |

### **üí° Ejemplo Comparativo**

- **M√≠nimo**: $15,000, **M√°ximo**: $180,000

**Recomendaci√≥n**: Para datos de ingresos (t√≠picamente asim√©tricos), usar Freedman-Diaconis.

### **üìè Paso 2: Definir L√≠mites de Intervalos**

#### **Principios Fundamentales:**
1. **Cobertura Completa**: Todos los datos deben estar incluidos

#### **Notaci√≥n de Intervalos:**

- **Intervalos intermedios**: (l√≠mite_{i-1}, l√≠mite_i]

## üìä **PARTE 5: Histogramas - Visualizaci√≥n de Distribuciones**

### **üìà ¬øQu√© es un Histograma?**

> Un **histograma** es una representaci√≥n gr√°fica de la distribuci√≥n de frecuencias de una variable continua, usando barras rectangulares cuya √°rea representa la frecuencia de cada intervalo.

### **üîß Componentes del Histograma**

#### **Ejes:**
- **Eje X (horizontal)**: Valores de la variable (intervalos de clase)
- **Eje Y (vertical)**: Frecuencia (absoluta, relativa, o densidad)

#### **Barras:**
- **Base**: Amplitud del intervalo de clase
- **Altura**: Frecuencia del intervalo
- **√Årea**: Proporcional a la frecuencia (importante para intervalos desiguales)

### **üìä Tipos de Histogramas**

#### **1. Histograma de Frecuencias Absolutas**
- **Eje Y**: N√∫mero de observaciones
- **Uso**: Mostrar conteos reales
- **Limitaci√≥n**: Dificulta comparaci√≥n entre datasets de diferente tama√±o

#### **2. Histograma de Frecuencias Relativas**
- **Eje Y**: Proporci√≥n o porcentaje
- **Uso**: Comparaci√≥n entre datasets
- **Ventaja**: Suma total = 1.0 (100%)

#### **3. Histograma de Densidad**
- **Eje Y**: Densidad = frecuencia / amplitud
- **Uso**: Intervalos de diferente amplitud
- **Propiedad**: √Årea total = 1.0

### **üéØ Interpretaci√≥n de Formas**

#### **Distribuci√≥n Sim√©trica (Normal)**
```
      üìä
     /|\
    / | \
   /  |  \
  /   |   \
 /_______|_______\
```
- **Caracter√≠sticas**: Media ‚âà Mediana ‚âà Moda
- **Ejemplos**: Altura, peso, errores de medici√≥n
- **Implicaciones**: Aplicable regla emp√≠rica (68-95-99.7)

#### **Distribuci√≥n Asim√©trica Positiva**
```
üìä
|\
| \
|  \
|   \____
|_______\
```
- **Caracter√≠sticas**: Media > Mediana > Moda
- **Ejemplos**: Ingresos, precios de casas, tiempo de vida
- **Implicaciones**: Usar mediana como medida central

#### **Distribuci√≥n Asim√©trica Negativa**
```
    üìä
   /|
  / |
 /  |
/   |

```
- **Caracter√≠sticas**: Moda > Mediana > Media
- **Ejemplos**: Edad al morir, calificaciones en examen f√°cil
- **Implicaciones**: Considerar transformaciones

#### **Distribuci√≥n Bimodal**
```
üìä       üìä

```
- **Caracter√≠sticas**: Dos picos claramente diferenciados
- **Ejemplos**: Mezcla de dos poblaciones
- **Implicaciones**: Analizar por separado cada grupo

#### **Distribuci√≥n Uniforme**
```
üìäüìäüìäüìäüìä
|  |  |  |  |
|__|__|__|__|
```
- **Caracter√≠sticas**: Frecuencias aproximadamente iguales
- **Ejemplos**: N√∫meros aleatorios, algunos procesos industriales
- **Implicaciones**: Todos los valores igualmente probables

### **‚ö†Ô∏è Problemas Comunes con Histogramas**

#### **1. N√∫mero Inadecuado de Intervalos**
- **Muy pocos**: P√©rdida de detalle, sobre-suavizado
- **Muy muchos**: Ruido excesivo, dificulta interpretaci√≥n

#### **2. L√≠mites Inapropiados**
- **L√≠mites artificiales**: Crear picos falsos
- **Intervalos desiguales**: Distorsionar la interpretaci√≥n

#### **3. Mala Selecci√≥n de Escala**
- **Escala Y inapropiada**: Exagerar o minimizar diferencias
- **Supresi√≥n del cero**: Crear impresiones falsas

## üìä **PARTE 6: Identificaci√≥n de Patrones Distributivos**

### **üîç An√°lisis Sistem√°tico de Distribuciones**

#### **Paso 1: Evaluaci√≥n Visual**
1. **Forma general**: ¬øSim√©trica, asim√©trica, multimodal?
2. **Concentraci√≥n**: ¬øDatos concentrados o dispersos?
3. **Colas**: ¬øPesadas, ligeras, una m√°s larga que otra?
4. **Gaps**: ¬øHay espacios sin datos?
5. **Outliers**: ¬øValores extremos aislados?

#### **Paso 2: M√©tricas Cuantitativas**
1. **Asimetr√≠a (Skewness)**: Confirma observaciones visuales
2. **Curtosis**: Eval√∫a concentraci√≥n y comportamiento de colas
3. **Coeficiente de variaci√≥n**: Eval√∫a dispersi√≥n relativa

#### **Paso 3: Comparaci√≥n con Distribuciones Te√≥ricas**
1. **Normal**: ¬øSe ajusta a la campana de Gauss?
2. **Exponencial**: ¬øDecrecimiento exponencial?
3. **Uniforme**: ¬øFrecuencias relativamente constantes?
4. **Log-normal**: ¬øAsim√©trica que se normaliza con logaritmo?

### **üìö Cat√°logo de Distribuciones Comunes**

#### **1. Distribuci√≥n Normal (Gaussiana)**
- **Forma**: Campana sim√©trica
- **Par√°metros**: Œº (media) y œÉ (desviaci√≥n est√°ndar)
- **Aplicaciones**: Errores, medidas biol√≥gicas, promedios
- **Identificaci√≥n**: Simetr√≠a + curtosis ‚âà 0

#### **2. Distribuci√≥n Exponencial**
- **Forma**: Decrecimiento exponencial desde el origen
- **Par√°metro**: Œª (tasa)
- **Aplicaciones**: Tiempo entre eventos, duraciones
- **Identificaci√≥n**: Asimetr√≠a muy positiva + inicio en cero

#### **3. Distribuci√≥n Uniforme**
- **Forma**: Rectangular, frecuencias constantes
- **Par√°metros**: a (m√≠nimo) y b (m√°ximo)
- **Aplicaciones**: N√∫meros aleatorios, procesos controlados
- **Identificaci√≥n**: Curtosis muy negativa + forma plana

#### **4. Distribuci√≥n Log-Normal**
- **Forma**: Asim√©trica positiva con cola larga
- **Origen**: Variable cuyo logaritmo es normal
- **Aplicaciones**: Ingresos, tama√±os de part√≠culas, precios
- **Identificaci√≥n**: Fuerte asimetr√≠a positiva + log(X) ‚âà Normal

#### **5. Distribuci√≥n Bimodal**
- **Forma**: Dos picos distintivos
- **Origen**: Mezcla de dos distribuciones
- **Aplicaciones**: Mezcla de poblaciones, procesos duales
- **Identificaci√≥n**: Dos modas claras + valle intermedio

### **üéØ Casos Especiales en Datos Reales**

#### **Distribuciones Truncadas**
- **Causa**: L√≠mites f√≠sicos o artificiales en la medici√≥n
- **Ejemplos**: Calificaciones (0-5), edades en estudios espec√≠ficos
- **Identificaci√≥n**: Corte abrupto en uno o ambos extremos

#### **Distribuciones Contaminadas**
- **Causa**: Mezcla de datos de diferentes fuentes
- **Ejemplos**: Mediciones con errores sistem√°ticos
- **Identificaci√≥n**: Forma irregular, m√∫ltiples modas

#### **Distribuciones con Gaps**
- **Causa**: Ausencia natural de ciertos valores
- **Ejemplos**: Datos discretos agrupados, mediciones con precisi√≥n limitada
- **Identificaci√≥n**: Espacios vac√≠os regulares o irregulares

## 1. Carga y Exploraci√≥n del Dataset Iris

````python
# 1. Dataset Iris (Variables categ√≥ricas y num√©ricas)
# Carga del dataset Iris
from sklearn.datasets import load_iris

# Cargar datos
iris_data = load_iris()
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target
iris_df['species_name'] = iris_df['species'].map({
    0: 'setosa', 
    1: 'versicolor', 
    2: 'virginica'
})

print("üìä Dataset Iris cargado")
print(f"Dimensiones: {iris_df.shape}")
print("\nüîç Primeras 5 filas:")
print(iris_df.head())
````

````python
# Informaci√≥n b√°sica del dataset
print("üîç Informaci√≥n b√°sica del dataset:")
print(iris_df.info())
print("\nüìà Estad√≠sticas descriptivas:")
print(iris_df.describe())
````

## üè∑Ô∏è An√°lisis de Variable Categ√≥rica: Especies

````python
# PASO A PASO: Tabla de frecuencias para variable categ√≥rica

print("=" * 60)
print("üå∏ AN√ÅLISIS DE FRECUENCIAS - ESPECIES DE IRIS")
print("=" * 60)

# Paso 1: Seleccionar la variable
especie = iris_df['species_name']
print(f"Variable analizada: {especie.name}")
print(f"Tipo de datos: {especie.dtype}")
print(f"Valores √∫nicos: {list(especie.unique())}")

# Paso 2: Calcular frecuencias absolutas
freq_absoluta = especie.value_counts().sort_index()
print(f"\nüìä Frecuencias Absolutas:")
print(freq_absoluta)

# Paso 3: Calcular frecuencias relativas
total_datos = len(especie)
freq_relativa = freq_absoluta / total_datos
print(f"\nüìà Frecuencias Relativas:")
print(freq_relativa.round(4))

# Paso 4: Calcular porcentajes
freq_porcentaje = freq_relativa * 100
print(f"\nüìã Porcentajes:")
print(freq_porcentaje.round(1))

# Paso 5: Tabla completa
tabla_especies = pd.DataFrame({
    'Especie': freq_absoluta.index,
    'Frecuencia_Absoluta': freq_absoluta.values,
    'Frecuencia_Relativa': freq_relativa.values.round(4),
    'Porcentaje': freq_porcentaje.values.round(1)
})

print(f"\nüóÇÔ∏è TABLA COMPLETA DE FRECUENCIAS:")
print(tabla_especies)

# Verificaci√≥n
print(f"\n‚úÖ Verificaci√≥n:")
print(f"Suma frecuencias absolutas: {tabla_especies['Frecuencia_Absoluta'].sum()}")
print(f"Suma porcentajes: {tabla_especies['Porcentaje'].sum():.1f}%")
````

### üìä Visualizaci√≥n de Variable Categ√≥rica

````python
# Crear visualizaciones para especies
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('An√°lisis de Frecuencias - Especies de Iris', 
             fontsize=16, fontweight='bold')

# Gr√°fico de barras
axes[0].bar(tabla_especies['Especie'], tabla_especies['Frecuencia_Absoluta'],
           color=['lightcoral', 'lightblue', 'lightgreen'], alpha=0.8)
axes[0].set_title('Frecuencias Absolutas')
axes[0].set_ylabel('Frecuencia')
axes[0].set_xlabel('Especie')

# A√±adir valores en las barras
for i, v in enumerate(tabla_especies['Frecuencia_Absoluta']):
    axes[0].text(i, v + 0.5, str(v), ha='center', va='bottom', fontweight='bold')

# Gr√°fico de barras horizontales
axes[1].barh(tabla_especies['Especie'], tabla_especies['Porcentaje'],
            color=['lightcoral', 'lightblue', 'lightgreen'], alpha=0.8)
axes[1].set_title('Distribuci√≥n Porcentual')
axes[1].set_xlabel('Porcentaje (%)')
axes[1].set_ylabel('Especie')

# Gr√°fico de torta
axes[2].pie(tabla_especies['Frecuencia_Absoluta'], 
           labels=tabla_especies['Especie'],
           autopct='%1.1f%%', startangle=90,
           colors=['lightcoral', 'lightblue', 'lightgreen'])
axes[2].set_title('Distribuci√≥n por Especies')

plt.tight_layout()
plt.show()
````

## üî¢ An√°lisis de Variable Num√©rica: Longitud de S√©palo

````python
print("\n" + "=" * 70)
print("üìè AN√ÅLISIS DE FRECUENCIAS - LONGITUD DE S√âPALO")
print("=" * 70)

# Paso 1: Seleccionar variable num√©rica
longitud_sepalo = iris_df['sepal length (cm)']
print(f"Variable: {longitud_sepalo.name}")
print(f"N√∫mero de datos: {len(longitud_sepalo)}")
print(f"Rango: {longitud_sepalo.min():.2f} - {longitud_sepalo.max():.2f} cm")
print(f"Media: {longitud_sepalo.mean():.2f} cm")

# Paso 2: Determinar n√∫mero de intervalos (Regla de Sturges)
n = len(longitud_sepalo)
k_sturges = int(1 + 3.322 * np.log10(n))
print(f"\nüî¢ C√°lculo de intervalos:")
print(f"N√∫mero de datos: {n}")
print(f"Intervalos seg√∫n Sturges: {k_sturges}")

# Paso 3: Calcular amplitud de intervalos
rango = longitud_sepalo.max() - longitud_sepalo.min()
amplitud = rango / k_sturges
print(f"Rango de datos: {rango:.3f} cm")
print(f"Amplitud por intervalo: {amplitud:.3f} cm")

# Paso 4: Crear intervalos
bins = np.linspace(longitud_sepalo.min(), longitud_sepalo.max(), k_sturges + 1)
print(f"\nüìä L√≠mites de intervalos:")
for i, limite in enumerate(bins):
    print(f"  {i}: {limite:.3f} cm")
````

### üìã Construcci√≥n de Tabla de Frecuencias Num√©rica

````python
# Crear tabla de frecuencias para variable num√©rica
print(f"\nüìã CONSTRUCCI√ìN DE TABLA DE FRECUENCIAS:")

# Calcular frecuencias con numpy
freq_abs, bin_edges = np.histogram(longitud_sepalo, bins=k_sturges)

# Crear etiquetas de intervalos
etiquetas_intervalos = []
for i in range(len(bin_edges) - 1):
    etiqueta = f"[{bin_edges[i]:.2f}, {bin_edges[i+1]:.2f})"
    etiquetas_intervalos.append(etiqueta)

# Calcular estad√≠sticas adicionales
freq_rel = freq_abs / n
freq_porcentaje = freq_rel * 100
freq_acumulada = np.cumsum(freq_abs)
freq_porcentaje_acum = np.cumsum(freq_porcentaje)
puntos_medios = (bin_edges[:-1] + bin_edges[1:]) / 2

# Crear tabla completa
tabla_numerica = pd.DataFrame({
    'Intervalo': etiquetas_intervalos,
    'L√≠mite_Inferior': np.round(bin_edges[:-1], 3),
    'L√≠mite_Superior': np.round(bin_edges[1:], 3),
    'Punto_Medio': np.round(puntos_medios, 3),
    'Frecuencia_Absoluta': freq_abs,
    'Frecuencia_Relativa': np.round(freq_rel, 4),
    'Porcentaje': np.round(freq_porcentaje, 1),
    'Frecuencia_Acumulada': freq_acumulada,
    'Porcentaje_Acumulado': np.round(freq_porcentaje_acum, 1)
})

print("üóÇÔ∏è TABLA COMPLETA DE FRECUENCIAS - LONGITUD DE S√âPALO:")
print(tabla_numerica)

# Verificaciones
print(f"\n‚úÖ Verificaciones:")
print(f"Suma frecuencias absolutas: {tabla_numerica['Frecuencia_Absoluta'].sum()}")
print(f"Suma frecuencias relativas: {tabla_numerica['Frecuencia_Relativa'].sum():.4f}")
print(f"√öltimo valor acumulado: {tabla_numerica['Frecuencia_Acumulada'].iloc[-1]}")
````

### üìà Creaci√≥n de Histogramas

````python
# Crear histogramas profesionales
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('An√°lisis de Distribuci√≥n - Longitud de S√©palo', 
             fontsize=16, fontweight='bold')

# Histograma b√°sico
axes[0, 0].hist(longitud_sepalo, bins=bin_edges, alpha=0.7, 
                color='skyblue', edgecolor='black')
axes[0, 0].set_title('Histograma de Frecuencias')
axes[0, 0].set_xlabel('Longitud de S√©palo (cm)')
axes[0, 0].set_ylabel('Frecuencia Absoluta')
axes[0, 0].grid(True, alpha=0.3)

# Histograma de densidad
axes[0, 1].hist(longitud_sepalo, bins=bin_edges, alpha=0.7, 
                color='lightgreen', edgecolor='black', density=True)
axes[0, 1].set_title('Histograma de Densidad')
axes[0, 1].set_xlabel('Longitud de S√©palo (cm)')
axes[0, 1].set_ylabel('Densidad')
axes[0, 1].grid(True, alpha=0.3)

# Gr√°fico de barras desde tabla
axes[1, 0].bar(range(len(puntos_medios)), tabla_numerica['Frecuencia_Absoluta'],
               color='coral', alpha=0.7, edgecolor='black')
axes[1, 0].set_title('Frecuencias por Intervalo')
axes[1, 0].set_xlabel('Intervalo')
axes[1, 0].set_ylabel('Frecuencia')
axes[1, 0].set_xticks(range(len(puntos_medios)))
axes[1, 0].set_xticklabels([f'{pm:.2f}' for pm in puntos_medios], rotation=45)

# A√±adir valores en las barras
for i, v in enumerate(tabla_numerica['Frecuencia_Absoluta']):
    axes[1, 0].text(i, v + 0.2, str(v), ha='center', va='bottom', fontweight='bold')

# Frecuencias acumuladas
axes[1, 1].plot(puntos_medios, tabla_numerica['Frecuencia_Acumulada'], 
                'bo-', linewidth=2, markersize=6, color='darkblue')
axes[1, 1].fill_between(puntos_medios, tabla_numerica['Frecuencia_Acumulada'], 
                        alpha=0.3, color='darkblue')
axes[1, 1].set_title('Frecuencias Acumuladas')
axes[1, 1].set_xlabel('Longitud de S√©palo (cm)')
axes[1, 1].set_ylabel('Frecuencia Acumulada')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

# ‚öôÔ∏è Funci√≥n Automatizada para An√°lisis de Frecuencias

````python
def analizar_frecuencias_completo(serie, n_bins=None, titulo="An√°lisis de Frecuencias"):
    """
    Funci√≥n completa para an√°lisis de frecuencias
    
    Par√°metros:
    serie: pandas Series con los datos
    n_bins: n√∫mero de intervalos (None para usar Sturges)
    titulo: t√≠tulo del an√°lisis
    
    Retorna:
    DataFrame con tabla de frecuencias y bins utilizados
    """
    
    print(f"\n{'='*15} {titulo} {'='*15}")
    
    # Informaci√≥n b√°sica
    n = len(serie)
    print(f"üìä Variable: {serie.name}")
    print(f"üìà Datos: {n} observaciones")
    print(f"üìè Rango: {serie.min():.3f} - {serie.max():.3f}")
    print(f"üìç Media: {serie.mean():.3f}")
    
    # Determinar bins
    if n_bins is None:
        n_bins = int(1 + 3.322 * np.log10(n))
    
    print(f"üî¢ Intervalos: {n_bins} (Regla de Sturges)")
    
    # Crear tabla de frecuencias
    freq_abs, bin_edges = np.histogram(serie, bins=n_bins)
    
    # Calcular estad√≠sticas
    intervalos = [f"[{bin_edges[i]:.2f}, {bin_edges[i+1]:.2f})" 
                 for i in range(len(bin_edges) - 1)]
    puntos_medios = (bin_edges[:-1] + bin_edges[1:]) / 2
    freq_rel = freq_abs / n
    freq_pct = freq_rel * 100
    freq_acum = np.cumsum(freq_abs)
    freq_pct_acum = np.cumsum(freq_pct)
    
    # Crear DataFrame
    tabla = pd.DataFrame({
        'Intervalo': intervalos,
        'Punto_Medio': np.round(puntos_medios, 3),
        'Frecuencia': freq_abs,
        'Freq_Relativa': np.round(freq_rel, 4),
        'Porcentaje': np.round(freq_pct, 1),
        'Freq_Acumulada': freq_acum,
        'Pct_Acumulado': np.round(freq_pct_acum, 1)
    })
    
    print(f"\nüìã TABLA DE FRECUENCIAS:")
    print(tabla)
    
    # Interpretaciones
    idx_max = tabla['Frecuencia'].idxmax()
    intervalo_max = tabla.loc[idx_max, 'Intervalo']
    freq_max = tabla.loc[idx_max, 'Frecuencia']
    
    print(f"\nüîç INTERPRETACIONES:")
    print(f"‚Ä¢ Intervalo m√°s frecuente: {intervalo_max} ({freq_max} casos)")
    print(f"‚Ä¢ Representa el {tabla.loc[idx_max, 'Porcentaje']:.1f}% de los datos")
    
    return tabla, bin_edges

# Usar la funci√≥n
tabla_automatica, bins_auto = analizar_frecuencias_completo(
    iris_df['petal length (cm)'], 
    titulo="LONGITUD DE P√âTALO"
)
````
